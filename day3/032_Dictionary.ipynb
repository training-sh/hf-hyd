# Dictionaries

A dictionary stores data as key-value pairs. Each key identifies a value, making dictionaries suitable for product records, order details, inventory lookup, configuration, grouping, and aggregation.

## 1. Creating Dictionaries

```python
empty_product = {}
product = {
    "sku": "P100",
    "name": "Laptop",
    "price": 65000.0,
    "in_stock": True,
}

print(empty_product)
print(product)
print(type(product))
print(len(product))
```

## 2. Keys and Values

A key must be unique and hashable. Values can be repeated and may contain any Python object. Assigning the same key again replaces its previous value.

```python
product = {
    "sku": "P200",
    "name": "Mouse",
    "price": 899,
    "price": 799,
}

print(product)
print(product["price"])
```

## 3. Accessing Values

Square brackets return the value for an existing key and raise `KeyError` when the key is absent. `get` returns `None` or a supplied default instead.

```python
order = {"order_id": 1001, "customer": "Asha", "amount": 2499.0}

print(order["customer"])
print(order.get("amount"))
print(order.get("coupon"))
print(order.get("coupon", "No coupon"))

try:
    print(order["status"])
except KeyError as error:
    print(type(error).__name__, error)
```

## 4. Adding and Updating Entries

Assignment adds a missing key or changes the value of an existing key. `update` applies several key-value pairs.

```python
product = {"sku": "P300", "name": "Keyboard", "price": 1499}
product["stock"] = 20
product["price"] = 1399
product.update({"rating": 4.5, "category": "Accessories"})

print(product)
```

## 5. Removing Entries

`pop` removes a key and returns its value. `popitem` removes and returns the most recently inserted pair. `del` removes a known key, and `clear` removes all entries.

```python
product = {"sku": "P400", "name": "Monitor", "price": 12999, "stock": 8}
stock = product.pop("stock")
last_pair = product.popitem()
del product["name"]

print(stock)
print(last_pair)
print(product)

product.clear()
print(product)
```

## 6. Membership

The `in` operator checks dictionary keys, not values.

```python
order = {"order_id": 1001, "status": "paid", "amount": 2499.0}

print("status" in order)
print("customer" not in order)
print("paid" in order)
print("paid" in order.values())
```

## 7. Iterating Over a Dictionary

Direct iteration produces keys. The key can then be used to retrieve its value.

```python
product = {"sku": "P100", "name": "Laptop", "price": 65000}

for key in product:
    print(key, product[key])
```

## 8. Keys, Values, and Items Views

`keys`, `values`, and `items` return dynamic view objects. The views reflect later changes to the dictionary.

```python
product = {"sku": "P200", "name": "Mouse", "price": 799}
product_keys = product.keys()
product_values = product.values()
product_items = product.items()

product["stock"] = 35

print(product_keys)
print(product_values)
print(product_items)
print(list(product_items))
```

## 9. Iterating Over Key-Value Pairs

Each value from `items()` is a two-item tuple that can be unpacked.

```python
order = {"order_id": 1002, "customer": "Bala", "amount": 12999, "status": "shipped"}

for field, value in order.items():
    print(f"{field:<10}: {value}")
```

## 10. Insertion Order

Dictionaries preserve insertion order. Updating a value does not move its key, while deleting and reinserting a key places it at the end.

```python
product = {}
product["sku"] = "P100"
product["name"] = "Laptop"
product["price"] = 65000
product["name"] = "Gaming Laptop"

print(list(product))

del product["sku"]
product["sku"] = "P100"
print(list(product))
```

## 11. Creating With `dict`

`dict` accepts keyword arguments, key-value pairs, or another mapping.

```python
product_a = dict(sku="P100", name="Laptop", price=65000)
product_b = dict([("sku", "P200"), ("name", "Mouse"), ("price", 799)])
product_c = dict(product_b)

print(product_a)
print(product_b)
print(product_c)
```

## 12. Creating a Dictionary With `zip`

```python
fields = ["order_id", "customer", "amount", "status"]
values = [1001, "Asha", 2499.0, "paid"]
order = dict(zip(fields, values))

print(order)
```

## 13. `fromkeys`

`fromkeys` creates keys with one shared default value. Immutable defaults are safe. A comprehension should be used when every key needs an independent mutable value.

```python
warehouses = ["Chennai", "Bengaluru", "Hyderabad"]
stock_status = dict.fromkeys(warehouses, "pending")
print(stock_status)

shared_bins = dict.fromkeys(warehouses, [])
shared_bins["Chennai"].append("P100")
print(shared_bins)

independent_bins = {warehouse: [] for warehouse in warehouses}
independent_bins["Chennai"].append("P100")
print(independent_bins)
```

## 14. `setdefault`

`setdefault` returns the existing value. When the key is missing, it inserts and returns the supplied default.

```python
order = {"order_id": 1001, "status": "paid"}
existing_status = order.setdefault("status", "pending")
items = order.setdefault("items", [])
items.append("Laptop")

print(existing_status)
print(order)
```

## 15. Dictionary Comprehensions

A dictionary comprehension can transform, filter, or reverse key-value relationships.

```python
prices = {"Laptop": 65000, "Mouse": 799, "Keyboard": 1499, "Monitor": 12999}
discounted_prices = {product: round(price * 0.90, 2) for product, price in prices.items()}
premium_products = {product: price for product, price in prices.items() if price >= 10000}

print(discounted_prices)
print(premium_products)
```

## 16. Nested Dictionaries

Nested dictionaries represent records containing other named records.

```python
order = {
    "order_id": 1001,
    "customer": {"name": "Asha", "city": "Chennai"},
    "payment": {"method": "UPI", "status": "paid"},
    "shipping": {"provider": "BlueDart", "tracking_id": "BD1001"},
}

print(order["customer"]["name"])
print(order["payment"]["status"])
print(order.get("shipping", {}).get("tracking_id"))
```

## 17. Lists Inside Dictionaries

An order dictionary can contain a changing list of line-item dictionaries.

```python
order = {
    "order_id": 1001,
    "customer": "Asha",
    "items": [
        {"sku": "P100", "quantity": 1, "unit_price": 65000},
        {"sku": "P200", "quantity": 2, "unit_price": 799},
    ],
}

order_total = sum(item["quantity"] * item["unit_price"] for item in order["items"])
print(order_total)
```

## 18. A List of Dictionaries

A list preserves record order, and each dictionary provides named fields.

```python
products = [
    {"sku": "P100", "name": "Laptop", "price": 65000, "stock": 8},
    {"sku": "P200", "name": "Mouse", "price": 799, "stock": 35},
    {"sku": "P300", "name": "Keyboard", "price": 1499, "stock": 0},
]

available_products = [product for product in products if product["stock"] > 0]
print(available_products)
```

## 19. Indexing Records by a Key

A list of records can be converted into a lookup dictionary for direct access by SKU.

```python
products_by_sku = {product["sku"]: product for product in products}

print(products_by_sku["P200"])
print(products_by_sku.get("P999", {"name": "Unknown product"}))
```

## 20. Merging Dictionaries

The union operator creates a new dictionary. For duplicate keys, values from the right side win. `|=` and `update` modify the existing dictionary.

```python
base_product = {"sku": "P100", "name": "Laptop", "price": 65000}
inventory_data = {"stock": 8, "warehouse": "Chennai"}
promotion = {"price": 62000, "campaign": "Weekend Sale"}

complete_product = base_product | inventory_data | promotion
print(complete_product)
print(base_product)

base_product |= inventory_data
print(base_product)
```

## 21. Shallow Copy and Aliasing

Assignment creates another reference to the same dictionary. `copy()` creates a new outer dictionary, but nested mutable values remain shared.

```python
order = {"order_id": 1001, "items": ["Laptop", "Mouse"]}
order_alias = order
order_copy = order.copy()

order_alias["status"] = "paid"
order_copy["items"].append("Keyboard")

print(order)
print(order_copy)
print(order_alias is order)
print(order_copy is order)
```

## 22. Deep Copy

A deep copy recursively copies nested mutable values.

```python
from copy import deepcopy

order = {"order_id": 1001, "items": [{"sku": "P100", "quantity": 1}]}
independent_order = deepcopy(order)
independent_order["items"][0]["quantity"] = 3

print(order)
print(independent_order)
```

## 23. Valid and Invalid Key Types

Strings, numbers, and tuples containing hashable values can be keys. Lists, dictionaries, and sets are mutable and cannot be keys.

```python
inventory = {
    "P100": 12,
    1001: "order lookup",
    ("Chennai", "P200"): 35,
}
print(inventory)

try:
    invalid = {["Chennai", "P200"]: 35}
except TypeError as error:
    print(type(error).__name__, error)
```

## 24. Dictionary Equality

Dictionaries are equal when they contain the same key-value pairs. Insertion order does not affect equality.

```python
product_a = {"sku": "P100", "price": 65000}
product_b = {"price": 65000, "sku": "P100"}
product_c = {"sku": "P100", "price": 62000}

print(product_a == product_b)
print(product_a == product_c)
print(product_a is product_b)
```

## 25. Grouping Records

`setdefault` can accumulate records under a grouping key.

```python
orders = [
    {"order_id": 1001, "status": "paid", "amount": 2499},
    {"order_id": 1002, "status": "shipped", "amount": 12999},
    {"order_id": 1003, "status": "paid", "amount": 799},
    {"order_id": 1004, "status": "cancelled", "amount": 5499},
]

orders_by_status = {}
for order in orders:
    orders_by_status.setdefault(order["status"], []).append(order)

for status, status_orders in orders_by_status.items():
    print(status, [order["order_id"] for order in status_orders])
```

## 26. Counting With a Dictionary

```python
purchased_skus = ["P100", "P200", "P200", "P300", "P200", "P100"]
sku_counts = {}

for sku in purchased_skus:
    sku_counts[sku] = sku_counts.get(sku, 0) + 1

print(sku_counts)
```

## 27. `defaultdict` and `Counter`

`defaultdict` supplies a missing value through a factory. `Counter` specializes dictionary-based counting.

```python
from collections import Counter, defaultdict

revenue_by_status = defaultdict(float)
for order in orders:
    revenue_by_status[order["status"]] += order["amount"]

status_counts = Counter(order["status"] for order in orders)

print(dict(revenue_by_status))
print(status_counts)
print(status_counts.most_common(2))
```

## 28. Sorting Dictionary Data

Sorting a dictionary directly sorts its keys. Sorting `items()` can use either the key or the value.

```python
sales_by_product = {"Laptop": 8, "Mouse": 35, "Keyboard": 20, "Monitor": 12}

print(sorted(sales_by_product))
print(sorted(sales_by_product.items()))
print(sorted(sales_by_product.items(), key=lambda item: item[1], reverse=True))
```

## 29. Safe Mutation During Iteration

The size of a dictionary should not change during direct iteration. Iterate over a copied list of keys or build a new dictionary.

```python
inventory = {"P100": 8, "P200": 0, "P300": 20, "P400": 0}

for sku in list(inventory):
    if inventory[sku] == 0:
        del inventory[sku]

print(inventory)

original_inventory = {"P100": 8, "P200": 0, "P300": 20}
available_inventory = {sku: stock for sku, stock in original_inventory.items() if stock > 0}
print(available_inventory)
```

## 30. Dictionary View Set Operations

Key and item views support set-like operations when their values are hashable.

```python
catalog_fields = {"sku": "P100", "name": "Laptop", "price": 65000}
inventory_fields = {"sku": "P100", "stock": 8, "warehouse": "Chennai"}

common_fields = catalog_fields.keys() & inventory_fields.keys()
all_fields = catalog_fields.keys() | inventory_fields.keys()
catalog_only = catalog_fields.keys() - inventory_fields.keys()

print(common_fields)
print(all_fields)
print(catalog_only)
```

## 31. Dictionary Performance

Key lookup, insertion, update, and deletion are typically constant time on average. Iteration and copying are linear in the number of entries. Dictionaries use additional memory to provide fast key-based access.

```python
large_catalog = {
    f"SKU-{number:05d}": {"price": number * 10, "stock": number % 25}
    for number in range(10_000)
}

product = large_catalog["SKU-07500"]
large_catalog["SKU-10000"] = {"price": 100000, "stock": 10}

print(product)
print(len(large_catalog))
```

## 32. Complete Ecommerce Aggregation

```python
order_lines = [
    {"order_id": 1001, "customer": "Asha", "category": "Electronics", "sku": "P100", "quantity": 1, "unit_price": 65000, "status": "paid"},
    {"order_id": 1001, "customer": "Asha", "category": "Accessories", "sku": "P200", "quantity": 2, "unit_price": 799, "status": "paid"},
    {"order_id": 1002, "customer": "Bala", "category": "Accessories", "sku": "P300", "quantity": 1, "unit_price": 1499, "status": "shipped"},
    {"order_id": 1003, "customer": "Charan", "category": "Electronics", "sku": "P400", "quantity": 1, "unit_price": 12999, "status": "cancelled"},
    {"order_id": 1004, "customer": "Divya", "category": "Accessories", "sku": "P200", "quantity": 3, "unit_price": 799, "status": "delivered"},
]

summary = {
    "revenue": 0.0,
    "units_sold": 0,
    "revenue_by_category": {},
    "units_by_sku": {},
    "orders_by_customer": {},
}

for line in order_lines:
    if line["status"] == "cancelled":
        continue

    line_total = line["quantity"] * line["unit_price"]
    summary["revenue"] += line_total
    summary["units_sold"] += line["quantity"]

    category = line["category"]
    summary["revenue_by_category"][category] = (
        summary["revenue_by_category"].get(category, 0) + line_total
    )

    sku = line["sku"]
    summary["units_by_sku"][sku] = (
        summary["units_by_sku"].get(sku, 0) + line["quantity"]
    )

    customer = line["customer"]
    summary["orders_by_customer"].setdefault(customer, set()).add(line["order_id"])

summary["orders_by_customer"] = {
    customer: len(order_ids)
    for customer, order_ids in summary["orders_by_customer"].items()
}

print(f"Revenue: ₹{summary['revenue']:,.2f}")
print(f"Units sold: {summary['units_sold']}")
print(summary["revenue_by_category"])
print(summary["units_by_sku"])
print(summary["orders_by_customer"])
```
