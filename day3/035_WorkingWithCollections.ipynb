# Working With Collections

Collection processing transforms raw values into useful results. Python provides iterable functions such as `map`, `filter`, `sorted`, `zip`, `enumerate`, `any`, `all`, `min`, `max`, and `sum`. These functions can work with lists, tuples, sets, dictionary views, generators, and many other iterable objects.

## 1. Ecommerce Dataset

```python
products = [
    {"sku": "P100", "name": "Laptop", "category": "Electronics", "price": 65000.0, "stock": 8, "rating": 4.6},
    {"sku": "P200", "name": "Mouse", "category": "Accessories", "price": 799.0, "stock": 35, "rating": 4.4},
    {"sku": "P300", "name": "Keyboard", "category": "Accessories", "price": 1499.0, "stock": 0, "rating": 4.5},
    {"sku": "P400", "name": "Monitor", "category": "Electronics", "price": 12999.0, "stock": 12, "rating": 4.7},
    {"sku": "P500", "name": "Laptop Stand", "category": "Accessories", "price": 2499.0, "stock": 6, "rating": 4.3},
]

orders = [
    {"order_id": 1001, "customer": "Asha", "sku": "P100", "quantity": 1, "status": "paid"},
    {"order_id": 1001, "customer": "Asha", "sku": "P200", "quantity": 2, "status": "paid"},
    {"order_id": 1002, "customer": "Bala", "sku": "P300", "quantity": 1, "status": "cancelled"},
    {"order_id": 1003, "customer": "Charan", "sku": "P400", "quantity": 2, "status": "shipped"},
    {"order_id": 1004, "customer": "Divya", "sku": "P500", "quantity": 1, "status": "delivered"},
    {"order_id": 1005, "customer": "Asha", "sku": "P200", "quantity": 3, "status": "delivered"},
]
```

## 2. `map`: Transform Every Value

`map(function, iterable)` applies a function to each value and returns a lazy iterator. Convert it to a collection when the results must be stored, displayed, indexed, or reused.

```python
product_names = list(map(lambda product: product["name"], products))
discounted_prices = list(map(lambda product: round(product["price"] * 0.90, 2), products))

print(product_names)
print(discounted_prices)
```

## 3. `map` With a Named Function

A named function is useful when the transformation contains several steps or will be reused.

```python
def add_tax(product):
    return {
        **product,
        "tax": round(product["price"] * 0.18, 2),
        "price_with_tax": round(product["price"] * 1.18, 2),
    }


taxed_products = list(map(add_tax, products))
print(taxed_products[0])
print(products[0])
```

## 4. `map` With Multiple Iterables

When several iterables are supplied, `map` passes one value from each iterable to the function and stops at the shortest iterable.

```python
skus = ["P100", "P200", "P300"]
quantities = [1, 2, 3]
unit_prices = [65000, 799, 1499]

line_totals = list(map(lambda quantity, price: quantity * price, quantities, unit_prices))
sku_totals = list(map(lambda sku, total: (sku, total), skus, line_totals))

print(line_totals)
print(sku_totals)
```

## 5. `filter`: Keep Matching Values

`filter(predicate, iterable)` returns the values for which the predicate is truthy. It also returns a lazy iterator.

```python
available_products = list(filter(lambda product: product["stock"] > 0, products))
successful_orders = list(filter(lambda order: order["status"] != "cancelled", orders))

print([product["name"] for product in available_products])
print([order["order_id"] for order in successful_orders])
```

## 6. `filter(None, iterable)`

Passing `None` removes falsy values such as `None`, `False`, numeric zero, and empty strings or collections.

```python
coupon_inputs = ["SAVE10", "", None, "FREESHIP", False, "WELCOME"]
valid_coupon_inputs = list(filter(None, coupon_inputs))
print(valid_coupon_inputs)
```

## 7. Comprehensions as Map and Filter

A comprehension often reads more naturally when transformation and filtering are combined. It creates the complete result immediately.

```python
premium_product_names = [
    product["name"].upper()
    for product in products
    if product["price"] >= 10000 and product["stock"] > 0
]

discount_by_sku = {
    product["sku"]: round(product["price"] * 0.90, 2)
    for product in products
    if product["stock"] > 0
}

active_categories = {product["category"] for product in products if product["stock"] > 0}

print(premium_product_names)
print(discount_by_sku)
print(active_categories)
```

## 8. `sorted`: Create a Sorted List

`sorted(iterable)` accepts any iterable and returns a new list. The input remains unchanged.

```python
prices = {65000, 799, 1499, 12999, 2499}
ascending_prices = sorted(prices)
descending_prices = sorted(prices, reverse=True)

print(ascending_prices)
print(descending_prices)
print(prices)
```

## 9. Sorting Records With `key`

The key function extracts the comparison value from each record.

```python
products_by_price = sorted(products, key=lambda product: product["price"])
products_by_rating = sorted(products, key=lambda product: product["rating"], reverse=True)

print([(product["name"], product["price"]) for product in products_by_price])
print([(product["name"], product["rating"]) for product in products_by_rating])
```

## 10. Sorting With Multiple Fields

A tuple key applies comparisons from left to right. Negating a numeric field produces descending order for that field while retaining ascending order for another.

```python
ranked_products = sorted(
    products,
    key=lambda product: (-product["rating"], product["price"], product["sku"]),
)

for product in ranked_products:
    print(product["sku"], product["rating"], product["price"])
```

## 11. `itemgetter` and `attrgetter`

`operator.itemgetter` extracts positions or mapping keys. `attrgetter` extracts object attributes. They are useful as reusable sorting and grouping keys.

```python
from operator import itemgetter

by_category_then_price = sorted(products, key=itemgetter("category", "price"))
print([(product["category"], product["name"], product["price"]) for product in by_category_then_price])
```

## 12. Stable Sorting

Python sorting is stable: records with equal keys retain their existing relative order. Multiple sort passes can therefore build mixed-direction ordering.

```python
stable_products = products.copy()
stable_products.sort(key=itemgetter("price"))
stable_products.sort(key=itemgetter("category"))

print([(product["category"], product["price"]) for product in stable_products])
```

## 13. `min` and `max` With Records

The `key` function selects the field used to find the minimum or maximum while the complete record is returned.

```python
cheapest_product = min(products, key=itemgetter("price"))
highest_rated_product = max(products, key=itemgetter("rating"))

print(cheapest_product)
print(highest_rated_product)
```

## 14. `sum`, `any`, and `all`

`sum` aggregates numbers. `any` stops at the first truthy result, and `all` stops at the first falsy result.

```python
total_inventory_units = sum(product["stock"] for product in products)
has_out_of_stock_product = any(product["stock"] == 0 for product in products)
all_prices_positive = all(product["price"] > 0 for product in products)

print(total_inventory_units)
print(has_out_of_stock_product)
print(all_prices_positive)
```

## 15. `enumerate`

`enumerate` adds a counter without manually maintaining an index. It yields `(index, value)` tuples.

```python
for rank, product in enumerate(ranked_products, start=1):
    print(f"{rank}. {product['name']} - {product['rating']}")
```

## 16. `zip`

`zip` combines parallel iterables and stops at the shortest input. Its tuples can be converted into records or passed to `dict` when one iterable contains unique keys.

```python
skus = ["P100", "P200", "P300"]
stocks = [8, 35, 0]
ratings = [4.6, 4.4, 4.5]

stock_by_sku = dict(zip(skus, stocks))
combined_rows = list(zip(skus, stocks, ratings))

print(stock_by_sku)
print(combined_rows)
```

## 17. Joining Two Collections

Build a dictionary index for the lookup side, then enrich each record using average `O(1)` key lookup rather than repeatedly scanning another list.

```python
products_by_sku = {product["sku"]: product for product in products}
enriched_orders = [
    {
        **order,
        "product_name": products_by_sku[order["sku"]]["name"],
        "line_total": order["quantity"] * products_by_sku[order["sku"]]["price"],
    }
    for order in orders
    if order["sku"] in products_by_sku
]

print(enriched_orders[0])
```

## 18. Grouping With a Dictionary

`setdefault` or `defaultdict` can accumulate several records under one lookup key.

```python
from collections import defaultdict

products_by_category = defaultdict(list)
for product in products:
    products_by_category[product["category"]].append(product)

for category, category_products in products_by_category.items():
    print(category, [product["name"] for product in category_products])
```

## 19. Grouping Consecutive Values With `groupby`

`itertools.groupby` groups adjacent equal keys. Sort by the grouping key first when matching records are not already consecutive.

```python
from itertools import groupby

orders_by_status = sorted(orders, key=itemgetter("status"))
status_groups = {
    status: list(group)
    for status, group in groupby(orders_by_status, key=itemgetter("status"))
}

for status, status_orders in status_groups.items():
    print(status, [order["order_id"] for order in status_orders])
```

## 20. Counting With `Counter`

`Counter` counts hashable values and provides frequency-oriented operations such as `most_common`.

```python
from collections import Counter

status_counts = Counter(order["status"] for order in orders)
sku_quantities = Counter()
for order in orders:
    sku_quantities[order["sku"]] += order["quantity"]

print(status_counts)
print(sku_quantities)
print(sku_quantities.most_common(2))
```

## 21. `reduce` for Cumulative Transformations

`functools.reduce` combines values two at a time into one result. `sum`, `min`, `max`, `any`, and `all` are clearer for their specific operations.

```python
from functools import reduce

cart_factors = [0.90, 0.95, 1.18]
combined_factor = reduce(lambda current, factor: current * factor, cart_factors, 1.0)
final_price = round(1000 * combined_factor, 2)

print(combined_factor)
print(final_price)
```

## 22. Chaining and Flattening

`chain` traverses several iterables as one sequence without first concatenating them. `chain.from_iterable` flattens one level.

```python
from itertools import chain

warehouse_a_skus = ["P100", "P200"]
warehouse_b_skus = ["P300", "P400"]
all_warehouse_skus = list(chain(warehouse_a_skus, warehouse_b_skus))

order_item_groups = [["P100", "P200"], ["P300"], ["P200", "P500"]]
all_order_skus = list(chain.from_iterable(order_item_groups))

print(all_warehouse_skus)
print(all_order_skus)
```

## 23. Batching With Slices

Slicing divides a materialized sequence into batches. The final batch may contain fewer values.

```python
order_ids = list(range(1001, 1012))
batch_size = 4
order_batches = [
    order_ids[start:start + batch_size]
    for start in range(0, len(order_ids), batch_size)
]

print(order_batches)
```

## 24. Lazy Processing and `islice`

Generators, `map`, and `filter` produce values on demand. `islice` reads a selected range without materializing the complete input.

```python
from itertools import islice

large_sku_stream = (f"SKU-{number:06d}" for number in range(1_000_000))
first_page = list(islice(large_sku_stream, 5))
second_page = list(islice(large_sku_stream, 5))

print(first_page)
print(second_page)
```

## 25. Pipeline: Filter, Transform, Sort, and Aggregate

A processing pipeline is easier to verify when each stage has one responsibility and a clearly named result.

```python
active_orders = filter(lambda order: order["status"] != "cancelled", orders)
enriched_active_orders = map(
    lambda order: {
        **order,
        "product": products_by_sku[order["sku"]]["name"],
        "line_total": order["quantity"] * products_by_sku[order["sku"]]["price"],
    },
    active_orders,
)
ranked_order_lines = sorted(
    enriched_active_orders,
    key=itemgetter("line_total"),
    reverse=True,
)
revenue = sum(order["line_total"] for order in ranked_order_lines)

print(f"Revenue: ₹{revenue:,.2f}")
for rank, order in enumerate(ranked_order_lines, start=1):
    print(f"{rank}. {order['product']} - ₹{order['line_total']:,.2f}")
```

## 26. Complete Collection Processing

The final result combines lookup, filtering, transformation, aggregation, grouping, distinct values, and sorting.

```python
valid_orders = [
    order
    for order in orders
    if order["status"] != "cancelled" and order["sku"] in products_by_sku
]

revenue_by_customer = defaultdict(float)
units_by_sku = Counter()
order_ids_by_customer = defaultdict(set)

for order in valid_orders:
    product = products_by_sku[order["sku"]]
    line_total = order["quantity"] * product["price"]
    revenue_by_customer[order["customer"]] += line_total
    units_by_sku[order["sku"]] += order["quantity"]
    order_ids_by_customer[order["customer"]].add(order["order_id"])

customer_report = sorted(
    (
        {
            "customer": customer,
            "orders": len(order_ids_by_customer[customer]),
            "revenue": revenue,
        }
        for customer, revenue in revenue_by_customer.items()
    ),
    key=lambda row: (-row["revenue"], row["customer"]),
)

sold_skus = set(units_by_sku)
available_skus = {product["sku"] for product in products if product["stock"] > 0}
available_but_unsold = sorted(available_skus - sold_skus)

print(customer_report)
print(units_by_sku.most_common())
print(available_but_unsold)
```
