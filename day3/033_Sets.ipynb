# Sets

A set is an unordered collection of unique, hashable values. Sets are useful for removing duplicates, fast membership checks, comparing catalogs, finding common customers, detecting missing products, and combining distinct values.

## 1. Creating Sets

Curly braces containing values create a set. An empty pair of braces creates a dictionary, so an empty set must be created with `set()`.

```python
categories = {"Electronics", "Books", "Fashion"}
empty_set = set()
empty_braces = {}

print(categories)
print(type(categories))
print(empty_set, type(empty_set))
print(empty_braces, type(empty_braces))
```

## 2. Duplicate Values

A set stores only one occurrence of each equal value. Repeated values are removed during creation.

```python
purchased_skus = {"P100", "P200", "P100", "P300", "P200"}
print(purchased_skus)
print(len(purchased_skus))
```

## 3. Removing Duplicates From a Sequence

Converting a list to a set removes duplicates but does not preserve sequence order. Dictionary keys can be used when first-seen order must be retained.

```python
search_history = ["laptop", "mouse", "laptop", "monitor", "mouse"]
unique_searches = set(search_history)
ordered_unique_searches = list(dict.fromkeys(search_history))

print(unique_searches)
print(ordered_unique_searches)
```

## 4. Sets Are Unordered

A set does not support positional indexing or slicing. Its displayed or iteration order should not be treated as fixed.

```python
warehouses = {"Chennai", "Bengaluru", "Hyderabad"}

for warehouse in warehouses:
    print(warehouse)

try:
    print(warehouses[0])
except TypeError as error:
    print(type(error).__name__, error)
```

## 5. Sorting Set Values

A set itself cannot be sorted in place. `sorted()` reads the set and returns a new list. A reverse order or a custom key can also be supplied.

```python
prices = {1499, 499, 65000, 799, 12999}
ascending_prices = sorted(prices)
descending_prices = sorted(prices, reverse=True)

product_names = {"Laptop Stand", "mouse", "USB Hub", "keyboard"}
alphabetical_names = sorted(product_names, key=str.casefold)

print(ascending_prices, type(ascending_prices))
print(descending_prices)
print(alphabetical_names)
```

## 6. Casting to a Set

`set()` consumes an iterable. Converting a dictionary produces its keys unless a view is passed explicitly.

```python
print(set(["P100", "P200", "P100"]))
print(set(("paid", "shipped", "paid")))
print(set("CART"))

product = {"sku": "P100", "name": "Laptop", "price": 65000}
print(set(product))
print(set(product.values()))
```

## 7. Membership

Membership checks are one of the main uses of a set and are typically constant time on average.

```python
blocked_skus = {"P120", "P215", "P410"}

print("P215" in blocked_skus)
print("P100" not in blocked_skus)
```

## 8. Adding Values

`add` inserts one value. `update` consumes one or more iterables and adds all their values.

```python
featured_skus = {"P100", "P200"}
featured_skus.add("P300")
featured_skus.add("P100")
featured_skus.update(["P400", "P500"], {"P600"})

print(featured_skus)
```

## 9. `add` and `update`

`add` treats its argument as one value. `update` iterates over its argument. A string passed to `update` is therefore added character by character.

```python
tags = set()
tags.add("sale")
print(tags)

letters = set()
letters.update("sale")
print(letters)
```

## 10. Removing Values

`remove` raises `KeyError` when the value is absent. `discard` does nothing when it is absent. `pop` removes and returns an arbitrary value, and `clear` empties the set.

```python
active_coupons = {"SAVE10", "FREESHIP", "WELCOME"}
active_coupons.remove("SAVE10")
active_coupons.discard("EXPIRED")
removed_coupon = active_coupons.pop()

print(removed_coupon)
print(active_coupons)

try:
    active_coupons.remove("EXPIRED")
except KeyError as error:
    print(type(error).__name__, error)

active_coupons.clear()
print(active_coupons)
```

## 11. Union

A union contains every distinct value present in either set. `|` and `union()` create a new set.

```python
online_skus = {"P100", "P200", "P300"}
store_skus = {"P200", "P300", "P400"}

all_skus_operator = online_skus | store_skus
all_skus_method = online_skus.union(store_skus, {"P500"})

print(all_skus_operator)
print(all_skus_method)
```

## 12. Intersection

An intersection contains values present in every participating set.

```python
online_skus = {"P100", "P200", "P300"}
store_skus = {"P200", "P300", "P400"}
warehouse_skus = {"P200", "P400", "P500"}

common_two = online_skus & store_skus
common_all = online_skus.intersection(store_skus, warehouse_skus)

print(common_two)
print(common_all)
```

## 13. Difference

A difference contains values from the left set that are absent from the right set. The operation is directional.

```python
catalog_skus = {"P100", "P200", "P300", "P400"}
inventory_skus = {"P100", "P300", "P500"}

missing_inventory = catalog_skus - inventory_skus
unlisted_inventory = inventory_skus.difference(catalog_skus)

print(missing_inventory)
print(unlisted_inventory)
```

## 14. Symmetric Difference

A symmetric difference contains values found in exactly one of the sets, excluding shared values.

```python
yesterday_customers = {"C101", "C102", "C103"}
today_customers = {"C102", "C103", "C104"}

changed_customers = yesterday_customers ^ today_customers
same_result = yesterday_customers.symmetric_difference(today_customers)

print(changed_customers)
print(same_result)
```

## 15. Updating Set Operations

The methods ending in `_update` change the existing set instead of creating a new one.

```python
catalog_skus = {"P100", "P200", "P300"}
catalog_skus.update({"P300", "P400"})
print(catalog_skus)

catalog_skus.intersection_update({"P200", "P300", "P500"})
print(catalog_skus)

catalog_skus.difference_update({"P300"})
print(catalog_skus)

catalog_skus.symmetric_difference_update({"P200", "P600"})
print(catalog_skus)
```

## 16. Operator and Method Forms

Set operators require set-like operands. Method forms accept any iterable, making them convenient with lists, tuples, and generators.

```python
catalog_skus = {"P100", "P200", "P300"}
incoming_skus = ["P200", "P400", "P400"]

print(catalog_skus.union(incoming_skus))
print(catalog_skus.intersection(incoming_skus))
print(catalog_skus | set(incoming_skus))
```

## 17. Subsets and Supersets

A subset contains no value outside the other set. A proper subset is smaller than the other set. Superset comparisons express the reverse relationship.

```python
required_fields = {"sku", "name", "price"}
product_fields = {"sku", "name", "price", "stock", "rating"}

print(required_fields <= product_fields)
print(required_fields < product_fields)
print(product_fields >= required_fields)
print(product_fields > required_fields)
print(required_fields.issubset(product_fields))
print(product_fields.issuperset(required_fields))
```

## 18. Disjoint Sets

Two sets are disjoint when they have no value in common.

```python
paid_order_ids = {1001, 1002, 1004}
cancelled_order_ids = {1003, 1005}
shipped_order_ids = {1002, 1004}

print(paid_order_ids.isdisjoint(cancelled_order_ids))
print(paid_order_ids.isdisjoint(shipped_order_ids))
```

## 19. Set Comprehensions

A set comprehension transforms or filters values while automatically keeping only distinct results.

```python
orders = [
    {"order_id": 1001, "customer": "Asha", "city": "Chennai", "status": "paid"},
    {"order_id": 1002, "customer": "Bala", "city": "Bengaluru", "status": "shipped"},
    {"order_id": 1003, "customer": "Asha", "city": "Chennai", "status": "paid"},
    {"order_id": 1004, "customer": "Divya", "city": "Chennai", "status": "cancelled"},
]

customer_names = {order["customer"].upper() for order in orders}
active_cities = {order["city"] for order in orders if order["status"] != "cancelled"}

print(customer_names)
print(active_cities)
```

## 20. Hashable Set Elements

Set elements must be hashable. Strings, numbers, and tuples containing hashable values can be stored. Lists, dictionaries, and sets cannot be elements.

```python
warehouse_products = {("Chennai", "P100"), ("Bengaluru", "P200")}
print(warehouse_products)

try:
    invalid_set = {["Chennai", "P100"]}
except TypeError as error:
    print(type(error).__name__, error)
```

## 21. Equality

Set equality depends only on membership. Order is irrelevant, and duplicates supplied during creation do not affect the result.

```python
set_a = {"P100", "P200", "P300"}
set_b = {"P300", "P100", "P200"}
set_c = set(["P100", "P200", "P200", "P300"])

print(set_a == set_b)
print(set_a == set_c)
print(set_a is set_b)
```

## 22. Aliasing and Copying

Assignment creates another reference to the same set. `copy()` or `set(...)` creates a separate set.

```python
sale_skus = {"P100", "P200"}
sale_alias = sale_skus
sale_copy = sale_skus.copy()

sale_alias.add("P300")
sale_copy.add("P400")

print(sale_skus)
print(sale_copy)
print(sale_alias is sale_skus)
print(sale_copy is sale_skus)
```

## 23. Mutating During Iteration

The size of a set should not change during direct iteration. Iterate over a copy or build a new set.

```python
coupon_codes = {"SAVE10", "EXPIRED5", "FREESHIP", "EXPIRED20"}

for code in coupon_codes.copy():
    if code.startswith("EXPIRED"):
        coupon_codes.remove(code)

print(coupon_codes)

all_codes = {"SAVE10", "EXPIRED5", "FREESHIP"}
active_codes = {code for code in all_codes if not code.startswith("EXPIRED")}
print(active_codes)
```

## 24. `frozenset`

A `frozenset` is an immutable set. It supports non-mutating set operations and can be used as a dictionary key or as an element of another set.

```python
bundle = frozenset({"P100", "P200"})
bundle_prices = {
    frozenset({"P100", "P200"}): 64000,
    frozenset({"P200", "P300"}): 1999,
}

print(bundle)
print(bundle | {"P300"})
print(bundle_prices[frozenset({"P200", "P100"})])

try:
    bundle.add("P300")
except AttributeError as error:
    print(type(error).__name__, error)
```

## 25. Sets Containing Frozen Sets

Ordinary sets cannot contain other mutable sets, but they can contain frozen sets.

```python
product_bundles = {
    frozenset({"Laptop", "Mouse"}),
    frozenset({"Keyboard", "Mouse"}),
}

print(product_bundles)
print(frozenset({"Mouse", "Laptop"}) in product_bundles)
```

## 26. Dictionary Views and Set Operations

Dictionary key views behave like sets and can be compared without first converting them.

```python
catalog = {"P100": "Laptop", "P200": "Mouse", "P300": "Keyboard"}
inventory = {"P100": 8, "P300": 20, "P400": 5}

print(catalog.keys() & inventory.keys())
print(catalog.keys() - inventory.keys())
print(inventory.keys() - catalog.keys())
```

## 27. `min`, `max`, `sum`, `any`, and `all`

General iterable functions work with sets. Their results do not depend on iteration order.

```python
stock_levels = {0, 3, 8, 12, 20}
prices = {499, 799, 1499, 12999}

print(min(prices))
print(max(prices))
print(sum(prices))
print(any(stock == 0 for stock in stock_levels))
print(all(stock >= 0 for stock in stock_levels))
```

## 28. Relationship Comparisons

`<`, `<=`, `>`, and `>=` mean subset and superset relationships for sets. They do not provide numeric or alphabetical sorting.

```python
required_fields = {"sku", "name"}
complete_fields = {"sku", "name", "price", "stock"}
unrelated_fields = {"customer", "city"}

print(required_fields < complete_fields)
print(complete_fields > required_fields)
print(required_fields.isdisjoint(unrelated_fields))
```

## 29. Set Performance

Membership, insertion, and deletion are typically constant time on average. Union, intersection, and difference depend on the sizes of the participating sets. Sets use additional memory to provide fast hash-based operations.

```python
catalog_skus = {f"SKU-{number:05d}" for number in range(10_000)}
incoming_skus = {f"SKU-{number:05d}" for number in range(7_500, 12_500)}

print("SKU-07500" in catalog_skus)
print(len(catalog_skus & incoming_skus))
print(len(incoming_skus - catalog_skus))
```

## 30. Complete Ecommerce Set Analysis

```python
orders = [
    {"order_id": 1001, "customer_id": "C101", "sku": "P100", "category": "Electronics", "status": "paid"},
    {"order_id": 1001, "customer_id": "C101", "sku": "P200", "category": "Accessories", "status": "paid"},
    {"order_id": 1002, "customer_id": "C102", "sku": "P300", "category": "Accessories", "status": "shipped"},
    {"order_id": 1003, "customer_id": "C103", "sku": "P400", "category": "Electronics", "status": "cancelled"},
    {"order_id": 1004, "customer_id": "C101", "sku": "P200", "category": "Accessories", "status": "delivered"},
    {"order_id": 1005, "customer_id": "C104", "sku": "P500", "category": "Home", "status": "delivered"},
]

catalog_skus = {"P100", "P200", "P300", "P400", "P500", "P600"}
inventory_skus = {"P100", "P200", "P300", "P500", "P700"}
active_orders = [order for order in orders if order["status"] != "cancelled"]

unique_customers = {order["customer_id"] for order in active_orders}
sold_skus = {order["sku"] for order in active_orders}
active_categories = {order["category"] for order in active_orders}
out_of_stock_skus = catalog_skus - inventory_skus
unlisted_inventory_skus = inventory_skus - catalog_skus
unsold_skus = catalog_skus - sold_skus

print("Unique customers       :", sorted(unique_customers))
print("Sold SKUs              :", sorted(sold_skus))
print("Active categories      :", sorted(active_categories))
print("Catalog without stock  :", sorted(out_of_stock_skus))
print("Inventory not cataloged:", sorted(unlisted_inventory_skus))
print("Unsold catalog SKUs    :", sorted(unsold_skus))
```
