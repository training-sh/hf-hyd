# Tuples

A tuple is an ordered, immutable collection. A list usually represents a changing sequence of similar values, while a tuple often represents one fixed record whose positions have specific meanings.

An ecommerce event such as `(1001, "Asha", 2499.0, "paid")` can be read as one order record: order ID, customer, amount, and status.

## 1. Empty, Single-Item, and Multi-Item Tuples

The comma creates a tuple. Parentheses make the structure clearer. A single-item tuple requires a trailing comma.

```python
empty_order = ()
single_order_id = (1001,)
two_fields = (1001, "paid")
order_record = (1001, "Asha", 2499.0, "paid")

print(empty_order, type(empty_order), len(empty_order))
print(single_order_id, type(single_order_id), len(single_order_id))
print(two_fields, type(two_fields))
print(order_record)
```

## 2. Parentheses Without a Comma

Parentheses alone only group an expression. The trailing comma distinguishes a one-item tuple from an ordinary value.

```python
order_id = (1001)
order_id_record = (1001,)

print(order_id, type(order_id))
print(order_id_record, type(order_id_record))
```

## 3. Tuple Packing

Comma-separated values are packed into a tuple even when parentheses are omitted.

```python
product_record = "P100", "Laptop", 65000.0, True
warehouse_location = "Chennai", "A-12"

print(product_record)
print(type(product_record))
print(warehouse_location)
```

## 4. A Tuple as One Record

Each position has a defined meaning. Keeping the fields in a known order makes the tuple a compact record.

```python
# (order_id, customer, amount, status)
order = (1001, "Asha", 2499.0, "paid")

print(f"Order ID : {order[0]}")
print(f"Customer : {order[1]}")
print(f"Amount   : ₹{order[2]:,.2f}")
print(f"Status   : {order[3]}")
```

## 5. Record Unpacking

Unpacking gives names to the fields and makes their meanings explicit. The number of variables must match the number of values.

```python
order = (1002, "Bala", 12999.0, "shipped")
order_id, customer, amount, status = order

print(order_id)
print(customer)
print(amount)
print(status)
```

## 6. Accessing Tuple Fields

Tuple indexing and slicing use the same syntax as lists. A slice returns a new tuple.

```python
product = ("P100", "Laptop", "Electronics", 65000.0, 4.6)

print(product[0])
print(product[-1])
print(product[1:4])
print(product[::-1])
```

## 7. Tuple Immutability

A tuple cannot have its fields added, removed, or reassigned after creation. This protects the shape of a record from accidental structural changes.

```python
order = (1001, "Asha", 2499.0, "paid")

try:
    order[3] = "shipped"
except TypeError as error:
    print(type(error).__name__, error)
```

## 8. Creating an Updated Record

Because a tuple is immutable, an update creates another tuple rather than modifying the existing one.

```python
order = (1001, "Asha", 2499.0, "paid")
updated_order = (*order[:3], "shipped")

print(order)
print(updated_order)
print(order is updated_order)
```

## 9. Tuple and List

A list is suitable when the collection grows, shrinks, or changes. A tuple is suitable when the number and meaning of fields should remain fixed.

```python
cart_items = ["Laptop", "Mouse"]
order_identity = (1001, "2026-07-19")

cart_items.append("Keyboard")
print(cart_items)
print(order_identity)

print(hasattr(cart_items, "append"))
print(hasattr(order_identity, "append"))
```

### Major Differences

| List | Tuple |
|---|---|
| Mutable | Immutable |
| Written with `[]` | Commonly written with `()` |
| Suitable for a changing collection | Suitable for a fixed record |
| Supports methods such as `append`, `extend`, and `remove` | Provides only `count` and `index` as tuple-specific methods |
| Cannot be a dictionary key | Can be a dictionary key when all contained values are hashable |
| Usually uses slightly more memory | Usually uses slightly less memory |

## 10. Casting Between Lists and Tuples

Casting is useful when values need to be collected in a list and later fixed as a record, or temporarily converted for modification.

```python
order_fields = [1001, "Asha", 2499.0, "paid"]
order_record = tuple(order_fields)
editable_fields = list(order_record)
editable_fields[3] = "shipped"
revised_record = tuple(editable_fields)

print(order_record)
print(revised_record)
```

## 11. Tuple Unpacking With Remaining Fields

Extended unpacking stores the remaining values in a list, even when the source is a tuple.

```python
order = (1001, "Asha", "Laptop", "Mouse", "USB Hub")
order_id, customer, *items = order

print(order_id)
print(customer)
print(items, type(items))
```

## 12. Ignoring Unused Fields

An underscore is conventionally used for a field that is deliberately ignored.

```python
order = (1003, "Charan", 799.0, "delivered")
order_id, _, amount, status = order

print(order_id, amount, status)
```

## 13. Swapping Values

The right side is packed and then unpacked, so a temporary variable is unnecessary.

```python
minimum_price = 499
maximum_price = 65000

minimum_price, maximum_price = maximum_price, minimum_price
print(minimum_price, maximum_price)
```

## 14. Returning Multiple Values

A function returning comma-separated values returns one tuple. The result can remain a record or be unpacked immediately.

```python
def order_metrics(unit_price, quantity, discount_rate=0):
    gross = unit_price * quantity
    discount = gross * discount_rate
    net = gross - discount
    return gross, discount, net


metrics = order_metrics(1499, 3, 0.10)
gross, discount, net = metrics

print(metrics, type(metrics))
print(gross, discount, net)
```

## 15. A List of Tuple Records

The outer list is mutable, so records can be added or removed. Each inner tuple keeps the field structure fixed.

```python
# Each record: (order_id, customer, amount, status)
orders = [
    (1001, "Asha", 2499.0, "paid"),
    (1002, "Bala", 12999.0, "shipped"),
    (1003, "Charan", 799.0, "cancelled"),
]

orders.append((1004, "Divya", 5499.0, "delivered"))

for order_id, customer, amount, status in orders:
    print(f"{order_id} | {customer:<7} | ₹{amount:>9,.2f} | {status}")
```

## 16. Searching, Counting, and Membership

Tuples support the common sequence operations and provide `count` and `index` methods.

```python
order_statuses = ("paid", "shipped", "paid", "cancelled")

print("shipped" in order_statuses)
print(order_statuses.count("paid"))
print(order_statuses.index("cancelled"))
```

## 17. Concatenation and Repetition

These operations do not alter an existing tuple. They create a new tuple.

```python
identity = (1001, "Asha")
transaction = (2499.0, "paid")
order = identity + transaction
placeholder_fields = (None,) * 3

print(order)
print(placeholder_fields)
```

## 18. Immutable Tuple, Mutable Member

Tuple immutability applies to its references. A mutable object stored inside the tuple can still change.

```python
order = (1001, "Asha", ["Laptop", "Mouse"])
order[2].append("Keyboard")

print(order)

try:
    hash(order)
except TypeError as error:
    print(type(error).__name__, error)
```

## 19. Tuples as Dictionary Keys

A tuple can be used as a dictionary key when every contained value is hashable. A composite key can identify a product within a warehouse.

```python
inventory = {
    ("Chennai", "P100"): 12,
    ("Chennai", "P200"): 35,
    ("Bengaluru", "P100"): 8,
}

print(inventory[("Chennai", "P100")])

for (warehouse, sku), quantity in inventory.items():
    print(warehouse, sku, quantity)
```

## 20. Enumeration and `zip`

`enumerate` produces `(index, value)` tuples. `zip` produces tuples containing one value from each input.

```python
products = ("Laptop", "Mouse", "Keyboard")
prices = (65000, 799, 1499)

for rank, product in enumerate(products, start=1):
    print(rank, product)

catalog_records = tuple(zip(products, prices))
print(catalog_records)
```

## 21. Sorting Tuple Records

`sorted` returns a list even when its input contains tuples. A key function selects the record field used for sorting.

```python
# Each record: (sku, product, price, rating)
products = [
    ("P100", "Laptop", 65000, 4.6),
    ("P200", "Mouse", 799, 4.4),
    ("P300", "Keyboard", 1499, 4.5),
    ("P400", "Monitor", 12999, 4.7),
]

by_price = sorted(products, key=lambda product: product[2])
by_rating = sorted(products, key=lambda product: product[3], reverse=True)

print(by_price)
print(by_rating)
```

## 22. Named Tuples

A named tuple keeps tuple immutability and unpacking while allowing fields to be accessed by name. It is useful when positional indexes make a record difficult to read.

```python
from typing import NamedTuple


class Order(NamedTuple):
    order_id: int
    customer: str
    amount: float
    status: str


order = Order(1001, "Asha", 2499.0, "paid")

print(order)
print(order.order_id)
print(order.customer)
print(order[2])
```

## 23. Updating a Named Tuple Record

`_replace` creates a new record and leaves the original unchanged.

```python
shipped_order = order._replace(status="shipped")

print(order)
print(shipped_order)
```

## 24. Pattern Matching With Tuple Records

Structural pattern matching can select records by their shape and field values. This requires Python 3.10 or later.

```python
def fulfillment_message(order):
    match order:
        case (order_id, customer, _, "cancelled"):
            return f"Order {order_id} for {customer} requires no dispatch"
        case (order_id, customer, amount, "paid") if amount >= 5000:
            return f"Order {order_id} for {customer} gets priority dispatch"
        case (order_id, customer, _, status):
            return f"Order {order_id} for {customer}: {status}"


print(fulfillment_message((1001, "Asha", 65000.0, "paid")))
print(fulfillment_message((1002, "Bala", 799.0, "cancelled")))
```

## 25. Complete Order Record Processing

```python
# Record format: (order_id, customer, product, quantity, unit_price, status)
order_records = [
    (1001, "Asha", "Laptop", 1, 65000.0, "paid"),
    (1002, "Bala", "Mouse", 2, 799.0, "shipped"),
    (1003, "Charan", "Monitor", 1, 12999.0, "cancelled"),
    (1004, "Divya", "Keyboard", 3, 1499.0, "delivered"),
]

completed_records = [record for record in order_records if record[5] != "cancelled"]
revenue = sum(quantity * unit_price for _, _, _, quantity, unit_price, _ in completed_records)
ranked_records = sorted(
    completed_records,
    key=lambda record: record[3] * record[4],
    reverse=True,
)

print(f"Revenue: ₹{revenue:,.2f}")
for rank, (order_id, customer, product, quantity, unit_price, status) in enumerate(ranked_records, start=1):
    line_total = quantity * unit_price
    print(f"{rank}. {order_id} | {customer} | {product} | ₹{line_total:,.2f} | {status}")
```
