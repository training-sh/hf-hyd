# Lists

A list is an ordered, mutable collection. It can store duplicate values, mixed data types, and references to other objects. Lists are commonly used for product catalogs, carts, order lines, prices, and inventory updates.

## 1. Creating Lists

```python
products = ["Laptop", "Mouse", "Keyboard"]
prices = [65000, 799, 1499]
empty_cart = []
mixed_order = [1001, "paid", 2499.50, True]

print(products)
print(type(products))
print(len(empty_cart))
```

## 2. Ordered Values, Duplicates, and Mixed Types

A list preserves insertion order and permits repeated values. Although mixed types are valid, consistent structures are usually easier to process.

```python
cart_items = ["Mouse", "USB Cable", "Mouse", "Laptop Stand"]
order_summary = [1002, "Asha", 3298.0, "shipped"]

print(cart_items)
print(cart_items.count("Mouse"))
print(order_summary)
```

## 3. Indexing

Positive indexes start from the left at `0`. Negative indexes start from the right at `-1`.

```python
categories = ["Electronics", "Books", "Home", "Fashion"]

print(categories[0])
print(categories[2])
print(categories[-1])
print(categories[-2])
```

## 4. Slicing

The slice form is `list[start:stop:step]`. The start is included and the stop is excluded. Omitted boundaries use the beginning or end of the list.

```python
daily_sales = [12500, 18200, 9400, 22100, 17500, 24800, 19600]

print(daily_sales[1:4])
print(daily_sales[:3])
print(daily_sales[4:])
print(daily_sales[::2])
print(daily_sales[::-1])
print(daily_sales[-3:])
```

## 5. Slice Assignment

Slice assignment can replace, insert, or remove multiple elements. The replacement length does not have to match the selected length unless an extended slice uses a step other than `1`.

```python
featured = ["Phone", "Tablet", "Camera", "Speaker"]
featured[1:3] = ["Smartwatch", "Earbuds", "Monitor"]
print(featured)

featured[1:1] = ["Laptop"]
print(featured)

featured[-2:] = []
print(featured)
```

## 6. Mutability

The list object keeps the same identity while its contents can change.

```python
inventory = [12, 8, 20]
inventory_id = id(inventory)

inventory[1] = 7
inventory.append(15)

print(inventory)
print(inventory_id == id(inventory))
```

## 7. Aliasing and Identity

Assignment does not copy a list. Both names refer to the same object, so a change through either name is visible through the other.

```python
cart = ["Mouse", "Keyboard"]
active_cart = cart
active_cart.append("USB Hub")

print(cart)
print(active_cart is cart)
print(active_cart == cart)
```

## 8. Shallow Copies

`copy()`, `list(...)`, and `[:]` create a new outer list. Nested mutable objects are still shared.

```python
original_cart = ["Laptop", "Mouse"]
cart_copy = original_cart.copy()
cart_from_constructor = list(original_cart)
cart_from_slice = original_cart[:]

cart_copy.append("Sleeve")
print(original_cart)
print(cart_copy)
print(original_cart is cart_copy)
```

## 9. Shallow Copy and Deep Copy

A deep copy recursively copies nested mutable objects. It is useful when a completely independent order structure is required.

```python
from copy import deepcopy

order_lines = [["Laptop", 1], ["Mouse", 2]]
shallow_order = order_lines.copy()
deep_order = deepcopy(order_lines)

shallow_order[0][1] = 2
deep_order[1][1] = 5

print(order_lines)
print(shallow_order)
print(deep_order)
```

## 10. Casting to a List

`list()` consumes an iterable and stores its values in a new list. Dictionary conversion uses keys unless a view such as `values()` or `items()` is passed explicitly.

```python
print(list(("Laptop", "Mouse")))
print(list({"paid", "shipped", "cancelled"}))
print(list("CART"))
print(list(range(1001, 1006)))

product = {"sku": "P100", "name": "Mouse", "price": 799}
print(list(product))
print(list(product.values()))
print(list(product.items()))
```

## 11. Adding Elements

```python
cart = ["Laptop"]
cart.append("Mouse")
cart.extend(["Keyboard", "USB Hub"])
cart.insert(1, "Laptop Sleeve")

print(cart)
```

## 12. `append` and `extend`

`append` adds one object. `extend` iterates over its argument and adds each value.

```python
cart_a = ["Laptop"]
cart_a.append(["Mouse", "Keyboard"])

cart_b = ["Laptop"]
cart_b.extend(["Mouse", "Keyboard"])

print(cart_a)
print(cart_b)
```

## 13. Removing Elements

`remove` deletes the first matching value. `pop` removes and returns an item. `del` removes by index or slice, and `clear` empties the list.

```python
cart = ["Laptop", "Mouse", "Keyboard", "Mouse", "USB Hub"]
cart.remove("Mouse")
last_item = cart.pop()
del cart[1]

print(cart)
print(last_item)

cart.clear()
print(cart)
```

## 14. Searching and Counting

```python
statuses = ["paid", "shipped", "paid", "cancelled", "paid"]

print("shipped" in statuses)
print("returned" not in statuses)
print(statuses.count("paid"))
print(statuses.index("cancelled"))
```

## 15. Concatenation and Repetition

`+` and `*` create new lists. Repetition repeats references, which matters when the repeated value is mutable.

```python
electronics = ["Laptop", "Mouse"]
accessories = ["Sleeve", "USB Hub"]
catalog = electronics + accessories
promotion_slots = ["featured"] * 3

print(catalog)
print(promotion_slots)
```

## 16. The Repeated Nested List Trap

Multiplying a list containing a mutable object repeats the same reference. A comprehension creates independent inner lists.

```python
shared_bins = [[]] * 3
shared_bins[0].append("P100")
print(shared_bins)

independent_bins = [[] for _ in range(3)]
independent_bins[0].append("P100")
print(independent_bins)
```

## 17. Iteration

```python
products = ["Laptop", "Mouse", "Keyboard"]

for product_name in products:
    print(product_name.upper())
```

## 18. Enumeration

`enumerate` produces `(index, value)` pairs. The starting number can be changed when row numbers or ranks should begin at `1`.

```python
top_products = ["Laptop", "Smartphone", "Headphones"]

for rank, product_name in enumerate(top_products, start=1):
    print(f"{rank}. {product_name}")

ranked_products = list(enumerate(top_products, start=1))
print(ranked_products)
```

## 19. Parallel Iteration With `zip`

`zip` stops at the shortest input. Converting it to a list materializes the generated pairs.

```python
product_names = ["Laptop", "Mouse", "Keyboard"]
product_prices = [65000, 799, 1499]
stock_levels = [8, 35, 20]

catalog_rows = list(zip(product_names, product_prices, stock_levels))
for name, price, stock in catalog_rows:
    print(f"{name}: ₹{price:,} | stock={stock}")
```

## 20. List Comprehensions

A list comprehension builds a new list from an iterable. It can transform values and filter them in one expression.

```python
prices = [499, 799, 1499, 2499, 65000]
discounted_prices = [round(price * 0.90, 2) for price in prices]
premium_prices = [price for price in prices if price >= 2000]
price_labels = ["premium" if price >= 2000 else "standard" for price in prices]

print(discounted_prices)
print(premium_prices)
print(price_labels)
```

## 21. Nested Comprehensions

The `for` clauses follow the same order as equivalent nested loops.

```python
warehouses = ["Chennai", "Bengaluru"]
skus = ["P100", "P200", "P300"]
warehouse_skus = [(warehouse, sku) for warehouse in warehouses for sku in skus]
print(warehouse_skus)

quantity_matrix = [[4, 8, 2], [7, 3, 9]]
all_quantities = [quantity for row in quantity_matrix for quantity in row]
print(all_quantities)
```

## 22. Unpacking Lists

Extended unpacking collects remaining values into a new list.

```python
order = [1001, "Asha", "Laptop", "Mouse", "USB Hub"]
order_id, customer, *items = order
first_item, *other_items = items

print(order_id)
print(customer)
print(first_item)
print(other_items)
```

## 23. Sorting Lists

`sort()` changes the existing list and returns `None`. `sorted()` accepts any iterable and returns a new list.

```python
prices = [1499, 499, 65000, 799, 2499]
sorted_prices = sorted(prices)
descending_prices = sorted(prices, reverse=True)

print(prices)
print(sorted_prices)
print(descending_prices)

prices.sort()
print(prices)
```

## 24. Sorting Structured Data

The `key` function extracts the value used for comparison. Python sorting is stable, so records with equal keys keep their previous relative order.

```python
products = [
    {"sku": "P100", "name": "Laptop", "category": "Electronics", "price": 65000, "rating": 4.6},
    {"sku": "P200", "name": "Mouse", "category": "Accessories", "price": 799, "rating": 4.4},
    {"sku": "P300", "name": "Keyboard", "category": "Accessories", "price": 1499, "rating": 4.4},
    {"sku": "P400", "name": "Monitor", "category": "Electronics", "price": 12999, "rating": 4.7},
]

by_price = sorted(products, key=lambda product: product["price"])
by_rating_then_price = sorted(
    products,
    key=lambda product: (-product["rating"], product["price"]),
)

print([(product["name"], product["price"]) for product in by_price])
print([(product["name"], product["rating"]) for product in by_rating_then_price])
```

## 25. Reversing a List

`reverse()` changes the list. `reversed()` returns an iterator, while `[::-1]` creates a new list.

```python
recent_orders = [1001, 1002, 1003, 1004]
print(recent_orders[::-1])
print(list(reversed(recent_orders)))

recent_orders.reverse()
print(recent_orders)
```

## 26. Aggregation and Conditions

```python
order_amounts = [2499.0, 799.0, 12999.0, 1499.0]
stock_levels = [12, 0, 8, 3]

print(sum(order_amounts))
print(min(order_amounts))
print(max(order_amounts))
print(any(stock == 0 for stock in stock_levels))
print(all(stock >= 0 for stock in stock_levels))
```

## 27. Nested Lists

A nested list can represent rows and columns. Each index operation moves one level deeper.

```python
order_table = [
    [1001, "Asha", 2499.0, "paid"],
    [1002, "Bala", 799.0, "shipped"],
    [1003, "Charan", 12999.0, "cancelled"],
]

print(order_table[0])
print(order_table[1][1])
print([row[2] for row in order_table])
```

## 28. Lists of Dictionaries

A list supplies order and iteration; each dictionary gives the record named fields.

```python
orders = [
    {"order_id": 1001, "customer": "Asha", "amount": 2499.0, "status": "paid"},
    {"order_id": 1002, "customer": "Bala", "amount": 799.0, "status": "shipped"},
    {"order_id": 1003, "customer": "Charan", "amount": 12999.0, "status": "cancelled"},
    {"order_id": 1004, "customer": "Divya", "amount": 5499.0, "status": "paid"},
]

successful_orders = [order for order in orders if order["status"] != "cancelled"]
revenue = sum(order["amount"] for order in successful_orders)
high_value_customers = [
    order["customer"] for order in successful_orders if order["amount"] >= 5000
]

print(successful_orders)
print(revenue)
print(high_value_customers)
```

## 29. Safe Updates While Iterating

Removing elements from the list currently being traversed can skip values. Build a filtered list, assign to a slice, or iterate over a copy.

```python
stock_levels = [5, 0, 0, 8, 2, 0]
available_stock = [stock for stock in stock_levels if stock > 0]
print(available_stock)

stock_levels[:] = [stock for stock in stock_levels if stock > 0]
print(stock_levels)
```

## 30. Stack Operations

Appending and popping from the end are efficient and make a list suitable for last-in, first-out processing.

```python
packing_stack = []
packing_stack.append("Order-1001")
packing_stack.append("Order-1002")
packing_stack.append("Order-1003")

next_package = packing_stack.pop()
print(next_package)
print(packing_stack)
```

## 31. Queue Operations

Removing index `0` shifts all remaining list elements. `collections.deque` is preferred for frequent operations at both ends.

```python
from collections import deque

dispatch_queue = deque(["Order-1001", "Order-1002", "Order-1003"])
next_dispatch = dispatch_queue.popleft()

print(next_dispatch)
print(list(dispatch_queue))
```

## 32. Membership and Lookup Cost

List membership scans values and is linear in the list size. A set is usually more suitable for repeated membership checks, while a dictionary is suitable for lookup by key.

```python
blocked_skus = ["P120", "P215", "P410"]
blocked_sku_set = set(blocked_skus)
print("P215" in blocked_sku_set)

product_rows = [
    {"sku": "P100", "name": "Laptop"},
    {"sku": "P200", "name": "Mouse"},
]
products_by_sku = {product["sku"]: product for product in product_rows}
print(products_by_sku["P200"] )
```

## 33. Iterators and Materialization

Functions such as `map`, `filter`, `zip`, and `reversed` return lazy iterators in Python 3. `list()` materializes their results when indexing, reuse, or display as a list is required.

```python
prices = [499, 799, 1499, 2499]
taxed_price_iterator = map(lambda price: round(price * 1.18, 2), prices)
taxed_prices = list(taxed_price_iterator)

premium_price_iterator = filter(lambda price: price >= 1000, prices)
premium_prices = list(premium_price_iterator)

print(taxed_prices)
print(premium_prices)
```

## 34. Conditional Search With `next`

A generator expression with `next` avoids building an intermediate list when only the first match is needed. The second argument supplies a default.

```python
first_expensive_product = next(
    (product for product in products if product["price"] >= 10000),
    None,
)
print(first_expensive_product)
```

## 35. Deduplication While Preserving Order

Dictionary keys preserve insertion order. This form works when the list values are hashable.

```python
search_history = ["laptop", "mouse", "laptop", "monitor", "mouse"]
unique_searches = list(dict.fromkeys(search_history))
print(unique_searches)
```

## 36. Grouping Consecutive Values

`itertools.groupby` groups adjacent values. Sorting first is necessary when all equal keys should be grouped together.

```python
from itertools import groupby

orders_by_status = sorted(orders, key=lambda order: order["status"])
grouped_orders = {
    status: list(group)
    for status, group in groupby(orders_by_status, key=lambda order: order["status"])
}

for status, status_orders in grouped_orders.items():
    print(status, [order["order_id"] for order in status_orders])
```

## 37. Chunking a List

Slicing can split a list into batches without changing the original list.

```python
order_ids = list(range(1001, 1012))
batch_size = 4
order_batches = [
    order_ids[start:start + batch_size]
    for start in range(0, len(order_ids), batch_size)
]
print(order_batches)
```

## 38. Transposing Rows and Columns

Unpacking rows into `zip` changes a row-oriented structure into columns.

```python
sales_matrix = [
    [120, 90, 75],
    [140, 105, 88],
    [135, 110, 92],
]
sales_by_product = [list(column) for column in zip(*sales_matrix)]
print(sales_by_product)
```

## 39. Type Hints for Lists

Type hints document the expected element type. They are not enforced automatically at runtime.

```python
def calculate_cart_total(prices: list[float]) -> float:
    return round(sum(prices), 2)


cart_prices: list[float] = [65000.0, 799.0, 1499.0]
print(calculate_cart_total(cart_prices))
```

## 40. Sequence Pattern Matching

Structural pattern matching can inspect a list by its shape and values. This requires Python 3.10 or later.

```python
def describe_cart(cart):
    match cart:
        case []:
            return "empty cart"
        case [single_item]:
            return f"one item: {single_item}"
        case [first_item, *remaining_items]:
            return f"first={first_item}, remaining={len(remaining_items)}"


print(describe_cart([]))
print(describe_cart(["Mouse"]))
print(describe_cart(["Laptop", "Mouse", "Keyboard"]))
```

## 41. List Equality and Lexicographical Comparison

Lists compare element by element from left to right. Equality also requires the same length and corresponding values.

```python
cart_a = ["Laptop", "Mouse"]
cart_b = ["Laptop", "Mouse"]
cart_c = ["Laptop", "Keyboard"]

print(cart_a == cart_b)
print(cart_a is cart_b)
print(cart_c < cart_a)
```

## 42. List Performance Characteristics

Index access and append at the end are typically constant time. Searching, membership, insertion near the beginning, and removal by value are linear. Slicing creates a new list proportional to the slice size. Lists also over-allocate internal storage so repeated appends are efficient on average.

```python
large_catalog = [f"SKU-{number:05d}" for number in range(10_000)]

first_sku = large_catalog[0]
large_catalog.append("SKU-10000")
catalog_page = large_catalog[100:120]

print(first_sku)
print(len(large_catalog))
print(catalog_page[:3])
```

## 43. Complete Ecommerce Processing

```python
order_lines = [
    {"order_id": 1001, "sku": "P100", "product": "Laptop", "qty": 1, "unit_price": 65000, "status": "paid"},
    {"order_id": 1001, "sku": "P200", "product": "Mouse", "qty": 2, "unit_price": 799, "status": "paid"},
    {"order_id": 1002, "sku": "P300", "product": "Keyboard", "qty": 1, "unit_price": 1499, "status": "shipped"},
    {"order_id": 1003, "sku": "P400", "product": "Monitor", "qty": 1, "unit_price": 12999, "status": "cancelled"},
    {"order_id": 1004, "sku": "P200", "product": "Mouse", "qty": 3, "unit_price": 799, "status": "delivered"},
]

completed_lines = [line for line in order_lines if line["status"] != "cancelled"]
line_totals = [line["qty"] * line["unit_price"] for line in completed_lines]
revenue = sum(line_totals)
units_sold = sum(line["qty"] for line in completed_lines)
ranked_lines = sorted(
    completed_lines,
    key=lambda line: line["qty"] * line["unit_price"],
    reverse=True,
)

print(f"Revenue: ₹{revenue:,.2f}")
print(f"Units sold: {units_sold}")
for rank, line in enumerate(ranked_lines, start=1):
    total = line["qty"] * line["unit_price"]
    print(f"{rank}. {line['product']}: ₹{total:,.2f}")
```
