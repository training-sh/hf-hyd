# Regular Expressions

A regular expression, usually called regex, is a pattern used to search, validate, extract, split, or replace text. Python provides regular-expression operations through the `re` module.

## 1. Importing `re`

```python
import re
```

## 2. Literal Text

Ordinary letters and numbers match themselves. The pattern `Mouse` looks for those exact characters in that exact case.

```python
text = "The cart contains a Mouse and a Keyboard"
match = re.search(r"Mouse", text)

print(match)
print(match.group())
print(match.start())
print(match.end())
```

## 3. Raw Strings

Regex patterns are commonly written as raw strings with an `r` prefix. In a raw string, backslashes are passed to the regex engine without normal Python string escaping.

```python
pattern = r"\d+"
text = "Order 1001"

print(pattern)
print(re.search(pattern, text).group())
```

## 4. `search`

`re.search` looks anywhere in the text and returns the first match. It returns `None` when no match exists.

```python
text = "Order 1001 is paid"

print(re.search(r"paid", text))
print(re.search(r"cancelled", text))
```

## 5. Checking Whether a Match Exists

```python
text = "Order 1001 is paid"
match = re.search(r"paid", text)

if match:
    print("Status found")
else:
    print("Status not found")
```

## 6. `match`, `search`, and `fullmatch`

`match` checks only the beginning. `search` looks anywhere. `fullmatch` requires the complete text to match the pattern.

```python
text = "ORDER-1001 paid"

print(re.match(r"ORDER", text))
print(re.search(r"paid", text))
print(re.fullmatch(r"ORDER-1001 paid", text))
print(re.fullmatch(r"ORDER-1001", text))
```

## 7. `findall`

`findall` returns all non-overlapping matches as a list.

```python
text = "Orders 1001, 1002, and 1003 are ready"
order_numbers = re.findall(r"\d+", text)

print(order_numbers)
print(type(order_numbers))
```

## 8. `finditer`

`finditer` returns match objects one at a time. Each object contains the matched text and its position.

```python
text = "P100 P200 P300"

for match in re.finditer(r"P\d+", text):
    print(match.group(), match.start(), match.end(), match.span())
```

## 9. Any Character With `.`

A dot matches any character except a newline by default.

```python
text = "P100 P-00 P_00"
matches = re.findall(r"P.00", text)
print(matches)
```

## 10. Character Sets With `[]`

Square brackets match one character from the listed choices.

```python
text = "paid Paid said mail"
print(re.findall(r"[pP]aid", text))
print(re.findall(r"[aeiou]", "Laptop Mouse"))
```

## 11. Character Ranges

Ranges inside square brackets describe groups of characters. `[a-z]` matches one lowercase letter, `[A-Z]` one uppercase letter, and `[0-9]` one digit.

```python
text = "SKU P100 costs 799"

print(re.findall(r"[a-z]", text))
print(re.findall(r"[A-Z]", text))
print(re.findall(r"[0-9]", text))
print(re.findall(r"[A-Za-z]", text))
```

## 12. Negated Character Sets

A caret immediately after `[` means any character except the listed characters.

```python
text = "P100-Mouse"

print(re.findall(r"[^0-9]", text))
print(re.findall(r"[^A-Za-z]", text))
```

## 13. Digits: `\d` and `\D`

`\d` matches one digit. `\D` matches one non-digit.

```python
text = "Order 1001 costs 2499"

print(re.findall(r"\d", text))
print(re.findall(r"\d+", text))
print(re.findall(r"\D+", text))
```

## 14. Word Characters: `\w` and `\W`

`\w` matches Unicode letters, digits, and underscore. `\W` matches characters outside that group.

```python
text = "ORDER_1001: paid!"

print(re.findall(r"\w+", text))
print(re.findall(r"\W+", text))
```

## 15. Alphabetic and Alphanumeric Patterns

Explicit ranges are useful when only English ASCII letters should match. A plus sign means one or more characters.

```python
values = ["Mouse", "P100", "ORDER_1001", "799", "Laptop Stand"]

for value in values:
    alphabetic = bool(re.fullmatch(r"[A-Za-z]+", value))
    alphanumeric = bool(re.fullmatch(r"[A-Za-z0-9]+", value))
    print(value, alphabetic, alphanumeric)
```

## 16. Whitespace: `\s` and `\S`

`\s` matches whitespace such as a space, tab, or newline. `\S` matches a non-whitespace character.

```python
text = "Laptop Mouse\tKeyboard\nMonitor"

print(re.findall(r"\s", text))
print(re.findall(r"\S+", text))
```

## 17. Space, Tab, and Newline

A literal space matches a space. `\t` matches a tab, and `\n` matches a newline.

```python
text = "P100 Mouse\t2\nP200 Keyboard\t1"

print(re.findall(r" ", text))
print(re.findall(r"\t", text))
print(re.findall(r"\n", text))
print(re.findall(r"P\d+\s+[A-Za-z]+\s+\d+", text))
```

## 18. Beginning and End: `^` and `$`

`^` matches the beginning of the text and `$` matches the end. They match positions rather than visible characters.

```python
order_code = "ORDER-1001"

print(bool(re.search(r"^ORDER", order_code)))
print(bool(re.search(r"1001$", order_code)))
print(bool(re.fullmatch(r"ORDER-\d+", order_code)))
```

## 19. Absolute Beginning and End: `\A` and `\Z`

`\A` matches only the beginning of the complete string. `\Z` matches only the end of the complete string. Unlike `^` and `$`, their behavior is not changed by multiline mode.

```python
text = "ORDER-1001"

print(bool(re.search(r"\AORDER", text)))
print(bool(re.search(r"1001\Z", text)))
```

## 20. Zero or More: `*`

An asterisk allows the previous pattern to occur zero or more times.

```python
values = ["P", "P1", "P100", "PABC"]

for value in values:
    print(value, bool(re.fullmatch(r"P\d*", value)))
```

## 21. One or More: `+`

A plus sign requires the previous pattern at least once.

```python
values = ["P", "P1", "P100", "PABC"]

for value in values:
    print(value, bool(re.fullmatch(r"P\d+", value)))
```

## 22. Optional Pattern: `?`

A question mark allows the previous pattern zero or one time.

```python
values = ["color", "colour", "colouur"]

for value in values:
    print(value, bool(re.fullmatch(r"colou?r", value)))
```

## 23. Exact and Ranged Repetition

`{n}` requires exactly `n` repetitions. `{m,n}` allows from `m` through `n` repetitions. `{m,}` requires at least `m`.

```python
values = ["P1", "P100", "P1234", "P12345"]

for value in values:
    exactly_three = bool(re.fullmatch(r"P\d{3}", value))
    three_to_four = bool(re.fullmatch(r"P\d{3,4}", value))
    at_least_three = bool(re.fullmatch(r"P\d{3,}", value))
    print(value, exactly_three, three_to_four, at_least_three)
```

## 24. Alternation With `|`

The vertical bar means OR. Parentheses control which alternatives belong together.

```python
statuses = ["paid", "shipped", "cancelled", "pending"]
pattern = r"paid|shipped|delivered"

for status in statuses:
    print(status, bool(re.fullmatch(pattern, status)))
```

## 25. Capturing Groups

Parentheses group a pattern and capture its matched text. `group(0)` is the complete match, and numbered groups begin at `1`.

```python
text = "ORDER-1001"
match = re.fullmatch(r"([A-Z]+)-(\d+)", text)

print(match.group(0))
print(match.group(1))
print(match.group(2))
print(match.groups())
```

## 26. Named Groups

A named group uses `(?P<name>pattern)`. It can be accessed by name and converted to a dictionary.

```python
text = "ORDER-1001-PAID"
pattern = r"(?P<prefix>[A-Z]+)-(?P<order_id>\d+)-(?P<status>[A-Z]+)"
match = re.fullmatch(pattern, text)

print(match.group("order_id"))
print(match.groupdict())
```

## 27. Non-Capturing Groups

`(?:pattern)` groups alternatives or repetition without adding a captured group.

```python
text = "Payment: UPI"
match = re.search(r"Payment: (?:UPI|CARD|CASH)", text)

print(match.group())
print(match.groups())
```

## 28. Word Boundaries: `\b`

A word boundary matches the position between a word character and a non-word character. It prevents a short word from matching inside a longer word.

```python
text = "A cart contains a carton and a cart"

print(re.findall(r"cart", text))
print(re.findall(r"\bcart\b", text))
```

## 29. Escaping Special Characters

Characters such as `.`, `+`, `*`, `?`, `(`, `)`, `[`, `]`, `{`, `}`, `^`, `$`, `|`, and `\` have special meanings. A backslash makes them literal.

```python
text = "Price: 799.50 | Discount: 10%"

print(re.findall(r"\d+\.\d+", text))
print(re.findall(r"\d+%", text))
print(re.escape("799.50"))
```

## 30. Greedy and Non-Greedy Matching

Repetition is greedy by default and takes as much text as possible. Adding `?` after a quantifier makes it take as little as possible.

```python
text = "<product>Laptop</product><product>Mouse</product>"

print(re.findall(r"<product>.*</product>", text))
print(re.findall(r"<product>.*?</product>", text))
```

## 31. Splitting Text

`re.split` separates text using a pattern. This is useful when several delimiters are accepted.

```python
text = "P100, P200;P300  P400\tP500"
skus = re.split(r"[,;\s]+", text)
print(skus)
```

## 32. Replacing Text

`re.sub` replaces every match by default. `count` limits how many replacements are made.

```python
phone = "9876543210"
masked_phone = re.sub(r"\d(?=\d{4})", "*", phone)
print(masked_phone)

messy_name = "  Laptop    Stand  "
clean_name = re.sub(r"\s+", " ", messy_name).strip()
print(clean_name)
```

## 33. Replacement With Captured Groups

Captured groups can be reused in replacement text with `\1`, `\2`, or the clearer `\g<name>` form.

```python
date = "19-07-2026"
formatted_date = re.sub(
    r"(?P<day>\d{2})-(?P<month>\d{2})-(?P<year>\d{4})",
    r"\g<year>/\g<month>/\g<day>",
    date,
)
print(formatted_date)
```

## 34. Case-Insensitive Matching

`re.IGNORECASE`, also written as `re.I`, allows uppercase and lowercase forms to match.

```python
text = "Paid PAID paid PaId"
print(re.findall(r"paid", text, flags=re.IGNORECASE))
```

## 35. Multiline Mode

With `re.MULTILINE`, `^` and `$` match the beginning and end of each line, not only the complete string.

```python
text = "ORDER-1001 paid\nORDER-1002 shipped\nINVALID LINE"
order_lines = re.findall(r"^ORDER-\d+.*$", text, flags=re.MULTILINE)
print(order_lines)
```

## 36. Dot-All Mode

With `re.DOTALL`, a dot also matches newline characters.

```python
text = "START\nLaptop\nMouse\nEND"

print(re.search(r"START.*END", text))
print(re.search(r"START.*END", text, flags=re.DOTALL).group())
```

## 37. Compiling a Pattern

`re.compile` creates a reusable pattern object. It keeps a repeated pattern and its flags together.

```python
sku_pattern = re.compile(r"P\d{3}")

print(sku_pattern.fullmatch("P100"))
print(sku_pattern.fullmatch("P10"))
print(sku_pattern.findall("P100 P200 X300"))
```

## 38. Validating Common Ecommerce Values

```python
patterns = {
    "sku": re.compile(r"P\d{3}"),
    "order_id": re.compile(r"ORDER-\d{4}"),
    "pin_code": re.compile(r"[1-9]\d{5}"),
    "phone": re.compile(r"[6-9]\d{9}"),
    "date": re.compile(r"\d{2}-\d{2}-\d{4}"),
    "amount": re.compile(r"\d+(?:\.\d{1,2})?"),
}

values = {
    "sku": "P100",
    "order_id": "ORDER-1001",
    "pin_code": "600001",
    "phone": "9876543210",
    "date": "19-07-2026",
    "amount": "2499.50",
}

for field, value in values.items():
    print(field, bool(patterns[field].fullmatch(value)))
```

## 39. A Simple Email Pattern

This pattern handles common email forms for basic input checks. Complete email-address rules are more extensive than this pattern.

```python
email_pattern = re.compile(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}")
emails = ["asha@example.com", "bala.s@example.co.in", "invalid@", "no-at-sign.com"]

for email in emails:
    print(email, bool(email_pattern.fullmatch(email)))
```

## 40. Regex Results Are Strings

Regex extracts text. Convert a matched value when the program requires an integer, float, date, or another Python type.

```python
text = "quantity=3 price=799.50 in_stock=True"
quantity_text = re.search(r"quantity=(\d+)", text).group(1)
price_text = re.search(r"price=(\d+(?:\.\d+)?)", text).group(1)
stock_text = re.search(r"in_stock=(True|False)", text).group(1)

quantity = int(quantity_text)
price = float(price_text)
in_stock = stock_text == "True"

print(quantity, type(quantity))
print(price, type(price))
print(in_stock, type(in_stock))
```

## 41. Extracting an Order Record

```python
order_text = "ORDER-1001 | customer=Asha | sku=P200 | quantity=2 | amount=1598.00 | status=paid"
order_pattern = re.compile(
    r"ORDER-(?P<order_id>\d+) \| "
    r"customer=(?P<customer>[A-Za-z]+) \| "
    r"sku=(?P<sku>P\d{3}) \| "
    r"quantity=(?P<quantity>\d+) \| "
    r"amount=(?P<amount>\d+(?:\.\d{1,2})?) \| "
    r"status=(?P<status>paid|shipped|delivered|cancelled)"
)

match = order_pattern.fullmatch(order_text)
order = match.groupdict()
order["order_id"] = int(order["order_id"])
order["quantity"] = int(order["quantity"])
order["amount"] = float(order["amount"])

print(order)
```

## 42. Pattern Summary

| Pattern | Meaning |
|---|---|
| `abc` | Literal text |
| `.` | Any character except newline by default |
| `[abc]` | One listed character |
| `[^abc]` | One character not listed |
| `[a-z]` | One character in a range |
| `\d` / `\D` | Digit / non-digit |
| `\w` / `\W` | Word / non-word character |
| `\s` / `\S` | Whitespace / non-whitespace |
| `^` / `$` | Beginning / end of text or line |
| `\A` / `\Z` | Absolute beginning / end of string |
| `*` | Zero or more |
| `+` | One or more |
| `?` | Zero or one |
| `{m,n}` | From `m` through `n` repetitions |
| `a|b` | `a` or `b` |
| `(pattern)` | Capturing group |
| `(?:pattern)` | Non-capturing group |
| `(?P<name>pattern)` | Named capturing group |
| `\b` | Word boundary |
