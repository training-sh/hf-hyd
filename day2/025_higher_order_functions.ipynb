# Higher Order Functions

A higher order function is a function that receives another function as an argument, returns a function, or does both.

This is useful when the calculation or business rule must change dynamically.

## 1. Passing a Function as an Argument

The function `apply_and_sum` receives another function and applies it to every input value.

```python
import math

def identity(value):
    return value

def square(value):
    return value * value

def cube(value):
    return value * value * value

def square_root(value):
    return math.sqrt(value)

def apply_and_sum(operation, *values):
    total = 0

    for value in values:
        total = total + operation(value)

    return total
```

## 2. Using Different Computations

The same higher order function can run different calculations.

```python
print(apply_and_sum(identity, 1, 2, 3, 4, 5))
print(apply_and_sum(square, 1, 2, 3, 4, 5))
print(apply_and_sum(cube, 1, 2, 3, 4, 5))
print(apply_and_sum(square_root, 4, 9, 16, 25))
```

## 3. Returning Detailed Results

The function can also return the transformed values along with the final total.

```python
def apply_and_report(operation, *values):
    results = [operation(value) for value in values]
    return {
        "input_values": values,
        "computed_values": results,
        "total": sum(results)
    }

report = apply_and_report(square, 2, 4, 6, 8)

print(report)
```

## 4. Higher Order Function With Named Operation

A function can receive both a label and an operation.

```python
def calculate_metric(metric_name, operation, *values):
    result = apply_and_sum(operation, *values)
    return f"{metric_name}: {result:.2f}"

print(calculate_metric("Sum of values", identity, 10, 20, 30))
print(calculate_metric("Sum of squares", square, 10, 20, 30))
```

## 5. Ecommerce Dataset

The same order data can be processed using different business rules.

```python
orders = [
    {"order_id": 1001, "customer": "Anu", "amount": 2400, "status": "paid", "tier": "gold"},
    {"order_id": 1002, "customer": "Bala", "amount": 1200, "status": "cancelled", "tier": "silver"},
    {"order_id": 1003, "customer": "Charu", "amount": 5200, "status": "paid", "tier": "platinum"},
    {"order_id": 1004, "customer": "Deepa", "amount": 900, "status": "paid", "tier": "regular"},
    {"order_id": 1005, "customer": "Eshan", "amount": 3100, "status": "returned", "tier": "gold"}
]
```

## 6. Passing a Filter Function

The filter rule can be passed as a function.

```python
def is_paid_order(order):
    return order["status"] == "paid"

def is_high_value_order(order):
    return order["amount"] >= 3000

def filter_orders(orders, condition):
    return [order for order in orders if condition(order)]

paid_orders = filter_orders(orders, is_paid_order)
high_value_orders = filter_orders(orders, is_high_value_order)

print(paid_orders)
print(high_value_orders)
```

## 7. Passing a Value Extractor

The function can decide which value should be extracted from each order.

```python
def get_amount(order):
    return order["amount"]

def get_customer(order):
    return order["customer"]

def collect_values(records, extractor):
    return [extractor(record) for record in records]

print(collect_values(orders, get_amount))
print(collect_values(orders, get_customer))
```

## 8. Revenue Calculation With Dynamic Filter

The same function can calculate revenue using different order selection rules.

```python
def calculate_revenue(orders, condition):
    total = 0

    for order in orders:
        if condition(order):
            total = total + order["amount"]

    return total

paid_revenue = calculate_revenue(orders, is_paid_order)
high_value_revenue = calculate_revenue(orders, is_high_value_order)

print(f"Paid Revenue       : {paid_revenue:.2f}")
print(f"High Value Revenue : {high_value_revenue:.2f}")
```

## 9. Dynamic Discount Rules

Different discount functions can be passed based on campaign or customer type.

```python
def no_discount(amount):
    return 0

def festival_discount(amount):
    return amount * 0.10

def high_value_discount(amount):
    if amount >= 3000:
        return amount * 0.15
    return amount * 0.05

def calculate_payable(amount, discount_rule):
    discount = discount_rule(amount)
    return amount - discount

print(calculate_payable(2400, no_discount))
print(calculate_payable(2400, festival_discount))
print(calculate_payable(5200, high_value_discount))
```

## 10. Applying Discount Rules to Orders

A higher order function can apply a selected discount rule to every order.

```python
def apply_discount_to_orders(orders, discount_rule):
    updated_orders = []

    for order in orders:
        discount = discount_rule(order["amount"])
        payable = order["amount"] - discount

        updated_orders.append({
            "order_id": order["order_id"],
            "customer": order["customer"],
            "amount": order["amount"],
            "discount": discount,
            "payable": payable
        })

    return updated_orders

festival_orders = apply_discount_to_orders(orders, festival_discount)

print(festival_orders)
```

## 11. Returning a Function

A function can create and return another function.

```python
def create_percentage_discount(rate):
    def discount(amount):
        return amount * rate

    return discount

ten_percent_discount = create_percentage_discount(0.10)
twenty_percent_discount = create_percentage_discount(0.20)

print(ten_percent_discount(1000))
print(twenty_percent_discount(1000))
```

## 12. Discount Rule Factory

A factory function can return a rule configured with campaign values.

```python
def create_minimum_cart_discount(minimum_amount, rate):
    def discount(amount):
        if amount >= minimum_amount:
            return amount * rate
        return 0

    return discount

weekend_discount = create_minimum_cart_discount(2000, 0.12)
premium_discount = create_minimum_cart_discount(5000, 0.18)

print(calculate_payable(2400, weekend_discount))
print(calculate_payable(5200, premium_discount))
```

## 13. Tax Rule as a Function

Tax calculation can also be passed dynamically.

```python
def standard_tax(amount):
    return amount * 0.18

def reduced_tax(amount):
    return amount * 0.05

def calculate_invoice(amount, discount_rule, tax_rule):
    discount = discount_rule(amount)
    taxable_amount = amount - discount
    tax = tax_rule(taxable_amount)
    payable = taxable_amount + tax

    return {
        "amount": amount,
        "discount": discount,
        "tax": tax,
        "payable": payable
    }

invoice = calculate_invoice(5200, premium_discount, standard_tax)

print(invoice)
```

## 14. Report Builder With Dynamic Columns

A report function can receive column extractors.

```python
def build_report(records, *extractors):
    report_rows = []

    for record in records:
        row = [extractor(record) for extractor in extractors]
        report_rows.append(row)

    return report_rows

def get_order_id(order):
    return order["order_id"]

def get_status(order):
    return order["status"]

report = build_report(orders, get_order_id, get_customer, get_amount, get_status)

print(report)
```

## 15. Using Lambda Functions

`lambda` creates small anonymous functions.

```python
paid_order_ids = collect_values(
    filter_orders(orders, lambda order: order["status"] == "paid"),
    lambda order: order["order_id"]
)

print(paid_order_ids)
```

## 16. Mini Ecommerce Pricing Engine

This example combines filter rules, discount rules, tax rules, and formatted output.

```python
def process_orders(orders, order_filter, discount_rule, tax_rule):
    processed = []

    for order in orders:
        if not order_filter(order):
            continue

        invoice = calculate_invoice(order["amount"], discount_rule, tax_rule)
        processed.append({
            "order_id": order["order_id"],
            "customer": order["customer"],
            "amount": invoice["amount"],
            "discount": invoice["discount"],
            "tax": invoice["tax"],
            "payable": invoice["payable"]
        })

    return processed

processed_orders = process_orders(
    orders,
    is_paid_order,
    weekend_discount,
    standard_tax
)

print(f"{'Order':<8}{'Customer':<10}{'Amount':>10}{'Discount':>12}{'Tax':>10}{'Payable':>12}")
print("-" * 62)

for order in processed_orders:
    print(
        f"{order['order_id']:<8}"
        f"{order['customer']:<10}"
        f"{order['amount']:>10.2f}"
        f"{order['discount']:>12.2f}"
        f"{order['tax']:>10.2f}"
        f"{order['payable']:>12.2f}"
    )
```

## Quick Recap

- Higher order functions receive functions, return functions, or both.
- Passing a function allows logic to change without rewriting the main function.
- `*args` allows the same operation to work with many input values.
- Business rules such as filters, discounts, and taxes can be passed as functions.
- A function can return another function configured with values such as discount rate or minimum cart amount.
- Lambda functions are useful for small one-time rules.
