# For Expressions

For expressions create a new output from an existing sequence.

They are useful when values must be transformed, filtered, or collected into a new list, set, dictionary, or generator.

## 1. Basic List Expression

A list expression creates a new list.

```python
numbers = [1, 2, 3, 4, 5]

squares = [number * number for number in numbers]

print(squares)
```

## 2. Creating a List From `range()`

`range()` is often used when a numeric sequence is needed.

```python
order_numbers = [number for number in range(1001, 1006)]

print(order_numbers)
```

## 3. Filtering With `if`

`if` at the end keeps only matching values.

```python
numbers = [10, 15, 20, 25, 30, 35]

even_numbers = [number for number in numbers if number % 2 == 0]

print(even_numbers)
```

## 4. Transforming and Filtering Together

The output expression can transform each selected value.

```python
prices = [250, 900, 1200, 450, 3000]

discounted_prices = [price * 0.90 for price in prices if price >= 1000]

print(discounted_prices)
```

## 5. Conditional Output With `if...else`

`if...else` before `for` chooses what value to place in the output.

```python
prices = [250, 900, 1200, 450, 3000]

price_labels = ["high value" if price >= 1000 else "regular" for price in prices]

print(price_labels)
```

## 6. Text Transformation

For expressions are useful for cleaning and formatting text.

```python
product_names = [" laptop bag ", "MOUSE", "keyBoard", " usb cable "]

clean_names = [name.strip().title() for name in product_names]

print(clean_names)
```

## 7. Set Expression

A set expression creates unique values.

```python
order_statuses = ["paid", "paid", "cancelled", "returned", "paid", "cancelled"]

unique_statuses = {status for status in order_statuses}

print(unique_statuses)
```

## 8. Dictionary Expression

A dictionary expression creates key-value pairs.

```python
products = ["Mouse", "Keyboard", "Webcam"]
prices = [499, 1299, 2499]

product_price_map = {product: price for product, price in zip(products, prices)}

print(product_price_map)
```

## 9. Dictionary Expression With Filtering

Filtering can be added to dictionary expressions.

```python
product_prices = {
    "Mouse": 499,
    "Keyboard": 1299,
    "Webcam": 2499,
    "USB Cable": 199
}

premium_products = {product: price for product, price in product_prices.items() if price >= 1000}

print(premium_products)
```

## 10. Generator Expression

A generator expression produces values one by one.

```python
prices = [499, 1299, 2499, 199]

total_amount = sum(price for price in prices)

print(total_amount)
```

## 11. Any and All With Generator Expressions

`any()` checks whether at least one value matches. `all()` checks whether every value matches.

```python
cart_quantities = [1, 2, 1, 3]

has_bulk_item = any(quantity >= 3 for quantity in cart_quantities)
all_items_selected = all(quantity > 0 for quantity in cart_quantities)

print("Has bulk item:", has_bulk_item)
print("All items selected:", all_items_selected)
```

## 12. Ecommerce Cart Line Totals

For expressions can calculate values from dictionaries.

```python
cart = [
    {"item": "Mouse", "quantity": 2, "price": 499},
    {"item": "Keyboard", "quantity": 1, "price": 1299},
    {"item": "USB Cable", "quantity": 3, "price": 199}
]

line_totals = [row["quantity"] * row["price"] for row in cart]

print(line_totals)
print("Cart Total:", sum(line_totals))
```

## 13. Adding a Calculated Field

A new list of dictionaries can include original values and calculated values.

```python
cart = [
    {"item": "Mouse", "quantity": 2, "price": 499},
    {"item": "Keyboard", "quantity": 1, "price": 1299},
    {"item": "USB Cable", "quantity": 3, "price": 199}
]

cart_with_total = [
    {
        "item": row["item"],
        "quantity": row["quantity"],
        "price": row["price"],
        "line_total": row["quantity"] * row["price"]
    }
    for row in cart
]

print(cart_with_total)
```

## 14. Filtering Ecommerce Records

Filtering keeps only the records that match a business condition.

```python
orders = [
    {"order_id": 1001, "amount": 2400, "status": "paid"},
    {"order_id": 1002, "amount": 1200, "status": "cancelled"},
    {"order_id": 1003, "amount": 5200, "status": "paid"},
    {"order_id": 1004, "amount": 900, "status": "paid"}
]

paid_orders = [order for order in orders if order["status"] == "paid"]
high_value_order_ids = [order["order_id"] for order in orders if order["amount"] >= 3000]

print(paid_orders)
print(high_value_order_ids)
```

## 15. Conditional Labels for Orders

For expressions can create labels from business rules.

```python
orders = [
    {"order_id": 1001, "amount": 2400},
    {"order_id": 1002, "amount": 1200},
    {"order_id": 1003, "amount": 5200},
    {"order_id": 1004, "amount": 900}
]

order_labels = [
    "high value" if order["amount"] >= 3000 else "regular"
    for order in orders
]

print(order_labels)
```

## 16. Nested For Expression

Nested for expressions can flatten values from nested collections.

```python
orders = [
    {"order_id": 1001, "items": ["Mouse", "Keyboard"]},
    {"order_id": 1002, "items": ["USB Cable"]},
    {"order_id": 1003, "items": ["Webcam", "Mouse"]}
]

all_items = [item for order in orders for item in order["items"]]

print(all_items)
```

## 17. Nested For Expression With Filtering

Filtering can be applied after nested loops.

```python
orders = [
    {"order_id": 1001, "items": ["Mouse", "Keyboard"]},
    {"order_id": 1002, "items": ["USB Cable"]},
    {"order_id": 1003, "items": ["Webcam", "Mouse"]}
]

mouse_orders = [
    order["order_id"]
    for order in orders
    for item in order["items"]
    if item == "Mouse"
]

print(mouse_orders)
```

## 18. Matrix Style Nested Expression

Nested expressions can create table-like values.

```python
quantities = [1, 2, 3]
prices = [100, 200, 300]

amount_matrix = [[quantity * price for price in prices] for quantity in quantities]

print(amount_matrix)
```

## 19. Unique Items From Orders

A nested set expression can collect unique item names from many orders.

```python
orders = [
    {"order_id": 1001, "items": ["Mouse", "Keyboard"]},
    {"order_id": 1002, "items": ["USB Cable"]},
    {"order_id": 1003, "items": ["Webcam", "Mouse"]}
]

unique_items = {item for order in orders for item in order["items"]}

print(unique_items)
```

## 20. Grouping Values With a Dictionary Expression

Dictionary expressions can create lookup-style outputs.

```python
orders = [
    {"order_id": 1001, "customer": "Anu", "amount": 2400},
    {"order_id": 1002, "customer": "Bala", "amount": 1200},
    {"order_id": 1003, "customer": "Charu", "amount": 5200}
]

order_amounts = {order["order_id"]: order["amount"] for order in orders}

print(order_amounts)
```

## 21. Mini Ecommerce Summary

This example combines filtering, transformation, and aggregation.

```python
orders = [
    {"order_id": 1001, "customer": "Anu", "amount": 2400, "status": "paid"},
    {"order_id": 1002, "customer": "Bala", "amount": 1200, "status": "cancelled"},
    {"order_id": 1003, "customer": "Charu", "amount": 5200, "status": "paid"},
    {"order_id": 1004, "customer": "Deepa", "amount": 900, "status": "paid"}
]

paid_amounts = [order["amount"] for order in orders if order["status"] == "paid"]
paid_order_ids = [order["order_id"] for order in orders if order["status"] == "paid"]
customer_labels = {
    order["customer"]: "high value" if order["amount"] >= 3000 else "regular"
    for order in orders
    if order["status"] == "paid"
}

print("Paid Order IDs:", paid_order_ids)
print("Paid Revenue:", sum(paid_amounts))
print("Customer Labels:", customer_labels)
```

## Quick Recap

- List expressions create lists from existing sequences.
- Set expressions create unique output values.
- Dictionary expressions create key-value output.
- Generator expressions produce values one by one.
- `if` at the end filters values.
- `if...else` before `for` chooses the output value.
- Nested for expressions process nested collections.
- For expressions are useful for transforming ecommerce carts, orders, prices, statuses, and reports.
