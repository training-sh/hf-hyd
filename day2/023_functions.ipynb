# Functions

Functions group reusable logic under a name.

In ecommerce applications, functions are useful for calculating totals, tax, discounts, delivery charges, invoices, and order summaries.

## 1. Basic Function

A function is created using `def`.

```python
def show_store_name():
    print("Online Retail Store")

show_store_name()
```

## 2. Function With Parameters

Parameters allow values to be passed into a function.

```python
def show_product(product_name, price):
    print(f"Product: {product_name}")
    print(f"Price  : {price:.2f}")

show_product("Mouse", 499)
```

## 3. Function With Return Value

`return` sends a value back from the function.

```python
def calculate_line_total(quantity, price):
    return quantity * price

line_total = calculate_line_total(3, 499)

print(f"Line Total: {line_total:.2f}")
```

## 4. Returning an Integer

A function can return an integer value.

```python
def count_cart_items(cart):
    return len(cart)

cart = ["Mouse", "Keyboard", "USB Cable"]
item_count = count_cart_items(cart)

print(item_count)
print(type(item_count))
```

## 5. Returning a Float

Money calculations often return float values.

```python
def calculate_tax(amount, tax_rate):
    return amount * tax_rate

tax_amount = calculate_tax(2500, 0.18)

print(tax_amount)
print(type(tax_amount))
```

## 6. Returning a String

A function can return a message or label.

```python
def get_order_label(amount):
    if amount >= 3000:
        return "high value"
    return "regular"

label = get_order_label(5200)

print(label)
print(type(label))
```

## 7. Returning a Boolean

Boolean return values are useful for checks and validations.

```python
def is_free_delivery(cart_total):
    return cart_total >= 1000

result = is_free_delivery(1250)

print(result)
print(type(result))
```

## 8. Returning a List

A function can return multiple values inside a list.

```python
def get_paid_order_ids(orders):
    return [order["order_id"] for order in orders if order["status"] == "paid"]

orders = [
    {"order_id": 1001, "status": "paid"},
    {"order_id": 1002, "status": "cancelled"},
    {"order_id": 1003, "status": "paid"}
]

paid_order_ids = get_paid_order_ids(orders)

print(paid_order_ids)
print(type(paid_order_ids))
```

## 9. Returning a Dictionary

Dictionaries are useful when a function returns named details.

```python
def build_invoice_summary(subtotal, tax, discount):
    payable = subtotal + tax - discount
    return {
        "subtotal": subtotal,
        "tax": tax,
        "discount": discount,
        "payable": payable
    }

summary = build_invoice_summary(2500, 450, 200)

print(summary)
print(type(summary))
```

## 10. Returning Multiple Values

A function can return more than one value as a tuple.

```python
def calculate_discount(cart_total, discount_rate):
    discount = cart_total * discount_rate
    payable = cart_total - discount
    return discount, payable

discount_amount, payable_amount = calculate_discount(4000, 0.10)

print(f"Discount: {discount_amount:.2f}")
print(f"Payable : {payable_amount:.2f}")
```

## 11. Default Arguments

Default values are used when an argument is not provided.

```python
def calculate_delivery_charge(cart_total, free_delivery_limit=1000, delivery_charge=50):
    if cart_total >= free_delivery_limit:
        return 0
    return delivery_charge

print(calculate_delivery_charge(750))
print(calculate_delivery_charge(1250))
print(calculate_delivery_charge(750, 1500, 80))
```

## 12. Named Arguments

Named arguments make function calls easier to read.

```python
def calculate_order_amount(subtotal, tax, discount):
    return subtotal + tax - discount

payable = calculate_order_amount(discount=200, subtotal=2500, tax=450)

print(f"Payable Amount: {payable:.2f}")
```

## 13. Positional and Named Arguments Together

Positional arguments must come before named arguments.

```python
def calculate_final_amount(subtotal, tax_rate=0.18, discount_rate=0.0):
    tax = subtotal * tax_rate
    discount = subtotal * discount_rate
    return subtotal + tax - discount

amount = calculate_final_amount(3000, discount_rate=0.10)

print(f"Final Amount: {amount:.2f}")
```

## 14. Variable Number of Arguments With `*args`

`*args` receives multiple positional values as a tuple.

```python
def calculate_cart_total(*prices):
    return sum(prices)

total = calculate_cart_total(499, 1299, 199, 899)

print(f"Cart Total: {total:.2f}")
```

## 15. `*args` With Looping

`*args` can be processed like any other sequence.

```python
def show_cart_items(*items):
    for index, item in enumerate(items, start=1):
        print(f"{index}. {item}")

show_cart_items("Mouse", "Keyboard", "USB Cable")
```

## 16. Keyword Arguments With `**kwargs`

`**kwargs` receives multiple named values as a dictionary.

```python
def show_order_details(**order):
    for key, value in order.items():
        print(f"{key}: {value}")

show_order_details(order_id=1001, customer="Anu", status="paid", amount=2400)
```

## 17. `**kwargs` With Defaults Inside Function

`get()` can provide default values when a keyword is missing.

```python
def create_shipping_label(**details):
    customer = details.get("customer", "Guest")
    city = details.get("city", "Not Provided")
    delivery_type = details.get("delivery_type", "standard")

    return f"Customer: {customer}, City: {city}, Delivery: {delivery_type}"

label = create_shipping_label(customer="Anu", city="Chennai")

print(label)
```

## 18. Mixed Arguments

A function can combine normal parameters, default arguments, `*args`, and `**kwargs`.

```python
def create_order_summary(customer, tax_rate=0.18, *prices, **details):
    subtotal = sum(prices)
    tax = subtotal * tax_rate
    discount = details.get("discount", 0)
    delivery_charge = details.get("delivery_charge", 0)
    payable = subtotal + tax - discount + delivery_charge

    return {
        "customer": customer,
        "subtotal": subtotal,
        "tax": tax,
        "discount": discount,
        "delivery_charge": delivery_charge,
        "payable": payable
    }

summary = create_order_summary("Anu", 0.18, 499, 1299, 199, discount=100, delivery_charge=40)

print(summary)
```

## 19. Function Calling Another Function

Larger calculations can be split into smaller functions.

```python
def get_subtotal(cart):
    return sum(row["quantity"] * row["price"] for row in cart)

def get_tax(subtotal, tax_rate=0.18):
    return subtotal * tax_rate

def get_payable(subtotal, tax, discount=0):
    return subtotal + tax - discount

cart = [
    {"item": "Mouse", "quantity": 2, "price": 499},
    {"item": "Keyboard", "quantity": 1, "price": 1299},
    {"item": "USB Cable", "quantity": 3, "price": 199}
]

subtotal = get_subtotal(cart)
tax = get_tax(subtotal)
payable = get_payable(subtotal, tax, discount=150)

print(f"Subtotal: {subtotal:.2f}")
print(f"Tax     : {tax:.2f}")
print(f"Payable : {payable:.2f}")
```

## 20. Function With Input and Output

A function can collect input, calculate values, and print a result.

```python
def print_product_bill():
    product = input("Enter product name: ")
    quantity = int(input("Enter quantity: "))
    price = float(input("Enter price: "))

    total = quantity * price

    print("\nBILL")
    print("-" * 30)
    print(f"Product : {product}")
    print(f"Quantity: {quantity}")
    print(f"Price   : {price:.2f}")
    print(f"Total   : {total:.2f}")

print_product_bill()
```

## 21. Mini Invoice Program

This example combines return values, default arguments, named arguments, and formatted output.

```python
def calculate_invoice(cart, tax_rate=0.18, discount_rate=0.0, delivery_charge=0):
    subtotal = sum(row["quantity"] * row["price"] for row in cart)
    tax = subtotal * tax_rate
    discount = subtotal * discount_rate
    payable = subtotal + tax - discount + delivery_charge

    return {
        "subtotal": subtotal,
        "tax": tax,
        "discount": discount,
        "delivery_charge": delivery_charge,
        "payable": payable
    }

cart = [
    {"item": "Mouse", "quantity": 2, "price": 499},
    {"item": "Keyboard", "quantity": 1, "price": 1299},
    {"item": "USB Cable", "quantity": 3, "price": 199}
]

invoice = calculate_invoice(cart, discount_rate=0.10, delivery_charge=40)

print("INVOICE SUMMARY")
print("=" * 35)
print(f"Subtotal        : {invoice['subtotal']:>10.2f}")
print(f"Tax             : {invoice['tax']:>10.2f}")
print(f"Discount        : {invoice['discount']:>10.2f}")
print(f"Delivery Charge : {invoice['delivery_charge']:>10.2f}")
print("-" * 35)
print(f"Payable         : {invoice['payable']:>10.2f}")
```

## Quick Recap

- Functions are created with `def`.
- Parameters receive input values.
- `return` sends output from a function.
- Functions can return integers, floats, strings, Booleans, lists, tuples, and dictionaries.
- Default arguments provide fallback values.
- Named arguments make calls more readable.
- `*args` receives multiple positional arguments.
- `**kwargs` receives multiple keyword arguments.
- Mixed arguments can combine normal parameters, defaults, `*args`, and `**kwargs`.
