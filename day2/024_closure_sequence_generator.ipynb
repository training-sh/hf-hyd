# Closure Sequence Generator

A closure is created when an inner function remembers values from its outer function even after the outer function has finished running.

Closures are useful when a function needs to keep private state.

## 1. Basic Closure

The inner function can use values from the outer function.

```python
def create_message(prefix):
    def show_message(value):
        return f"{prefix}: {value}"

    return show_message

order_message = create_message("Order ID")

print(order_message(1001))
print(order_message(1002))
```

## 2. Sequence Generator Closure

This function creates a sequence generator using a seed value and a step value.

```python
def create_sequence_generator(seed, step):
    current_value = seed

    def next_value():
        nonlocal current_value

        value = current_value
        current_value = current_value + step
        return value

    return next_value
```

## 3. Sequence Starting From 0 With Step 1

This sequence can be used for simple running numbers.

```python
seq_from_zero = create_sequence_generator(seed=0, step=1)

print(seq_from_zero())
print(seq_from_zero())
print(seq_from_zero())
print(seq_from_zero())
print(seq_from_zero())
```

## 4. Sequence Starting From 1000 With Step 5

This sequence can be used for order numbers or invoice numbers.

```python
seq_from_1000 = create_sequence_generator(seed=1000, step=5)

print(seq_from_1000())
print(seq_from_1000())
print(seq_from_1000())
print(seq_from_1000())
print(seq_from_1000())
```

## 5. Sequence Starting From 11111 With Step 2

This sequence can be used when generated values must start from a custom business number.

```python
seq_from_11111 = create_sequence_generator(seed=11111, step=2)

print(seq_from_11111())
print(seq_from_11111())
print(seq_from_11111())
print(seq_from_11111())
print(seq_from_11111())
```

## 6. Independent Sequence Generators

Each closure keeps its own private value.

```python
cart_line_seq = create_sequence_generator(seed=0, step=1)
order_seq = create_sequence_generator(seed=1000, step=5)
invoice_seq = create_sequence_generator(seed=11111, step=2)

print(f"Cart Line ID : {cart_line_seq()}")
print(f"Order ID     : {order_seq()}")
print(f"Invoice ID   : {invoice_seq()}")

print(f"Cart Line ID : {cart_line_seq()}")
print(f"Order ID     : {order_seq()}")
print(f"Invoice ID   : {invoice_seq()}")
```

## 7. Using the Generator in Ecommerce Data

A closure can assign generated values while preparing records.

```python
order_id_generator = create_sequence_generator(seed=1000, step=5)

customers = ["Anu", "Bala", "Charu"]

orders = []

for customer in customers:
    order = {
        "order_id": order_id_generator(),
        "customer": customer,
        "status": "created"
    }
    orders.append(order)

print(orders)
```

## 8. Formatted Sequence Output

Generated values can be displayed with labels and formatting.

```python
invoice_number = create_sequence_generator(seed=11111, step=2)

for _ in range(5):
    print(f"INV-{invoice_number()}")
```

## Quick Recap

- A closure is an inner function that remembers values from an outer function.
- The sequence generator remembers its current value.
- `nonlocal` allows the inner function to update a value from the outer function.
- Each generated function keeps its own independent sequence.
- Seed controls the starting value.
- Step controls how much the value increases each time.
