# Decision Making

Programs often need to choose different actions based on data.

In ecommerce applications, decisions are used for discounts, delivery charges, stock checks, payment status, customer tiers, and return eligibility.

## 1. Simple `if` Statement

An `if` statement runs a block of code only when a condition is true.

```python
cart_total = 1250

if cart_total >= 1000:
    print("Free delivery applied")

print("Checkout page loaded")
```

## 2. `if` and `else`

`else` runs when the `if` condition is false.

```python
stock_available = 0

if stock_available > 0:
    print("Item is available")
else:
    print("Item is out of stock")
```

## 3. `if`, `elif`, and `else`

`elif` is used when there are multiple possible conditions.

```python
order_status = "shipped"

if order_status == "placed":
    print("Order received")
elif order_status == "packed":
    print("Order packed and ready for shipment")
elif order_status == "shipped":
    print("Order is on the way")
elif order_status == "delivered":
    print("Order delivered")
else:
    print("Unknown order status")
```

## 4. Comparison Operators

Conditions usually compare values.

```python
cart_total = 750
minimum_order_value = 500

print(cart_total > minimum_order_value)
print(cart_total >= minimum_order_value)
print(cart_total == minimum_order_value)
print(cart_total != minimum_order_value)
```

## 5. Discount Based on Cart Value

Different cart ranges can receive different discount rates.

```python
cart_total = 4200

if cart_total >= 5000:
    discount_rate = 0.20
elif cart_total >= 3000:
    discount_rate = 0.15
elif cart_total >= 1000:
    discount_rate = 0.10
else:
    discount_rate = 0.00

discount_amount = cart_total * discount_rate
payable_amount = cart_total - discount_amount

print(f"Cart Total      : {cart_total:.2f}")
print(f"Discount Rate   : {discount_rate:.0%}")
print(f"Discount Amount : {discount_amount:.2f}")
print(f"Payable Amount  : {payable_amount:.2f}")
```

## 6. Logical `and`

`and` requires all conditions to be true.

```python
cart_total = 1800
customer_logged_in = True

if cart_total >= 1000 and customer_logged_in:
    print("Coupon can be applied")
else:
    print("Coupon cannot be applied")
```

## 7. Logical `or`

`or` requires at least one condition to be true.

```python
customer_tier = "gold"
cart_total = 650

if customer_tier == "gold" or cart_total >= 1000:
    print("Priority support enabled")
else:
    print("Standard support enabled")
```

## 8. Logical `not`

`not` reverses a Boolean value.

```python
payment_failed = False

if not payment_failed:
    print("Payment successful")
else:
    print("Payment failed")
```

## 9. Nested Conditions

A condition can be placed inside another condition.

```python
stock_available = 12
requested_quantity = 3
payment_status = "success"

if stock_available >= requested_quantity:
    print("Stock check passed")

    if payment_status == "success":
        print("Order confirmed")
    else:
        print("Payment pending")
else:
    print("Requested quantity is not available")
```

## 10. Delivery Charge Calculation

Delivery charge can depend on cart value and delivery location.

```python
cart_total = 850
delivery_zone = "local"

if cart_total >= 1000:
    delivery_charge = 0
elif delivery_zone == "local":
    delivery_charge = 40
else:
    delivery_charge = 80

final_amount = cart_total + delivery_charge

print(f"Cart Total      : {cart_total:.2f}")
print(f"Delivery Charge : {delivery_charge:.2f}")
print(f"Final Amount    : {final_amount:.2f}")
```

## 11. Coupon Validation

A coupon may require a valid code, minimum cart value, and active customer account.

```python
coupon_code = "SAVE10"
cart_total = 1500
account_active = True

if coupon_code == "SAVE10" and cart_total >= 1000 and account_active:
    print("Coupon applied successfully")
else:
    print("Coupon is not applicable")
```

## 12. Customer Tier Benefits

Customer tiers can control discounts and delivery benefits.

```python
customer_tier = "platinum"
cart_total = 2400

if customer_tier == "platinum":
    discount_rate = 0.15
    delivery_charge = 0
elif customer_tier == "gold":
    discount_rate = 0.10
    delivery_charge = 0 if cart_total >= 1000 else 40
elif customer_tier == "silver":
    discount_rate = 0.05
    delivery_charge = 40
else:
    discount_rate = 0.00
    delivery_charge = 60

discount = cart_total * discount_rate
payable = cart_total - discount + delivery_charge

print(f"Customer Tier   : {customer_tier}")
print(f"Cart Total      : {cart_total:.2f}")
print(f"Discount        : {discount:.2f}")
print(f"Delivery Charge : {delivery_charge:.2f}")
print(f"Payable Amount  : {payable:.2f}")
```

## 13. Return Eligibility

Return rules can check order status, number of days, and product category.

```python
order_status = "delivered"
days_since_delivery = 5
category = "electronics"

if order_status == "delivered":
    if days_since_delivery <= 7 and category != "grocery":
        print("Return is allowed")
    else:
        print("Return period is closed or category is not returnable")
else:
    print("Return is available only after delivery")
```

## 14. Payment Method Decision

Different payment methods may have different handling rules.

```python
payment_method = "cod"
cart_total = 3200

if payment_method == "card":
    print("Redirecting to card payment gateway")
elif payment_method == "upi":
    print("Generating UPI payment request")
elif payment_method == "cod" and cart_total <= 5000:
    print("Cash on delivery is available")
elif payment_method == "cod" and cart_total > 5000:
    print("Cash on delivery is not available for this order value")
else:
    print("Invalid payment method")
```

## 15. Mini Checkout Decision Program

This example combines conditions, nested conditions, and logical operators.

```python
cart_total = float(input("Enter cart total: "))
stock_available = int(input("Enter available stock: "))
requested_quantity = int(input("Enter requested quantity: "))
customer_tier = input("Enter customer tier: ").lower()
payment_status = input("Enter payment status: ").lower()

if stock_available >= requested_quantity:
    if payment_status == "success":
        if customer_tier == "gold" or customer_tier == "platinum":
            discount_rate = 0.10
        elif cart_total >= 3000:
            discount_rate = 0.05
        else:
            discount_rate = 0.00

        discount = cart_total * discount_rate
        payable = cart_total - discount

        print("\nORDER SUMMARY")
        print("=" * 35)
        print(f"Cart Total    : {cart_total:>10.2f}")
        print(f"Discount Rate : {discount_rate:>10.0%}")
        print(f"Discount      : {discount:>10.2f}")
        print(f"Payable       : {payable:>10.2f}")
        print("Order confirmed")
    else:
        print("Order cannot be confirmed because payment is not successful")
else:
    print("Order cannot be confirmed because requested quantity is unavailable")
```

## Quick Recap

- `if` runs code when a condition is true.
- `else` runs when the `if` condition is false.
- `elif` handles multiple possible conditions.
- Nested conditions place one decision inside another decision.
- `and` requires all conditions to be true.
- `or` requires at least one condition to be true.
- `not` reverses a Boolean value.
- Ecommerce applications use decisions for carts, coupons, payment, delivery, stock, and returns.
