# Looping Constructs

Loops are used when the same action must run multiple times.

In ecommerce applications, loops are useful for processing cart items, calculating totals, checking inventory, retrying payment, reading order records, and preparing reports.

## 1. Basic `for` Loop

A `for` loop is commonly used to process each item in a collection.

```python
cart_items = ["Laptop Bag", "Mouse", "Keyboard", "USB Cable"]

for item in cart_items:
    print(item)
```

## 2. Looping Through Prices

A loop can calculate a running total.

```python
prices = [899.00, 499.50, 1299.00, 199.00]
cart_total = 0

for price in prices:
    cart_total = cart_total + price

print(f"Cart Total: {cart_total:.2f}")
```

## 3. `for` Loop With `range()`

`range()` generates a sequence of numbers.

```python
for order_number in range(1, 6):
    print(f"Processing order #{order_number}")
```

## 4. `range()` With Start, Stop, and Step

The step value controls how the number changes.

```python
for discount in range(5, 31, 5):
    print(f"Discount option: {discount}%")
```

## 5. Looping With `enumerate()`

`enumerate()` gives both position and value.

```python
cart_items = ["Laptop Bag", "Mouse", "Keyboard"]

for index, item in enumerate(cart_items, start=1):
    print(f"{index}. {item}")
```

## 6. Looping Through a Dictionary

Dictionaries are useful for item names and prices.

```python
product_prices = {
    "Laptop Bag": 899.00,
    "Mouse": 499.50,
    "Keyboard": 1299.00
}

for product, price in product_prices.items():
    print(f"{product:<15} {price:>8.2f}")
```

## 7. Calculating Quantity Wise Total

A loop can process multiple fields from each order line.

```python
cart = [
    {"item": "Laptop Bag", "quantity": 1, "price": 899.00},
    {"item": "Mouse", "quantity": 2, "price": 499.50},
    {"item": "Keyboard", "quantity": 1, "price": 1299.00}
]

cart_total = 0

print(f"{'Item':<15}{'Qty':>5}{'Price':>10}{'Total':>10}")
print("-" * 40)

for row in cart:
    line_total = row["quantity"] * row["price"]
    cart_total = cart_total + line_total
    print(f"{row['item']:<15}{row['quantity']:>5}{row['price']:>10.2f}{line_total:>10.2f}")

print("-" * 40)
print(f"{'Cart Total':<30}{cart_total:>10.2f}")
```

## 8. Basic `while` Loop

A `while` loop continues as long as a condition is true.

```python
pending_orders = 5

while pending_orders > 0:
    print(f"Packing order. Pending orders: {pending_orders}")
    pending_orders = pending_orders - 1

print("All pending orders packed")
```

## 9. `while` Loop With Payment Retry

A `while` loop is useful when the number of attempts is limited.

```python
payment_success = False
attempt = 1
max_attempts = 3

while not payment_success and attempt <= max_attempts:
    print(f"Payment attempt {attempt}")

    if attempt == 2:
        payment_success = True
        print("Payment successful")
    else:
        print("Payment failed")

    attempt = attempt + 1
```

## 10. Python Style `do...while` Pattern

Python does not have a built-in `do...while` loop. A similar pattern can be written using `while True` and `break`.

```python
coupon_attempts = 0

while True:
    coupon_attempts = coupon_attempts + 1
    coupon_code = input("Enter coupon code: ")

    if coupon_code == "SAVE10":
        print("Coupon applied")
        break

    if coupon_attempts == 3:
        print("Maximum coupon attempts reached")
        break

    print("Invalid coupon. Try again.")
```

## 11. `break`

`break` stops a loop immediately.

```python
order_ids = [1001, 1002, 1003, 1004, 1005]
search_order_id = 1003

for order_id in order_ids:
    print(f"Checking order {order_id}")

    if order_id == search_order_id:
        print("Order found")
        break
```

## 12. `continue`

`continue` skips the current loop cycle and moves to the next one.

```python
orders = [
    {"order_id": 1001, "status": "paid"},
    {"order_id": 1002, "status": "cancelled"},
    {"order_id": 1003, "status": "paid"},
    {"order_id": 1004, "status": "returned"}
]

for order in orders:
    if order["status"] == "cancelled":
        continue

    print(f"Processing order {order['order_id']}")
```

## 13. `pass`

`pass` is used when a block is needed syntactically, but no action is required yet.

```python
order_statuses = ["paid", "packed", "refunded"]

for status in order_statuses:
    if status == "refunded":
        pass
    else:
        print(f"Status checked: {status}")
```

## 14. Loop With `else`

The `else` block after a loop runs when the loop completes without `break`.

```python
product_ids = [501, 502, 503, 504]
search_product_id = 509

for product_id in product_ids:
    if product_id == search_product_id:
        print("Product found")
        break
else:
    print("Product not found")
```

## 15. Nested Loops

Nested loops are useful when one collection contains another collection.

```python
orders = [
    {"order_id": 1001, "items": ["Mouse", "Keyboard"]},
    {"order_id": 1002, "items": ["Laptop Bag"]},
    {"order_id": 1003, "items": ["USB Cable", "Mouse", "Webcam"]}
]

for order in orders:
    print(f"Order ID: {order['order_id']}")

    for item in order["items"]:
        print(f"    Item: {item}")
```

## 16. Inventory Check With Loop

A loop can check stock availability for every cart item.

```python
cart = [
    {"item": "Mouse", "quantity": 2},
    {"item": "Keyboard", "quantity": 1},
    {"item": "Webcam", "quantity": 3}
]

inventory = {
    "Mouse": 10,
    "Keyboard": 0,
    "Webcam": 5
}

for row in cart:
    item = row["item"]
    requested_quantity = row["quantity"]
    available_quantity = inventory[item]

    if available_quantity >= requested_quantity:
        print(f"{item}: available")
    else:
        print(f"{item}: insufficient stock")
```

## 17. Counting Orders by Status

Loops are often used to count records by category.

```python
orders = ["paid", "paid", "cancelled", "returned", "paid", "cancelled"]

paid_count = 0
cancelled_count = 0
returned_count = 0

for status in orders:
    if status == "paid":
        paid_count = paid_count + 1
    elif status == "cancelled":
        cancelled_count = cancelled_count + 1
    elif status == "returned":
        returned_count = returned_count + 1

print(f"Paid Orders      : {paid_count}")
print(f"Cancelled Orders : {cancelled_count}")
print(f"Returned Orders  : {returned_count}")
```

## 18. Mini Order Processing Program

This example combines loops, conditions, calculations, and formatted output.

```python
orders = [
    {"order_id": 1001, "customer": "Anu", "amount": 2400, "status": "paid"},
    {"order_id": 1002, "customer": "Bala", "amount": 1200, "status": "cancelled"},
    {"order_id": 1003, "customer": "Charu", "amount": 5200, "status": "paid"},
    {"order_id": 1004, "customer": "Deepa", "amount": 900, "status": "paid"}
]

total_revenue = 0

print(f"{'Order ID':<10}{'Customer':<12}{'Amount':>10}{'Status':>12}")
print("-" * 44)

for order in orders:
    if order["status"] == "cancelled":
        continue

    total_revenue = total_revenue + order["amount"]
    print(f"{order['order_id']:<10}{order['customer']:<12}{order['amount']:>10.2f}{order['status']:>12}")

print("-" * 44)
print(f"{'Total Revenue':<32}{total_revenue:>12.2f}")
```

## Quick Recap

- `for` loops are useful for processing each value in a collection.
- `while` loops continue while a condition is true.
- Python does not have a native `do...while`, but `while True` with `break` can create a similar pattern.
- `break` stops a loop immediately.
- `continue` skips the current loop cycle.
- `pass` leaves a block empty temporarily.
- Loop `else` runs when the loop finishes without `break`.
- Nested loops handle data inside data, such as orders containing multiple items.
