# Lambda Functions

A lambda function is a small anonymous function written in a single expression.

Lambda functions are useful for short calculations, inline logic, sorting, filtering, and passing simple behavior into another function.

## 1. Lambda With No Parameter

A lambda can be created without parameters.

```python
store_name = lambda: "Online Retail Store"

print(store_name())
```

## 2. Lambda With One Parameter

A one-parameter lambda receives one input value.

```python
add_tax = lambda amount: amount + (amount * 0.18)

print(add_tax(1000))
```

## 3. Lambda With Two Parameters

A two-parameter lambda receives two input values.

```python
calculate_line_total = lambda quantity, price: quantity * price

print(calculate_line_total(3, 499))
```

## 4. Lambda With More Than Two Parameters

A lambda can accept multiple parameters, but it should still stay simple.

```python
calculate_payable = lambda subtotal, tax, discount: subtotal + tax - discount

print(calculate_payable(2500, 450, 200))
```

## 5. Lambda Returning a Boolean

Lambda functions are often used for simple checks.

```python
is_free_delivery = lambda cart_total: cart_total >= 1000

print(is_free_delivery(750))
print(is_free_delivery(1250))
```

## 6. Lambda With Conditional Expression

A lambda can include `if...else` as an expression.

```python
order_label = lambda amount: "high value" if amount >= 3000 else "regular"

print(order_label(1200))
print(order_label(5200))
```

## 7. Normal Function and Lambda Function

Both examples below calculate the same result.

```python
def calculate_discount_def(amount):
    return amount * 0.10

calculate_discount_lambda = lambda amount: amount * 0.10

print(calculate_discount_def(2000))
print(calculate_discount_lambda(2000))
```

## 8. Reusable Lambda Assigned to a Variable

A lambda can be assigned to a variable and reused.

```python
apply_festival_discount = lambda amount: amount - (amount * 0.10)

print(apply_festival_discount(1500))
print(apply_festival_discount(2500))
print(apply_festival_discount(5000))
```

## 9. Inline Lambda

An inline lambda is used directly without saving it under a separate name.

```python
def calculate_amount(amount, operation):
    return operation(amount)

print(calculate_amount(2000, lambda amount: amount * 0.90))
print(calculate_amount(2000, lambda amount: amount * 1.18))
```

## 10. Lambda With `map()`

`map()` applies a function to every item.

```python
prices = [499, 1299, 2499, 199]

prices_with_tax = list(map(lambda price: price * 1.18, prices))

print(prices_with_tax)
```

## 11. Lambda With `filter()`

`filter()` keeps values where the function returns `True`.

```python
prices = [499, 1299, 2499, 199, 5200]

premium_prices = list(filter(lambda price: price >= 1000, prices))

print(premium_prices)
```

## 12. Lambda With `sorted()`

Lambda functions are commonly used as sorting keys.

```python
orders = [
    {"order_id": 1001, "customer": "Anu", "amount": 2400},
    {"order_id": 1002, "customer": "Bala", "amount": 1200},
    {"order_id": 1003, "customer": "Charu", "amount": 5200},
    {"order_id": 1004, "customer": "Deepa", "amount": 900}
]

orders_by_amount = sorted(orders, key=lambda order: order["amount"])

print(orders_by_amount)
```

## 13. Sorting in Descending Order

`reverse=True` sorts the result in descending order.

```python
orders_by_amount_desc = sorted(orders, key=lambda order: order["amount"], reverse=True)

print(orders_by_amount_desc)
```

## 14. Lambda With List Expression

A lambda can be called inside a list expression.

```python
discount_rule = lambda amount: amount * 0.10 if amount >= 1000 else 0

discounts = [discount_rule(order["amount"]) for order in orders]

print(discounts)
```

## 15. Lambda Passed to a Custom Function

Lambda functions are useful when a custom function expects another function.

```python
def select_orders(orders, condition):
    return [order for order in orders if condition(order)]

paid_orders = select_orders(
    [
        {"order_id": 1001, "status": "paid", "amount": 2400},
        {"order_id": 1002, "status": "cancelled", "amount": 1200},
        {"order_id": 1003, "status": "paid", "amount": 5200}
    ],
    lambda order: order["status"] == "paid"
)

print(paid_orders)
```

## 16. Lambda With Two Inputs in `map()`

`map()` can read values from multiple sequences.

```python
quantities = [2, 1, 3]
prices = [499, 1299, 199]

line_totals = list(map(lambda quantity, price: quantity * price, quantities, prices))

print(line_totals)
```

## 17. Anonymous Function Behavior

A lambda does not need a function name when it is used immediately.

```python
result = (lambda amount: amount * 0.90)(2000)

print(result)
```

## 18. Lambda Limit: Single Expression

Lambda functions are best for short expression-based logic. Use `def` when the logic needs multiple statements.

```python
tax_amount = lambda amount: amount * 0.18

def print_invoice(amount):
    tax = amount * 0.18
    payable = amount + tax

    print(f"Amount : {amount:.2f}")
    print(f"Tax    : {tax:.2f}")
    print(f"Payable: {payable:.2f}")

print(tax_amount(1000))
print_invoice(1000)
```

## 19. Mini Ecommerce Lambda Example

This example uses lambda functions for filtering, sorting, discount calculation, and report values.

```python
orders = [
    {"order_id": 1001, "customer": "Anu", "amount": 2400, "status": "paid"},
    {"order_id": 1002, "customer": "Bala", "amount": 1200, "status": "cancelled"},
    {"order_id": 1003, "customer": "Charu", "amount": 5200, "status": "paid"},
    {"order_id": 1004, "customer": "Deepa", "amount": 900, "status": "paid"}
]

paid_orders = list(filter(lambda order: order["status"] == "paid", orders))
paid_orders = sorted(paid_orders, key=lambda order: order["amount"], reverse=True)

discount = lambda amount: amount * 0.10 if amount >= 1000 else 0

print(f"{'Order':<8}{'Customer':<10}{'Amount':>10}{'Discount':>12}{'Payable':>12}")
print("-" * 52)

for order in paid_orders:
    discount_amount = discount(order["amount"])
    payable = order["amount"] - discount_amount

    print(
        f"{order['order_id']:<8}"
        f"{order['customer']:<10}"
        f"{order['amount']:>10.2f}"
        f"{discount_amount:>12.2f}"
        f"{payable:>12.2f}"
    )
```

## Quick Recap

- Lambda functions are anonymous functions.
- A lambda can have no parameter, one parameter, two parameters, or more.
- Lambda functions contain a single expression.
- A lambda can be assigned to a variable and reused.
- Inline lambdas are useful for short one-time behavior.
- Use `def` for longer logic with multiple statements.
- Lambda functions are commonly used with `map()`, `filter()`, `sorted()`, and higher order functions.
