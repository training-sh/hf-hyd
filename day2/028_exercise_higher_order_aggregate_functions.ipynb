## Validation Cell

```python
def assert_equal(actual, expected, message):
    assert actual == expected, f"{message}. Expected {expected}, got {actual}"


def assert_close(actual, expected, message):
    assert actual is not None, f"{message}. Expected {expected}, got None"
    assert abs(actual - expected) < 0.000001, f"{message}. Expected {expected}, got {actual}"


invoice_quality_filters = [
    is_paid_invoice,
    has_customer_id,
    has_positive_qty,
    has_positive_unit_price,
    has_valid_total
]

order_quality_filters = [
    is_delivered_order,
    has_customer_id,
    has_positive_qty,
    has_positive_unit_price,
    has_valid_amount
]

assert_equal(aggregate_count(invoices, "total", invoice_quality_filters), 3, "Invoice count failed")
assert_equal(aggregate_min(invoices, "total", invoice_quality_filters), 1062, "Invoice min failed")
assert_equal(aggregate_max(invoices, "total", invoice_quality_filters), 1416, "Invoice max failed")
assert_equal(aggregate_sum(invoices, "total", invoice_quality_filters), 3568, "Invoice sum failed")
assert_close(aggregate_avg(invoices, "total", invoice_quality_filters), 1189.3333333333333, "Invoice avg failed")

assert_equal(aggregate_count(orders, "amount", order_quality_filters), 3, "Order count failed")
assert_equal(aggregate_min(orders, "amount", order_quality_filters), 900, "Order min failed")
assert_equal(aggregate_max(orders, "amount", order_quality_filters), 1200, "Order max failed")
assert_equal(aggregate_sum(orders, "amount", order_quality_filters), 3100, "Order sum failed")
assert_close(aggregate_avg(orders, "amount", order_quality_filters), 1033.3333333333333, "Order avg failed")

print("Main validation checks passed")
```
