# Destructuring and Unpacking

Destructuring means taking values from a collection and assigning them to separate variables.

In Python, this is usually called unpacking.

## 1. List Unpacking

Values from a list can be assigned to variables.

```python
items = ["Mouse", "Keyboard", "USB Cable"]

item1, item2, item3 = items

print(item1)
print(item2)
print(item3)
```

## 2. Tuple Unpacking

Tuple values can also be unpacked.

```python
product = (101, "Mouse", 499)

product_id, product_name, price = product

print(product_id)
print(product_name)
print(price)
```

## 3. Unpacking String Values

A string can be unpacked character by character.

```python
code = "ABC"

first, second, third = code

print(first)
print(second)
print(third)
```

## 4. Number of Variables Must Match

The number of variables should match the number of values.

```python
order = [1001, "paid"]

order_id, status = order

print(order_id)
print(status)
```

## 5. Swapping Values

Python can swap values without using a temporary variable.

```python
old_price = 499
new_price = 449

old_price, new_price = new_price, old_price

print(old_price)
print(new_price)
```

## 6. Function Returning Multiple Values

A function can return multiple values, and the result can be unpacked.

```python
def calculate_bill(amount):
    tax = amount * 0.18
    total = amount + tax
    return tax, total

tax_amount, total_amount = calculate_bill(1000)

print(tax_amount)
print(total_amount)
```

## 7. Ignoring a Value

`_` is often used for a value that is not needed.

```python
product = (101, "Mouse", 499)

product_id, product_name, _ = product

print(product_id)
print(product_name)
```

## 8. Starred Unpacking

`*` collects remaining values into a list.

```python
items = ["Mouse", "Keyboard", "USB Cable", "Webcam"]

first_item, *remaining_items = items

print(first_item)
print(remaining_items)
```

## 9. Starred Unpacking in the Middle

The starred variable can also appear in the middle.

```python
items = ["Mouse", "Keyboard", "USB Cable", "Webcam"]

first_item, *middle_items, last_item = items

print(first_item)
print(middle_items)
print(last_item)
```

## 10. Unpacking in a Loop

Each tuple can be unpacked while looping.

```python
products = [
    (101, "Mouse"),
    (102, "Keyboard"),
    (103, "USB Cable")
]

for product_id, product_name in products:
    print(product_id, product_name)
```

## 11. Unpacking Dictionary Items in a Loop

`items()` returns key-value pairs that can be unpacked.

```python
prices = {
    "Mouse": 499,
    "Keyboard": 1299,
    "USB Cable": 199
}

for product_name, price in prices.items():
    print(product_name, price)
```

## 12. Unpacking List Values Into Function Arguments

`*` can unpack a list or tuple into function arguments.

```python
def calculate_line_total(quantity, price):
    return quantity * price

line_values = [3, 499]

total = calculate_line_total(*line_values)

print(total)
```

## 13. Unpacking Dictionary Values Into Function Arguments

`**` can unpack a dictionary into named function arguments.

```python
def show_order(order_id, status):
    print(f"Order ID: {order_id}")
    print(f"Status  : {status}")

order_details = {
    "order_id": 1001,
    "status": "paid"
}

show_order(**order_details)
```

## 14. Unpacking With Function Return and Function Call

Returned values can be unpacked and passed to another function.

```python
def get_product_values():
    return "Mouse", 3, 499

def print_product(product_name, quantity, price):
    print(f"Product : {product_name}")
    print(f"Quantity: {quantity}")
    print(f"Price   : {price}")

product_values = get_product_values()

print_product(*product_values)
```

## Quick Recap

- Destructuring in Python is usually called unpacking.
- Lists, tuples, and strings can be unpacked.
- The number of variables usually must match the number of values.
- `_` can be used for a value that is not needed.
- `*` collects remaining values during unpacking.
- `*` unpacks list or tuple values into function arguments.
- `**` unpacks dictionary values into named function arguments.
- Function return values can be unpacked directly.
