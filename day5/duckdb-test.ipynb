```python
import duckdb

# Create an in-memory database
conn = duckdb.connect()

# Execute a SQL query
result = conn.execute("SELECT 'Hello, World!' AS message").fetchall()

print(result)

conn.close()
```

```python
# create db

import duckdb

# Creates hello.duckdb if it doesn't exist
conn = duckdb.connect("hello.duckdb")

conn.execute("""
CREATE TABLE IF NOT EXISTS messages (
    id INTEGER,
    text VARCHAR
)
""")

conn.execute("""
INSERT INTO messages VALUES
(1, 'Hello DuckDB')
""")

rows = conn.execute("SELECT * FROM messages").fetchall()

for row in rows:
    print(row)

conn.close()
```

```python
import duckdb
import pandas as pd

df = pd.DataFrame({
    "name": ["Alice", "Bob", "Charlie"],
    "salary": [60000, 75000, 90000]
})

result = duckdb.sql("""
SELECT *
FROM df
WHERE salary > 70000
""").df()

print(result)
```
