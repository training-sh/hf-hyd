# Garbage Collector Related APIs

Python manages memory automatically. Most objects are cleaned up when they are no longer referenced.

The `gc` module provides APIs to inspect and control Python's cyclic garbage collector.

## 1. Import the `gc` Module

`gc` stands for garbage collector.

```python
import gc

print(gc)
```

## 2. Check Whether Garbage Collection Is Enabled

`gc.isenabled()` returns `True` when automatic garbage collection is enabled.

```python
print(gc.isenabled())
```

## 3. Garbage Collection Counts

`gc.get_count()` returns current allocation counters for the three GC generations.

```python
before_count = gc.get_count()

print(before_count)
```

## 4. Garbage Collection Thresholds

`gc.get_threshold()` shows when automatic collection should run for each generation.

```python
thresholds = gc.get_threshold()

print(thresholds)
```

## 5. Create Many Python Objects

This creates many dictionaries and lists in memory.

```python
before_count = gc.get_count()

objects = []

for number in range(10000):
    row = {
        "id": number,
        "values": [number, number * 2, number * 3]
    }
    objects.append(row)

after_count = gc.get_count()

print("Before:", before_count)
print("After :", after_count)
print("Objects created:", len(objects))
```

## 6. Check Whether an Object Is Tracked

`gc.is_tracked()` checks whether the garbage collector is tracking an object.

```python
sample_list = [1, 2, 3]
sample_dict = {"name": "Python", "type": "language"}
sample_number = 100

print(gc.is_tracked(sample_list))
print(gc.is_tracked(sample_dict))
print(gc.is_tracked(sample_number))
```

## 7. Force Garbage Collection

`gc.collect()` asks Python to run garbage collection immediately.

```python
collected_objects = gc.collect()

print("Collected objects:", collected_objects)
print("GC count after collect:", gc.get_count())
```

## 8. Remove References Before Collection

Objects become available for cleanup only after references are removed.

```python
large_objects = []

for number in range(50000):
    large_objects.append([number, number + 1, number + 2])

print("Created:", len(large_objects))
print("Before deleting reference:", gc.get_count())

del large_objects

collected = gc.collect()

print("Collected:", collected)
print("After collection:", gc.get_count())
```

## 9. Cyclic References

Reference counting alone cannot clean objects that reference each other. The cyclic garbage collector handles these cases.

```python
list_a = []
list_b = []

list_a.append(list_b)
list_b.append(list_a)

print("list_a references list_b")
print("list_b references list_a")

del list_a
del list_b

collected = gc.collect()

print("Collected cyclic objects:", collected)
```

## 10. Disable and Enable Garbage Collection

`gc.disable()` turns off automatic cyclic garbage collection. `gc.enable()` turns it back on.

```python
print("Before disable:", gc.isenabled())

gc.disable()
print("After disable :", gc.isenabled())

gc.enable()
print("After enable  :", gc.isenabled())
```

## 11. Create Objects While GC Is Disabled

This shows how to manually collect after automatic GC has been disabled for a short block of code.

```python
gc.disable()

temporary_objects = []

for number in range(20000):
    temporary_objects.append({"number": number, "square": number * number})

print("GC enabled:", gc.isenabled())
print("Temporary objects:", len(temporary_objects))
print("GC count:", gc.get_count())

del temporary_objects

gc.enable()
collected = gc.collect()

print("GC enabled:", gc.isenabled())
print("Collected:", collected)
print("GC count:", gc.get_count())
```

## 12. Get Garbage Collector Statistics

`gc.get_stats()` returns collection statistics for each generation.

```python
stats = gc.get_stats()

for generation, generation_stats in enumerate(stats):
    print(f"Generation {generation}")
    print(generation_stats)
```

## 13. Get Tracked Objects Count

`gc.get_objects()` returns objects currently tracked by the garbage collector. The list can be large, so this example only prints the count.

```python
tracked_objects = gc.get_objects()

print("Tracked objects:", len(tracked_objects))
```

## 14. Inspect Garbage Collector Generations

`gc.collect()` can receive a generation number: `0`, `1`, or `2`.

```python
collected_gen0 = gc.collect(0)
collected_gen1 = gc.collect(1)
collected_gen2 = gc.collect(2)

print("Generation 0 collected:", collected_gen0)
print("Generation 1 collected:", collected_gen1)
print("Generation 2 collected:", collected_gen2)
```

## 15. Simple Before and After Function

This function creates objects, removes references, forces collection, and shows before/after counts.

```python
def create_and_clean_objects(size):
    before = gc.get_count()

    values = []
    for number in range(size):
        values.append({"number": number, "items": [number, number + 1]})

    after_create = gc.get_count()

    del values

    collected = gc.collect()
    after_collect = gc.get_count()

    return {
        "before": before,
        "after_create": after_create,
        "collected": collected,
        "after_collect": after_collect
    }

result = create_and_clean_objects(30000)

print(result)
```

## Quick Recap

- Python cleans most objects automatically.
- `gc.isenabled()` checks whether automatic garbage collection is enabled.
- `gc.get_count()` shows current GC counters.
- `gc.get_threshold()` shows automatic collection thresholds.
- `gc.collect()` forces garbage collection.
- `gc.disable()` and `gc.enable()` control automatic cyclic garbage collection.
- `gc.is_tracked()` checks whether an object is tracked by GC.
- `gc.get_stats()` returns collection statistics.
- `gc.get_objects()` returns currently tracked objects.
