# NumPy Basics — E-commerce Examples

NumPy stores numeric data in fast, fixed-type arrays. Run the notebook from top to bottom.

## 1. Import NumPy and create arrays

```python
import numpy as np

product_ids = np.array([101, 102, 103, 104, 105])
prices = np.array([799.0, 1299.0, 499.0, 2499.0, 999.0])
quantities = np.array([2, 1, 4, 1, 3])

print(product_ids)
print(type(prices))
```

## 2. Array properties

`shape`, `ndim`, `size`, and `dtype` describe an array.

```python
print("shape:", prices.shape)
print("dimensions:", prices.ndim)
print("items:", prices.size)
print("data type:", prices.dtype)
```

## 3. Useful array creators

```python
print("zeros:", np.zeros(5, dtype=int))       # initial stock
print("ones:", np.ones(5, dtype=int))         # one item each
print("range:", np.arange(100, 501, 100))     # coupon values
print("even steps:", np.linspace(0, 1, 5))    # discount rates
```

## 4. Indexing and slicing

Indexes start at `0`; the stop position of a slice is excluded.

```python
print("first price:", prices[0])
print("last price:", prices[-1])
print("first three:", prices[:3])
print("alternate products:", product_ids[::2])
```

## 5. Two-dimensional arrays

Rows below are orders; columns are product ID, quantity, and unit price.

```python
orders = np.array([
    [101, 2, 799],
    [102, 1, 1299],
    [103, 4, 499],
    [104, 1, 2499],
    [105, 3, 999],
])

print(orders)
print("one value:", orders[1, 2])
print("price column:", orders[:, 2])
print("first two rows:\n", orders[:2, :])
```

## 6. Vectorized calculations

Arithmetic works element by element—no Python loop is needed.

```python
line_totals = prices * quantities
tax = line_totals * 0.18
final_totals = line_totals + tax

print("line totals:", line_totals)
print("including tax:", final_totals.round(2))
```

## 7. Broadcasting

A scalar, or a compatible smaller array, is automatically applied across an array.

```python
discounted_prices = prices * 0.90                  # scalar broadcasting
regional_shipping = np.array([40, 70, 100])
shipping_by_product = prices[:, np.newaxis] + regional_shipping

print(discounted_prices)
print("Product prices plus three shipping rates:\n", shipping_by_product)
```

## 8. Boolean masks and selection

```python
premium_mask = prices >= 1000
print("mask:", premium_mask)
print("premium product IDs:", product_ids[premium_mask])

bulk_or_premium = (quantities >= 3) | (prices >= 2000)
print("selected IDs:", product_ids[bulk_or_premium])
print("shipping labels:", np.where(line_totals >= 2000, "Free", "Paid"))
```

## 9. Aggregations and `axis`

For a 2D array, `axis=0` reduces rows and `axis=1` reduces columns.

```python
print("revenue:", line_totals.sum())
print("average price:", prices.mean())
print("price range:", prices.min(), "to", prices.max())
print("highest-price index:", prices.argmax())

daily_orders = np.array([[2, 1, 4], [3, 0, 2]])
print("units per product:", daily_orders.sum(axis=0))
print("units per day:", daily_orders.sum(axis=1))
```

## 10. Reshape and transpose

The number of elements must stay the same. `-1` asks NumPy to infer a dimension.

```python
weekly_units = np.arange(1, 15).reshape(2, 7)
print("stores x days:\n", weekly_units)
print("days x stores:\n", weekly_units.T)
print("back to 1D:", weekly_units.reshape(-1))
```

## 11. Combine arrays

```python
morning_orders = np.array([1200, 850, 640])
evening_orders = np.array([990, 1750])
all_order_values = np.concatenate((morning_orders, evening_orders))

order_summary = np.column_stack((product_ids, quantities, prices, line_totals))
print(all_order_values)
print("ID, quantity, price, total:\n", order_summary)
```

## 12. Sort, rank, and find unique values

```python
sales_channels = np.array(["app", "web", "app", "store", "web", "app"])
channels, counts = np.unique(sales_channels, return_counts=True)
ranking = np.argsort(line_totals)[::-1]

print("sorted prices:", np.sort(prices))
print("products by revenue:", product_ids[ranking])
print("channel counts:", dict(zip(channels, counts)))
```

## 13. Missing values

`NaN` represents missing floating-point data. Use NumPy's `nan...` functions when it should be ignored.

```python
ratings = np.array([4.5, np.nan, 3.8, 4.9, np.nan])
print("missing mask:", np.isnan(ratings))
print("average available rating:", np.nanmean(ratings))
print("filled ratings:", np.nan_to_num(ratings, nan=0.0))
```

## 14. Views and copies

A slice is usually a view; changing it can change the original. Use `.copy()` for independent data.

```python
stock = np.array([12, 8, 20, 5, 15])
stock_view = stock[:2]
stock_copy = stock[:2].copy()

stock_view[0] = 10
stock_copy[1] = 99
print("original:", stock)
print("independent copy:", stock_copy)
```

## 15. Reproducible random data

Use a generator with a seed when examples must produce the same result each run.

```python
rng = np.random.default_rng(seed=42)
simulated_order_values = rng.integers(200, 3000, size=8)
sampled_product_ids = rng.choice(product_ids, size=3, replace=False)

print("order values:", simulated_order_values)
print("sampled products:", sampled_product_ids)
```

## 16. Timing Python and NumPy

Both cells sum the integers from 1 through `N`. The default 10 million keeps memory use reasonable. A Python list of 100 million integers may require several GB; use `N = 100_000_000` only on a high-memory machine.

```python
import timeit

N = 10_000_000  # Change to 100_000_000 only when enough RAM is available
python_numbers = list(range(1, N + 1))
start = timeit.default_timer()
python_sum = sum(python_numbers)
python_seconds = timeit.default_timer() - start

print(f"Python list sum: {python_sum:,}")
print(f"Python time: {python_seconds:.4f} seconds")
del python_numbers  # Release the large list before the NumPy test
```

```python
import numpy as np
numpy_numbers = np.arange(1, N + 1, dtype=np.int64)
start = timeit.default_timer()
numpy_sum = numpy_numbers.sum()
numpy_seconds = timeit.default_timer() - start

print(f"NumPy array sum: {numpy_sum:,}")
print(f"NumPy time: {numpy_seconds:.4f} seconds")
print(f"NumPy speed-up: {python_seconds / numpy_seconds:.1f}x")
assert python_sum == numpy_sum
del numpy_numbers
```

## Quick practice

Find the product ID with the largest line total, then calculate total revenue after a 5% discount.

```python
top_product_id = product_ids[np.argmax(line_totals)]
discounted_revenue = np.sum(line_totals * 0.95)

print("top product:", top_product_id)
print("discounted revenue:", discounted_revenue)
```
