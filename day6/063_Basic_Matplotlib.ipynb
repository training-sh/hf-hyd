# Basic Matplotlib with pandas

Matplotlib is Python's foundational plotting library. This notebook begins with a plain Python chart and then visualizes two small e-commerce DataFrames. We will create line, bar, pie, histogram, and area charts. Every dataset is hardcoded, so the results are predictable and no external file is required.

## 1. Import the libraries

`matplotlib.pyplot` is commonly imported as `plt`. It provides functions for creating figures, axes, labels, and charts. Pandas works naturally with Matplotlib: we can calculate summaries with pandas and pass their columns or indexes to Matplotlib.

If the import reports `ModuleNotFoundError`, install Matplotlib into the active notebook kernel with `%pip install matplotlib`, restart the kernel, and run the notebook again. The `%pip` command is a Jupyter command, not ordinary Python syntax.

```python
import matplotlib.pyplot as plt
import pandas as pd

pd.set_option("display.float_format", "{:,.2f}".format)
```

## 2. Your first line chart

A **Figure** is the complete drawing area. An **Axes** is the plotting area containing data, labels, and ticks. `plt.subplots()` creates both. Calling `ax.plot()` draws the line, while the remaining methods add meaning. `plt.show()` displays the completed figure.

```python
years = [2021, 2022, 2023, 2024]
sales = [12000, 15500, 18400, 22000]

fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(years, sales, marker="o", color="royalblue", label="Sales")
ax.set_title("Annual Sales")
ax.set_xlabel("Year")
ax.set_ylabel("Sales Amount")
ax.set_xticks(years)
ax.legend()
ax.grid(alpha=0.3)
fig.tight_layout()
plt.show()
```

## 3. Create the e-commerce DataFrames

The first DataFrame contains annual product records with year, price, quantity, and stock code. `Amount` is derived from price multiplied by quantity. The second DataFrame contains monthly quantities for three products and is useful for showing changes over time.

```python
product_sales = pd.DataFrame([
    {"Year": 2021, "StockCode": "A100", "Price": 5.00, "Quantity": 120},
    {"Year": 2021, "StockCode": "B200", "Price": 2.00, "Quantity": 180},
    {"Year": 2021, "StockCode": "C300", "Price": 8.00, "Quantity": 75},
    {"Year": 2022, "StockCode": "A100", "Price": 5.50, "Quantity": 150},
    {"Year": 2022, "StockCode": "B200", "Price": 2.20, "Quantity": 210},
    {"Year": 2022, "StockCode": "C300", "Price": 8.50, "Quantity": 90},
    {"Year": 2023, "StockCode": "A100", "Price": 6.00, "Quantity": 170},
    {"Year": 2023, "StockCode": "B200", "Price": 2.50, "Quantity": 240},
    {"Year": 2023, "StockCode": "C300", "Price": 9.00, "Quantity": 110},
    {"Year": 2024, "StockCode": "A100", "Price": 6.50, "Quantity": 200},
    {"Year": 2024, "StockCode": "B200", "Price": 2.75, "Quantity": 260},
    {"Year": 2024, "StockCode": "C300", "Price": 9.50, "Quantity": 130},
])
product_sales["Amount"] = product_sales["Price"] * product_sales["Quantity"]
product_sales
```

```python
monthly_quantity = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "A100": [12, 15, 14, 18, 21, 24],
    "B200": [20, 18, 25, 23, 28, 30],
    "C300": [8, 10, 9, 12, 14, 16],
})
monthly_quantity
```

## 4. Line chart: change over time

A line chart is suitable for ordered values such as years or months. First, pandas groups the rows by year and sums their amounts. Matplotlib then plots the year index against total revenue. Markers make the individual yearly observations visible.

```python
yearly_revenue = product_sales.groupby("Year")["Amount"].sum()

fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(yearly_revenue.index, yearly_revenue.values, marker="o", linewidth=2)
ax.set(title="Revenue by Year", xlabel="Year", ylabel="Revenue")
ax.set_xticks(yearly_revenue.index)
ax.grid(alpha=0.3)
fig.tight_layout()
plt.show()
```

## 5. Bar chart: compare categories

A bar chart compares separate categories. Here, each bar represents the total quantity sold for one SKU. Starting the vertical axis at zero helps viewers compare bar lengths honestly. The loop adds exact values above the bars.

```python
quantity_by_sku = (
    product_sales.groupby("StockCode")["Quantity"].sum().sort_values(ascending=False)
)

fig, ax = plt.subplots(figsize=(7, 4))
bars = ax.bar(quantity_by_sku.index, quantity_by_sku.values, color="seagreen")
ax.set(title="Total Quantity by SKU", xlabel="Stock Code", ylabel="Units Sold")
ax.set_ylim(bottom=0)
ax.bar_label(bars, padding=3)
fig.tight_layout()
plt.show()
```

## 6. Pie chart: show part of a whole

A pie chart shows how categories contribute to one total. It works best with a small number of categories. `autopct` displays percentages, and `startangle=90` begins the first slice at the top. Avoid pie charts when there are many categories or values are very similar.

```python
revenue_by_sku = product_sales.groupby("StockCode")["Amount"].sum()

fig, ax = plt.subplots(figsize=(6, 5))
ax.pie(
    revenue_by_sku.values, labels=revenue_by_sku.index,
    autopct="%1.1f%%", startangle=90,
)
ax.set_title("Revenue Share by SKU")
ax.axis("equal")
fig.tight_layout()
plt.show()
```

## 7. Histogram: understand a distribution

A histogram groups numeric observations into intervals called bins. The horizontal axis shows price ranges, while bar height shows how many records fall in each range. Unlike a bar chart, histogram bars represent continuous numeric intervals rather than named categories. Different bin counts can reveal or hide patterns.

```python
fig, ax = plt.subplots(figsize=(7, 4))
ax.hist(product_sales["Price"], bins=6, color="mediumpurple", edgecolor="black")
ax.set(title="Distribution of Product Prices", xlabel="Price", ylabel="Record Count")
fig.tight_layout()
plt.show()
```

## 8. Area chart: totals and composition over time

A stacked area chart shows both the total quantity and each SKU's contribution across months. `stackplot()` receives the x-axis followed by one numeric sequence per area. Stacking is useful for composition, but overlapping line charts may be clearer when exact product-to-product comparison is more important.

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.stackplot(
    monthly_quantity["Month"],
    monthly_quantity["A100"],
    monthly_quantity["B200"],
    monthly_quantity["C300"],
    labels=["A100", "B200", "C300"],
    alpha=0.8,
)
ax.set(title="Monthly Quantity by SKU", xlabel="Month", ylabel="Units")
ax.legend(loc="upper left")
fig.tight_layout()
plt.show()
```

## 9. Pandas plotting uses Matplotlib

Pandas provides convenient `.plot()` methods backed by Matplotlib. They return a Matplotlib Axes, so the same title, label, legend, and formatting methods still apply. This example creates a multi-line chart directly from the monthly DataFrame after setting Month as the index.

```python
ax = monthly_quantity.set_index("Month").plot(
    kind="line", marker="o", figsize=(8, 4)
)
ax.set_title("Monthly Product Quantities")
ax.set_xlabel("Month")
ax.set_ylabel("Units")
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

## Choosing a chart

- **Line:** change across ordered time.
- **Bar:** comparison between categories.
- **Pie:** simple part-to-whole comparison with few categories.
- **Histogram:** distribution of one numeric variable.
- **Area:** total and composition changing over time.

Start with the analytical question, then select the chart. Always include a clear title and axis labels, and avoid decoration that does not improve understanding.
