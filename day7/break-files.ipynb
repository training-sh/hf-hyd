```python
from pathlib import Path
import shutil

INPUT_FILE = Path(r"C:\data\weblogs\access.log")
OUTPUT_DIR = INPUT_FILE.parent / "parts"

# Delete previous output directory completely
if OUTPUT_DIR.exists():
    shutil.rmtree(OUTPUT_DIR)

OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

NUMBER_OF_PARTS = 20


def split_by_lines(
    input_file: Path,
    output_dir: Path,
    number_of_parts: int,
) -> None:

    if not input_file.exists():
        raise FileNotFoundError(f"File not found: {input_file}")

    total_size = input_file.stat().st_size
    approximate_part_size = total_size // number_of_parts

    print(f"Input size: {total_size:,} bytes")
    print(f"Approximate size per part: {approximate_part_size:,} bytes")

    part_number = 1
    current_part_size = 0

    output_file = None

    try:
        with input_file.open("rb") as source:

            for line in source:

                if (
                    output_file is None
                    or (
                        current_part_size >= approximate_part_size
                        and part_number < number_of_parts
                    )
                ):
                    if output_file is not None:
                        output_file.close()

                    filename = (
                        f"{input_file.stem}."
                        f"part{part_number:03d}-of-{number_of_parts:03d}.txt"
                    )

                    filepath = output_dir / filename

                    print(f"Creating: {filepath}")

                    output_file = filepath.open("wb")

                    current_part_size = 0
                    part_number += 1

                output_file.write(line)
                current_part_size += len(line)

    finally:
        if output_file is not None:
            output_file.close()

    print(f"\nCompleted. Created {part_number - 1} text files.")


split_by_lines(
    input_file=INPUT_FILE,
    output_dir=OUTPUT_DIR,
    number_of_parts=NUMBER_OF_PARTS,
)
```
