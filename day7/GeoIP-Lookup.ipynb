```python
from pathlib import Path

import geoip2.database
from geoip2.errors import AddressNotFoundError

mmdb_file = Path(r"C:\data\geoip\GeoLite2-City.mmdb")

ip_address = "8.8.8.8"

with geoip2.database.Reader(str(mmdb_file)) as reader:

    try:
        result = reader.city(ip_address)

        subdivision = result.subdivisions.most_specific

        print("IP Address          :", ip_address)
        print("Country             :", result.country.name)
        print("Country ISO         :", result.country.iso_code)
        print("State               :", subdivision.name)
        print("State ISO           :", subdivision.iso_code)
        print("City                :", result.city.name)
        print("Postal Code         :", result.postal.code)
        print("Latitude            :", result.location.latitude)
        print("Longitude           :", result.location.longitude)
        print("Accuracy Radius (km):", result.location.accuracy_radius)
        print("Time Zone           :", result.location.time_zone)
        print("Continent           :", result.continent.name)
        print("Continent Code      :", result.continent.code)
        print("Network             :", result.traits.network)

    except AddressNotFoundError:
        print(f"{ip_address} not found in GeoLite2 database.")

    except Exception as error:
        print(error)
```
