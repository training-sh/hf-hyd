## 1. Hello World

A `Hello World` program is usually the first program in any language. In data engineering, simple print statements are also useful while learning, debugging, or checking pipeline steps.

```python

# print() displays output on the screen.
# This is useful when we want to quickly verify what our code is doing.

print("Hello, World!")
```

## 2. Comments

Comments are notes for humans. Python ignores them while running the program. Good comments explain **why** something is done, not just **what** the code already says.

In data engineering, comments are useful when explaining business rules, assumptions, data quality checks, or transformations.

```python
# This is a single-line comment.
# Python will ignore this line during execution.

source_system = "sales_db"  # Inline comment: this tells us where the data comes from.

print(source_system)
```

## 3. Indentation

Python uses indentation to understand blocks of code. A block is a group of statements that belong together.

Languages like Java or C often use braces `{}`. Python uses spaces. This makes Python readable, but it also means indentation mistakes can break the program.

The standard style is **4 spaces** per indentation level.

```python

# The print statement below is indented because it belongs to the if block.

records_loaded = 120

if records_loaded > 0:
    print("Data load completed successfully.")
    print("Records loaded:", records_loaded)

print("This line is outside the if block.")
```

## 4. Nested Indentation

Sometimes one block exists inside another block. This is called nesting.

In data engineering, nested indentation commonly appears in loops, conditions, error handling, and data validation logic.

```python

# The second if statement is inside the first if statement.
# Each level of logic needs one more indentation level.

file_found = True
file_has_data = True

if file_found:
    print("File exists.")

    if file_has_data:
        print("File contains data. Ready to process.")
    else:
        print("File is empty. Do not process.")
else:
    print("File missing. Stop the pipeline.")
```

## 5. Common Indentation Error

If we forget indentation after `if`, Python raises an `IndentationError`.

We keep the incorrect example as text so the notebook can run without stopping.

```python

# This is an incorrect example written inside a string.

wrong_code = """
if True:
print("This line should be indented")
"""

print(wrong_code)
```

## 6. Line Continuation Using Backslash

A backslash `\` tells Python that the statement continues on the next line.

This works, but in modern Python code, parentheses are usually preferred because they are cleaner and less error-prone.

```python

# The backslash continues the expression onto the next line.
# Be careful: there should be no extra characters after the backslash.

total_records = 1000 + \
                250 + \
                75

print("Total records:", total_records)
```

## 7. Line Continuation Using Parentheses

Parentheses allow Python statements to continue naturally across multiple lines.

This is usually the preferred style for long calculations, long strings, lists, tuples, dictionaries, and function arguments.

```python

# Parentheses make long expressions easier to read.
# This style is cleaner than using backslash for most cases.

total_sales = (
    15000
    + 27500
    + 18400
)

print("Total sales:", total_sales)
```

## 8. Parentheses with Long Function Calls

In real projects, function calls can become long because they may have many arguments.

Putting each argument on a separate line improves readability and makes future changes easier.

```python

# This function call is split across multiple lines using parentheses.
# This is common in production Python code.

print(
    "Pipeline status:",
    "SUCCESS",
    "| Records processed:",
    1325,
    "| Source:",
    "customer_api",
)
```

## 9. Multi-line Strings

Triple quotes `"""` or `'''` are used to create strings that span multiple lines.

They are useful for messages, SQL queries, documentation strings, and sample data.

```python

# Triple quotes preserve line breaks inside the string.
# Data engineers often use this style for SQL queries.

query = """
SELECT
    customer_id,
    order_date,
    total_amount
FROM orders
WHERE order_date >= '2026-01-01'
"""

print(query)
```

## 10. Readability Practice

Python code should be readable. In data engineering, readable code matters because pipelines are maintained for months or years.

This final example combines comments, indentation, and line continuation.

```python

# This example combines the ideas from this notebook.
# We calculate whether a small data load meets a quality threshold.

source_name = "orders_file"
expected_records = 1000
actual_records = (
    800
    + 150
    + 50
)

load_percentage = actual_records / expected_records

if load_percentage >= 0.95:
    print("Quality check passed for", source_name)
    print("Load percentage:", load_percentage)
else:
    print("Quality check failed for", source_name)
    print("Investigate missing records.")
```

## Quick Recap

- `print()` is used to display output.
- Comments explain code to humans.
- Indentation defines code blocks in Python.
- Backslash `\` can continue a line, but should be used carefully.
- Parentheses are usually the cleaner way to write multi-line Python.
- Readable code is especially important in data engineering projects.
