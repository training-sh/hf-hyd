# Python Data Types, Objects, and Duck Typing

This notebook covers Python data types from a practical data engineering point of view.

We will cover:

- Common built-in data types
- How to inspect types
- Dynamic typing
- Duck typing
- Object identity and memory references
- Mutable vs immutable objects
- Why Python does not behave like primitive-type languages such as C or Java

## 1. Everything Starts with Values

A value is a piece of data. Every value in Python has a type.

For example, `100` is an integer, `"orders"` is a string, and `True` is a boolean.

```python

# type() tells us the type of a value.

record_count = 100
table_name = "orders"
is_valid = True

print(type(record_count))
print(type(table_name))
print(type(is_valid))
```

## 2. Common Built-in Data Types

Python gives us several built-in data types. These are the types students will use constantly in scripts, ETL jobs, API processing, and data validation.

Common types include:

- `int`: whole numbers
- `float`: decimal numbers
- `str`: text
- `bool`: true or false
- `NoneType`: absence of a value
- `list`: ordered, mutable collection
- `tuple`: ordered, immutable collection
- `dict`: key-value collection
- `set`: unique unordered collection

```python

# This cell shows common Python types used in data engineering.

examples = [
    42,
    99.95,
    "customer",
    True,
    None,
    ["csv", "json", "parquet"],
    ("host", "port"),
    {"source": "api", "active": True},
    {"orders", "customers", "orders"},
]

for value in examples:
    print(value, "=>", type(value))
```

## 3. Variables Do Not Store Types Permanently

Python is dynamically typed. This means a variable name can point to different types at different times.

This is flexible, but in production code we should avoid changing the meaning of a variable too often because it can confuse readers.

```python

# The variable name 'status' can point to different types.
# Python allows it, but we should use this flexibility carefully.

status = "SUCCESS"
print(status, type(status))

status = 1
print(status, type(status))

status = True
print(status, type(status))
```

## 4. Python Names Point to Objects

In Python, variables are names that refer to objects.

A simple way to say it:

- The object holds the actual value.
- The variable name points to that object.
- The type belongs to the object, not permanently to the variable name.

This is different from many languages where variables are thought of as fixed boxes that hold primitive values.

```python

# id() gives the identity of an object during the current Python session.
# Think of it as a way to observe whether two names point to the same object.

a = ["orders", "customers"]
b = a

print("a:", a)
print("b:", b)
print("id(a):", id(a))
print("id(b):", id(b))
print("Same object?", a is b)
```

## 5. Object vs Primitive Type Thinking

In languages like C, Java, or C#, you may hear about primitive types such as `int`, `float`, or `boolean`.

Python is different: values like integers, floats, strings, and booleans are objects. They have:

- A type
- An identity
- A value
- Methods or behavior

For example, a string is not just raw text. It is an object with methods like `.upper()`, `.lower()`, and `.replace()`.

```python

# Even simple-looking values are objects with behavior.

source = "customer_api"
count = 25

print(source.upper())
print(count.bit_length())
print(type(source))
print(type(count))
```

## 6. Exploring Objects with `dir()`

`dir()` shows the attributes and methods available on an object.

This is useful while learning, debugging, or exploring unfamiliar libraries.

```python

# dir() helps us discover what an object can do.
# We show only a few methods to keep the output readable.

file_name = "orders_2026.csv"

string_features = dir(file_name)

print("Does string have upper?", "upper" in string_features)
print("Does string have split?", "split" in string_features)
print("Does string have append?", "append" in string_features)
```

## 7. `isinstance()` for Type Checking

`type()` tells us the exact type. `isinstance()` checks whether a value belongs to a type or a family of types.

`isinstance()` is usually preferred when writing validation logic.

```python

# isinstance() is useful when validating inputs.

batch_size = 500
source_name = "orders"

print(isinstance(batch_size, int))
print(isinstance(source_name, str))
print(isinstance(batch_size, (int, float)))
```

## 8. Mutable and Immutable Types

Some objects can be changed after creation. These are called mutable objects.

Some objects cannot be changed after creation. These are called immutable objects.

Common examples:

- Mutable: `list`, `dict`, `set`
- Immutable: `int`, `float`, `str`, `bool`, `tuple`

```python

# Lists are mutable. We can change the same list object.

columns = ["order_id", "customer_id"]
print("Before:", columns)
print("Before id:", id(columns))

columns.append("order_date")

print("After:", columns)
print("After id:", id(columns))
```

## 9. Immutable Objects Create New Objects

Strings are immutable. When we appear to modify a string, Python creates a new string object.

This matters when working with transformations, memory usage, and performance.

```python

# The string itself is not changed.
# A new string object is created and assigned back to the name.

table = "orders"
print("Before:", table)
print("Before id:", id(table))

table = table.upper()

print("After:", table)
print("After id:", id(table))
```

## 10. Equality vs Identity

`==` checks whether two values are equal.

`is` checks whether two names point to the exact same object.

In normal data checks, we usually use `==`. Use `is` for checks like `value is None`.

```python

# Two lists can have the same content but still be different objects.

list_a = [1, 2, 3]
list_b = [1, 2, 3]
list_c = list_a

print("list_a == list_b:", list_a == list_b)
print("list_a is list_b:", list_a is list_b)
print("list_a is list_c:", list_a is list_c)

missing_value = None
print("missing_value is None:", missing_value is None)
```

## 11. Duck Typing

Duck typing means Python often cares more about what an object can do than what exact type it is.

The phrase comes from the idea: if it walks like a duck and talks like a duck, treat it like a duck.

In Python terms: if an object has the method or behavior we need, we can often use it.

```python

# This function does not ask whether the input is a list, tuple, or string.
# It only expects the object to work with len().

def show_size(data):
    print("Size:", len(data))

show_size(["orders", "customers", "products"])
show_size(("host", "port"))
show_size("customer_api")
```

## 12. Duck Typing with File-like Objects

Duck typing is powerful in data engineering. Many tools work with file-like objects: the object does not need to be a real file on disk, but it should behave like a file.

For example, an object with a `.read()` method can often be treated as readable input.

```python

# This function only requires a .read() method.
# It does not care whether the object is a real file, an in-memory stream, or a custom object.

from io import StringIO

def preview_data(readable_source):
    content = readable_source.read()
    print(content[:30])

sample_csv = StringIO("order_id,total\n1,500\n2,750")

preview_data(sample_csv)
```

## 13. Type Conversion

Data often arrives as text. CSV files, JSON payloads, API responses, and environment variables commonly represent numbers as strings.

Type conversion changes a value from one type to another when possible.

```python

# Convert values before calculation.

raw_quantity = "10"
raw_price = "99.5"

quantity = int(raw_quantity)
price = float(raw_price)
total = quantity * price

print(total)
print(type(total))
```

## 14. Handling Conversion Errors

Not every value can be converted. Production data often contains blanks, unexpected text, or invalid values.

We use `try` and `except` to handle these cases gracefully.

```python

# Never assume external data is clean.
# Handle conversion errors explicitly.

raw_value = "N/A"

try:
    numeric_value = float(raw_value)
    print("Converted value:", numeric_value)
except ValueError:
    print("Could not convert value:", raw_value)
```

## 15. Small Practice: Inspect a Record

A record in a pipeline is often represented as a dictionary. We can inspect each field and understand what Python type it has.

This is a useful habit when working with unfamiliar data.

```python

# Dictionaries are common for records from APIs and JSON files.
# We can loop through fields and inspect each value.

record = {
    "order_id": 101,
    "customer_id": "C001",
    "amount": 2500.75,
    "is_paid": True,
    "coupon_code": None,
}

for field_name, field_value in record.items():
    print(field_name, "=>", field_value, "=>", type(field_value))
```

## Quick Recap

- Every value in Python has a type.
- Python variables are names that point to objects.
- Python is dynamically typed.
- Python values are objects, even values like integers and strings.
- `type()`, `isinstance()`, `id()`, and `dir()` help us explore objects.
- Mutable objects can change in place; immutable objects cannot.
- `==` checks equality; `is` checks identity.
- Duck typing focuses on behavior rather than exact type.
- Type conversion is essential when working with external data.
