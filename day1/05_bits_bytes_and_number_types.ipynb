# Bits, Bytes, Binary Notation, and Number Types

This notebook explains the low-level vocabulary students often hear when learning programming and data engineering:

- Bit
- Byte
- Binary notation like `0b001`
- Boolean
- Integer
- Short, int, long
- Float and double
- How these ideas differ between Python and languages like C, C++, Java, and C#

## 1. What Is a Bit?

A bit is the smallest unit of data in computing.

A bit can have only two values:

- `0`
- `1`

Computers represent data using combinations of bits.

```python

# A single bit can represent two states.
# For example: off/on, false/true, no/yes.

bit_off = 0
bit_on = 1

print("Off bit:", bit_off)
print("On bit:", bit_on)
```

## 2. What Is a Byte?

A byte is usually 8 bits.

Example:

`01000001`

One byte can represent 256 different patterns because `2 ** 8 = 256`.

```python

# 8 bits can produce 256 possible combinations.

bits_in_byte = 8
patterns = 2 ** bits_in_byte

print("Bits in one byte:", bits_in_byte)
print("Possible patterns:", patterns)
```

## 3. Binary Notation in Python: `0b`

Python uses the prefix `0b` to write integer values in binary.

For example:

- `0b001` means binary `001`, which is decimal `1`
- `0b010` means binary `010`, which is decimal `2`
- `0b100` means binary `100`, which is decimal `4`

The `0b` prefix is just notation. The value is still a Python `int`.

```python

# Binary literals are still integers in Python.

read_permission = 0b001
write_permission = 0b010
execute_permission = 0b100

print(read_permission, type(read_permission))
print(write_permission, type(write_permission))
print(execute_permission, type(execute_permission))
```

## 4. Converting Numbers to Binary

`bin()` converts an integer into a binary string representation.

The result starts with `0b` to show that the representation is binary.

```python

# bin() helps us view the binary representation of an integer.

for number in range(8):
    print(number, "=>", bin(number))
```

## 5. Fixed-width Binary Display

Sometimes we want to display binary using a fixed number of digits.

For example, `001` is easier to understand than `1`  

```python

# format(number, '03b') shows a number as 3 binary digits.
# format(number, '08b') shows a number as 8 binary digits.

number = 5

print("Normal binary:", bin(number))
print("3-bit view:", format(number, "03b"))
print("8-bit view:", format(number, "08b"))
```

## 6. Decimal, Binary, Octal, and Hexadecimal

Python supports several integer literal formats:

- Decimal: `10`
- Binary: `0b1010`
- Octal: `0o12`
- Hexadecimal: `0xA`

All of these represent the same integer value: `10`.

```python

# Different notations can represent the same integer value.

decimal_value = 10
binary_value = 0b1010
octal_value = 0o12
hex_value = 0xA

print(decimal_value)
print(binary_value)
print(octal_value)
print(hex_value)
print(decimal_value == binary_value == octal_value == hex_value)
```

## 7. Boolean Values

A boolean represents one of two truth values:

- `True`
- `False`

In Python, `bool` is a subclass of `int`. That means `True` behaves like `1`, and `False` behaves like `0` in arithmetic.

```python

# Boolean values are objects in Python.
# They also behave numerically in some contexts.

print(True, type(True))
print(False, type(False))
print(isinstance(True, int))
print(True + True + False)
print(int(True), int(False))
```

## 8. Python `int`

Python's `int` represents whole numbers.

Unlike C, C++, Java, or C#, Python does not expose separate normal integer types like `short`, `int`, and `long` at the language level.

In Python 3, `int` can grow as large as memory allows.

```python

# Python integers are arbitrary precision.
# The same int type handles small and huge numbers.

small_number = 100
large_number = 10 ** 100

print(small_number, type(small_number))
print(large_number, type(large_number))
print("Large number bit length:", large_number.bit_length())
```

## 9. Bit Length

`bit_length()` tells us how many bits are needed to represent an integer in binary, excluding the sign and leading zeros.

This is useful when learning how numbers grow internally.

```python

# Larger numbers need more bits.

numbers = [0, 1, 2, 3, 7, 8, 255, 256, 1024]

for number in numbers:
    print(number, "binary:", bin(number), "bit_length:", number.bit_length())
```

## 10. Signed and Unsigned Integer Idea

In many lower-level languages, integer types can be signed or unsigned.

- Signed integers can represent negative and positive numbers.
- Unsigned integers represent only zero and positive numbers.

Python's normal `int` does not require you to choose signed or unsigned size. Python handles the sign and size for you.

```python

# Python int naturally supports negative and positive values.

negative_value = -25
positive_value = 25

print(negative_value, type(negative_value))
print(positive_value, type(positive_value))
print(abs(negative_value))
```

## 11. Common Integer Sizes in Other Languages

Languages such as C, C++, Java, and C# often talk about fixed-width integer types.

Typical names include:

- `byte`: often 8 bits
- `short`: often 16 bits
- `int`: often 32 bits
- `long`: often 64 bits, depending on language/platform

Exact sizes depend on the language and sometimes the platform. Java is more fixed. C and C++ can vary by platform.

## 12. Value Ranges by Bit Width

For unsigned integers:

- Minimum: `0`
- Maximum: `2 ** bits - 1`

For signed integers using common two's-complement representation:

- Minimum: `-(2 ** (bits - 1))`
- Maximum: `2 ** (bits - 1) - 1`

```python

# Fixed-width integer types have limits.
# Python int does not overflow like these fixed-width examples.

for bits in [8, 16, 32, 64]:
    unsigned_min = 0
    unsigned_max = 2 ** bits - 1
    signed_min = -(2 ** (bits - 1))
    signed_max = 2 ** (bits - 1) - 1

    print(bits, "bits")
    print("  unsigned:", unsigned_min, "to", unsigned_max)
    print("  signed:", signed_min, "to", signed_max)
```

## 13. Overflow in Fixed-width Types

In fixed-width integer systems, values can overflow when they exceed the available range.

Python's normal `int` does not overflow. It grows automatically, limited mainly by available memory.

However, libraries like NumPy, databases, file formats, and binary protocols often use fixed-width types.

```python

# Python int keeps growing instead of overflowing.

max_8_bit_unsigned = 255

print(max_8_bit_unsigned)
print(max_8_bit_unsigned + 1)
print(type(max_8_bit_unsigned + 1))
```

## 14. Simulating 8-bit Overflow with Modulo

We can simulate unsigned 8-bit behavior using modulo `256`.

This is useful for understanding binary protocols and low-level systems.

```python

# Unsigned 8-bit values wrap around after 255.

value = 255
wrapped_value = (value + 1) % 256

print("normal Python:", value + 1)
print("simulated unsigned 8-bit:", wrapped_value)
```

## 15. Python `float`

Python's `float` usually uses C `double` internally.

That means Python `float` is normally a double-precision floating point value, not a 32-bit single-precision float.

This is different from languages where `float` and `double` are separate normal types.

```python

# Python float is the language's standard decimal-number type.
# It is usually implemented as a C double.

pi = 3.141592653589793

print(pi)
print(type(pi))
print(pi.hex())
```

## 16. Float Precision Problem

Floating point numbers are stored in binary.

Some decimal values, such as `0.1`, cannot be represented exactly in binary floating point.

This is why `0.1 + 0.2` is not exactly equal to `0.3`.

```python

# This is not a Python bug. It is normal binary floating point behavior.

result = 0.1 + 0.2

print(result)
print(result == 0.3)
print("0.1:", (0.1).hex())
print("0.2:", (0.2).hex())
print("0.3:", (0.3).hex())
print("result:", result.hex())
```

## 17. Float vs Double Vocabulary

In many languages:

- `float` usually means 32-bit single precision
- `double` usually means 64-bit double precision

In Python:

- `float` usually means 64-bit double precision internally
- Python does not have a separate built-in `double` type

If you need explicit 32-bit or 64-bit numeric arrays, tools like NumPy provide types such as `float32`, `float64`, `int32`, and `int64`.

## 18. Checking Approximate Float Equality

When comparing floats, avoid exact equality for calculated results.

Use a tolerance or `math.isclose()`.

```python

# math.isclose() is safer for comparing calculated floats.

import math

result = 0.1 + 0.2

print(result == 0.3)
print(math.isclose(result, 0.3))
```

## 19. Decimal for Money and Exact Decimal Values

For money or exact decimal calculations, use `Decimal` instead of binary `float`.

Create `Decimal` values from strings, not from floats.

```python

# Decimal avoids many binary floating point surprises for decimal arithmetic.

from decimal import Decimal

price = Decimal("0.10")
tax = Decimal("0.20")

print(price + tax)
print((price + tax) == Decimal("0.30"))
```

## 20. Bytes in Python

`bytes` is a Python type used for raw binary data.

A `bytes` object contains integers from `0` to `255`.

This is different from `str`, which represents Unicode text.

```python

# bytes is for raw data. str is for text.

raw_data = b"ABC"

print(raw_data)
print(type(raw_data))
print(list(raw_data))

for byte_value in raw_data:
    print(byte_value, "=>", format(byte_value, "08b"))
```

## 21. Encoding Text to Bytes

Text must be encoded before it becomes bytes.

A common encoding is UTF-8.

In data engineering, encoding matters when reading files, APIs, logs, and database exports.

```python

# encode() converts str to bytes.
# decode() converts bytes back to str.

text = "Data"
encoded = text.encode("utf-8")
decoded = encoded.decode("utf-8")

print(text, type(text))
print(encoded, type(encoded))
print(decoded, type(decoded))
```

## 22. Reading and Writing Binary Structures with `struct`

The `struct` module converts between Python values and C-style binary data.

This is useful when working with binary files, network protocols, or fixed-width formats.

```python

# 'i' means C signed int, commonly 4 bytes.
# 'd' means C double, commonly 8 bytes.

import struct

packed_int = struct.pack("i", 100)
packed_double = struct.pack("d", 3.14)

print("packed int:", packed_int, "size:", len(packed_int))
print("packed double:", packed_double, "size:", len(packed_double))

print("unpacked int:", struct.unpack("i", packed_int)[0])
print("unpacked double:", struct.unpack("d", packed_double)[0])
```

## 23. Endianness

Endianness describes byte order in multi-byte values.

- Little-endian: least significant byte first
- Big-endian: most significant byte first

This matters when reading binary files or sending data over networks.

```python

# struct lets us choose byte order explicitly.
# '<I' means little-endian unsigned int.
# '>I' means big-endian unsigned int.

number = 0x12345678

little = struct.pack("<I", number)
big = struct.pack(">I", number)

print("number:", hex(number))
print("little-endian bytes:", little)
print("big-endian bytes:", big)
```

## 24. Memory Size of Python Objects

Python values are objects, so they have overhead.

`sys.getsizeof()` shows the size of a Python object in memory. This is different from the raw C payload size.

```python

# A Python int is not just 4 bytes like a typical C int.
# A Python float object is larger than the 8-byte double payload.

import sys

values = [0, 1, 255, 256, 10 ** 50, 3.14, True, b"A", "A"]

for value in values:
    print(repr(value), type(value).__name__, sys.getsizeof(value), "bytes")
```
