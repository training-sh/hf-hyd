# Python Operators

Operators are symbols or keywords that perform operations on values.

In Python, operators are used for arithmetic, comparison, logical decisions, assignment, membership checks, identity checks, and bit-level operations.

This notebook covers the major Python operators with examples useful for general programming and data engineering.

## 1. Arithmetic Operators

Arithmetic operators are used for mathematical calculations.

Common arithmetic operators:

- `+` addition
- `-` subtraction
- `*` multiplication
- `/` true division
- `//` floor division
- `%` modulo
- `**` exponentiation

```python
# Teaching point:
# Arithmetic operators are commonly used for metrics, totals, averages, and calculations.

a = 17
b = 5

print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)
print("a // b =", a // b)
print("a % b =", a % b)
print("a ** b =", a ** b)
```

## 2. Division: `/` vs `//`

`/` always performs true division and returns a floating point result.

`//` performs floor division. It gives the quotient rounded down to the nearest whole number.

```python
# Teaching point:
# Use / when you need an exact ratio.
# Use // when you need full groups or pages.

records = 105
batch_size = 20

print("records / batch_size:", records / batch_size)
print("records // batch_size:", records // batch_size)
print("records % batch_size:", records % batch_size)
```

## 3. Modulo Operator `%`

The modulo operator returns the remainder after division.

It is useful for checking even/odd numbers, cycling through values, and finding leftover records after batching.

```python
# Teaching point:
# If number % 2 is 0, the number is even.

number = 42

if number % 2 == 0:
    print(number, "is even")
else:
    print(number, "is odd")
```

## 4. Assignment Operators

Assignment operators assign or update values.

Common assignment operators:

- `=` assign
- `+=` add and assign
- `-=` subtract and assign
- `*=` multiply and assign
- `/=` divide and assign
- `//=` floor divide and assign
- `%=` modulo and assign
- `**=` exponent and assign

```python
# Teaching point:
# Augmented assignment updates an existing variable.

processed_records = 0
processed_records += 100
processed_records += 250

print("Processed records:", processed_records)

retry_count = 5
retry_count -= 1

print("Retries left:", retry_count)
```

## 5. Comparison Operators

Comparison operators return boolean values: `True` or `False`.

Common comparison operators:

- `==` equal to
- `!=` not equal to
- `>` greater than
- `<` less than
- `>=` greater than or equal to
- `<=` less than or equal to

```python
# Teaching point:
# Comparisons are used heavily in validation and filtering.

expected_records = 1000
actual_records = 975

print(actual_records == expected_records)
print(actual_records != expected_records)
print(actual_records > expected_records)
print(actual_records < expected_records)
print(actual_records >= 950)
print(actual_records <= expected_records)
```

## 6. Chained Comparisons

Python allows chained comparisons, which are readable and close to mathematical notation.

For example, `80 <= score <= 100` checks whether `score` is between 80 and 100.

```python
# Teaching point:
# Chained comparisons avoid repeating the variable name.

data_quality_score = 96

if 90 <= data_quality_score <= 100:
    print("Excellent data quality")
else:
    print("Needs review")
```

## 7. Logical Operators

Logical operators combine boolean expressions.

Python logical operators:

- `and`: true only if both conditions are true
- `or`: true if at least one condition is true
- `not`: reverses a boolean value

```python
# Teaching point:
# Logical operators are used to combine validation rules.

file_exists = True
file_has_data = True
schema_valid = False

print("Ready to process:", file_exists and file_has_data)
print("Needs attention:", not schema_valid)
print("Can attempt processing:", file_exists or file_has_data)
```

## 8. Short-circuit Evaluation

`and` and `or` use short-circuit evaluation.

For `and`, if the first condition is false, Python does not need to check the second condition.

For `or`, if the first condition is true, Python does not need to check the second condition.

```python
# Teaching point:
# Short-circuiting helps avoid errors and unnecessary work.

records = []

if records and records[0] == "header":
    print("Header found")
else:
    print("No records or no header")

# Because records is empty, Python does not evaluate records[0].
```

## 9. Membership Operators

Membership operators check whether a value exists inside a collection.

Python membership operators:

- `in`
- `not in`

```python
# Teaching point:
# Membership checks are useful for validating columns, keys, and allowed values.

required_columns = ["order_id", "customer_id", "amount"]
column_name = "amount"

print(column_name in required_columns)
print("discount" not in required_columns)

record = {"order_id": 101, "amount": 500}
print("order_id" in record)  # For dictionaries, in checks keys.
```

## 10. Identity Operators

Identity operators check whether two names refer to the same object.

Python identity operators:

- `is`
- `is not`

Use `==` for value equality. Use `is` mainly for identity checks like `value is None`.

```python
# Teaching point:
# == checks value equality. is checks object identity.

a = [1, 2, 3]
b = [1, 2, 3]
c = a

print("a == b:", a == b)
print("a is b:", a is b)
print("a is c:", a is c)

missing_value = None
print("missing_value is None:", missing_value is None)
```

## 11. Bitwise Operators

Bitwise operators work at the binary bit level.

Common bitwise operators:

- `&` bitwise AND
- `|` bitwise OR
- `^` bitwise XOR
- `~` bitwise NOT
- `<<` left shift
- `>>` right shift

These are less common in beginner Python, but useful in permissions, flags, masks, networking, binary formats, and performance-heavy systems.

```python
# Teaching point:
# bin() shows the binary representation of an integer.

x = 0b1010  # 10 in decimal
y = 0b1100  # 12 in decimal

print("x:", bin(x), x)
print("y:", bin(y), y)
print("x & y:", bin(x & y), x & y)
print("x | y:", bin(x | y), x | y)
print("x ^ y:", bin(x ^ y), x ^ y)
print("x << 1:", bin(x << 1), x << 1)
print("x >> 1:", bin(x >> 1), x >> 1)
```

## 12. Bitwise Operators for Flags

Flags are values that represent yes/no options using bits.

For example, a user or service account may have read, write, and execute permissions.

```python
# Teaching point:
# Bitwise OR can combine flags.
# Bitwise AND can check whether a flag is present.

READ = 0b001
WRITE = 0b010
EXECUTE = 0b100

permission = READ | WRITE

print("permission:", bin(permission))
print("Can read:", bool(permission & READ))
print("Can write:", bool(permission & WRITE))
print("Can execute:", bool(permission & EXECUTE))
```

## 13. Bitwise Assignment Operators

Python also supports assignment versions of bitwise operators:

- `&=`
- `|=`
- `^=`
- `<<=`
- `>>=`

```python
# Teaching point:
# Bitwise assignment updates the existing variable with a bitwise operation.

flags = 0b001
print("Initial:", bin(flags))

flags |= 0b010
print("After adding WRITE:", bin(flags))

flags &= 0b001
print("After keeping only READ:", bin(flags))
```

## 14. Unary Operators

Unary operators work with one value.

Common unary operators:

- `+x` positive
- `-x` negative
- `not x` logical negation
- `~x` bitwise inversion

```python
# Teaching point:
# Unary operators are placed before the value.

value = 10
is_loaded = False

print(+value)
print(-value)
print(not is_loaded)
print(~value)
```

## 15. Conditional Expression

Python has a compact conditional expression:

`value_if_true if condition else value_if_false`

This is sometimes called the ternary operator, though Python's syntax is different from languages like Java, C, or JavaScript.

```python
# Teaching point:
# Conditional expressions are useful for simple choices.
# Avoid making them too long or nested.

records_loaded = 0

status = "SUCCESS" if records_loaded > 0 else "EMPTY_FILE"

print(status)
```

## 16. Walrus Operator `:=`

The walrus operator assigns a value as part of an expression.

It can reduce repeated work, but it should be used carefully because it can make code harder to read if overused.

```python
# Teaching point:
# The walrus operator lets us assign and test in one expression.

records = ["row1", "row2", "row3"]

if (count := len(records)) > 0:
    print("Records found:", count)
else:
    print("No records found")
```

## 17. Operator Overloading

In Python, operators can behave differently depending on the object types.

For example, `+` adds numbers, joins strings, and combines lists. This is possible because Python objects define how operators should work for them.

```python
# Teaching point:
# Same operator, different behavior based on type.

print(10 + 20)
print("data" + " engineering")
print(["csv", "json"] + ["parquet"])

# This would fail because Python does not know how to add int and str directly:
# print(10 + "20")
```

## 18. Operators with Strings

Some arithmetic-looking operators also work with strings.

- `+` concatenates strings
- `*` repeats strings
- `in` checks substring membership

```python
# Teaching point:
# Operators can be useful for basic text processing.

prefix = "orders"
suffix = "2026"
file_name = prefix + "_" + suffix + ".csv"

print(file_name)
print("-" * 30)
print("orders" in file_name)
```

## 19. Operators with Lists

Lists support several operators:

- `+` combines lists
- `*` repeats lists
- `in` checks whether an item exists

```python
# Teaching point:
# List operators are useful, but be careful with repetition and nested mutable objects later.

batch_1 = ["order_1", "order_2"]
batch_2 = ["order_3"]

all_orders = batch_1 + batch_2

print(all_orders)
print(batch_2 * 3)
print("order_2" in all_orders)
```

## 20. Operator Precedence

Operator precedence decides which operation happens first.

For example, multiplication happens before addition.

When in doubt, use parentheses. Parentheses make code easier to read and reduce mistakes.

```python
# Teaching point:
# Parentheses make the intended order clear.

result_1 = 10 + 5 * 2
result_2 = (10 + 5) * 2

print(result_1)
print(result_2)

quality_check = (95 >= 90) and (1000 > 0)
print(quality_check)
```

## 21. Common Precedence Order

A simplified precedence order from higher to lower:

- Parentheses: `()`
- Exponentiation: `**`
- Unary operators: `+x`, `-x`, `~x`
- Multiplication, division, floor division, modulo: `*`, `/`, `//`, `%`
- Addition and subtraction: `+`, `-`
- Bit shifts: `<<`, `>>`
- Bitwise AND: `&`
- Bitwise XOR: `^`
- Bitwise OR: `|`
- Comparisons, membership, identity: `==`, `<`, `in`, `is`, etc.
- Logical NOT: `not`
- Logical AND: `and`
- Logical OR: `or`
- Conditional expression: `x if condition else y`
- Assignment expression: `:=`

## 22. Practice Example: Data Load Decision

This example combines arithmetic, comparison, logical, membership, and conditional operators.

```python
# Teaching point:
# Real validation logic often combines many operator categories.

required_columns = {"order_id", "customer_id", "amount"}
actual_columns = {"order_id", "customer_id", "amount", "order_date"}

expected_records = 1000
actual_records = 980

schema_ok = required_columns <= actual_columns
volume_ok = actual_records >= expected_records * 0.95
load_status = "ACCEPT" if schema_ok and volume_ok else "REJECT"

print("Schema OK:", schema_ok)
print("Volume OK:", volume_ok)
print("Load status:", load_status)
```
