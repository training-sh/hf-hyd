# Python Type Casting

Type casting means converting a value from one type to another.

In data engineering, type casting is everywhere:

- CSV values arrive as strings
- JSON fields may contain mixed types
- Database columns need exact types
- API payloads need validation
- Dates, numbers, booleans, and missing values need careful conversion

This notebook covers common type casting combinations in Python.

## 1. Basic Casting Functions

Python provides built-in functions for common conversions:

- `int()`
- `float()`
- `str()`
- `bool()`
- `list()`
- `tuple()`
- `set()`
- `dict()`
- `bytes()`

Some conversions work naturally. Some fail. Some work but may surprise beginners.

```python

# Casting creates a value of a target type when Python knows how to convert it.

print(int("100"))
print(float("99.5"))
print(str(250))
print(bool(1))
print(list("ABC"))
```

## 2. Check Type Before and After Casting

Always inspect values when learning type casting.

`type()` helps us confirm what happened.

```python

# The value may look similar after casting, but the type can be different.

raw_value = "100"
converted_value = int(raw_value)

print(raw_value, type(raw_value))
print(converted_value, type(converted_value))
```

## 3. String to Integer

`int()` converts a valid whole-number string into an integer.

This is common when reading CSV files or environment variables.

```python

# A string must contain a valid whole number for int() to work.

raw_count = "500"
count = int(raw_count)

print(count)
print(type(count))
```

## 4. String to Float

`float()` converts a valid decimal-number string into a floating point number.

```python

# Decimal-looking strings can be converted with float().

raw_amount = "1250.75"
amount = float(raw_amount)

print(amount)
print(type(amount))
```

## 5. Float String to Integer: Two-step Conversion

`int("10.5")` fails because `"10.5"` is not a whole-number string.

If needed, convert to `float` first, then to `int`. Remember: converting `float` to `int` removes the decimal part.

```python

# int() truncates toward zero when converting from float.

raw_value = "10.9"
number = int(float(raw_value))

print(number)
print(type(number))
```

## 6. Integer to Float and Float to Integer

`float()` can convert integers to floats.

`int()` can convert floats to integers, but it truncates the decimal part. It does not round.

```python

# int(99.99) becomes 99, not 100.

print(float(10), type(float(10)))
print(int(99.99), type(int(99.99)))
print(int(-99.99), type(int(-99.99)))
```

## 7. Rounding Before Integer Conversion

If you want rounding, use `round()` before or instead of `int()`.

```python

# int() truncates. round() rounds.

value = 99.99

print("int:", int(value))
print("round:", round(value))
print("int after round:", int(round(value)))
```

## 8. Number to String

`str()` converts numbers into text.

This is useful when building file names, log messages, SQL text, and API payloads.

```python

# Use str() when combining text with numbers.

batch_id = 25
file_name = "batch_" + str(batch_id) + ".csv"

print(file_name)
```

## 9. Boolean Casting Rules

`bool()` converts values into `True` or `False`.

Common false-like values:

- `False`
- `0`
- `0.0`
- `""` empty string
- `[]` empty list
- `{}` empty dictionary
- `set()` empty set
- `None`

Most other values become `True`.

```python

# Empty values are usually False. Non-empty values are usually True.

values = [False, True, 0, 1, 0.0, 0.5, "", "False", [], [1], {}, {"a": 1}, None]

for value in values:
    print(repr(value), "=>", bool(value))
```

## 10. Warning: `bool("False")` Is `True`

This is a common mistake.

Any non-empty string becomes `True`, even the string `"False"`.

```python

# bool() checks emptiness, not the meaning of the text.

print(bool("False"))
print(bool("0"))
print(bool("no"))
print(bool(""))
```

## 11. Safe String to Boolean

When converting strings to booleans, write explicit logic.

This is common with config files, environment variables, and API inputs.

```python

# Convert text to boolean using clear allowed values.

def parse_bool(value):
    normalized = value.strip().lower()

    if normalized in {"true", "yes", "y", "1"}:
        return True
    if normalized in {"false", "no", "n", "0"}:
        return False

    raise ValueError("Invalid boolean text: " + value)

print(parse_bool("true"))
print(parse_bool("No"))
print(parse_bool(" 1 "))
```

## 12. List, Tuple, and Set Casting

`list()`, `tuple()`, and `set()` can convert iterable values.

An iterable is something Python can loop over.

```python

# Strings, lists, tuples, sets, and dictionaries are iterable.

text = "ABC"
items = ["csv", "json", "csv"]

print(list(text))
print(tuple(items))
print(set(items))
```

## 13. String to List: Characters vs Split

`list("csv,json")` gives individual characters.

If you want fields, use `.split()`.

```python

# list() and split() solve different problems.

raw_columns = "order_id,customer_id,amount"

print(list(raw_columns))
print(raw_columns.split(","))
```

## 14. List to String

Use `join()` to combine a list of strings into one string.

`str(list_value)` creates a display representation, not a clean delimited string.

```python

# join() is the standard way to combine strings.

columns = ["order_id", "customer_id", "amount"]

print(str(columns))
print(",".join(columns))
```

## 15. Tuple to List and List to Tuple

Tuples are immutable. Lists are mutable.

Convert between them when you need different behavior.

```python

# list() makes a mutable list. tuple() makes an immutable tuple.

schema_tuple = ("order_id", "customer_id")
schema_list = list(schema_tuple)
schema_list.append("amount")

print(schema_tuple)
print(schema_list)
print(tuple(schema_list))
```

## 16. List to Set for Removing Duplicates

`set()` removes duplicate values because a set stores unique elements.

Important: sets are unordered collections.

```python

# set() removes duplicates, but order is not guaranteed.

raw_sources = ["api", "csv", "api", "database", "csv"]
unique_sources = set(raw_sources)

print(unique_sources)
print(list(unique_sources))
```

## 17. Preserve Order While Removing Duplicates

If order matters, use `dict.fromkeys()` as a common Python technique.

```python

# dict.fromkeys() removes duplicates while preserving first-seen order.

raw_sources = ["api", "csv", "api", "database", "csv"]
unique_ordered_sources = list(dict.fromkeys(raw_sources))

print(unique_ordered_sources)
```

## 18. Dictionary Casting

`dict()` can create a dictionary from key-value pairs.

Each pair must contain exactly two items: key and value.

```python

# A list of two-item tuples can be converted into a dictionary.

pairs = [("source", "api"), ("active", True), ("batch_size", 500)]
config = dict(pairs)

print(config)
print(type(config))
```

## 19. Dictionary to List, Tuple, and Set

When you cast a dictionary directly to `list`, `tuple`, or `set`, Python uses the dictionary keys.

Use `.items()`, `.values()`, or `.keys()` when you want to be explicit.

```python

# Direct iteration over a dictionary gives keys.

record = {"order_id": 101, "amount": 250.5, "status": "PAID"}

print(list(record))
print(tuple(record))
print(set(record))
print(list(record.items()))
print(list(record.values()))
```

## 20. Bytes and String Casting

`str` is text.

`bytes` is raw binary data.

Use `.encode()` to convert text to bytes. Use `.decode()` to convert bytes to text.

```python

# Encoding and decoding require an encoding name, commonly utf-8.

text = "orders"
raw_bytes = text.encode("utf-8")
decoded_text = raw_bytes.decode("utf-8")

print(text, type(text))
print(raw_bytes, type(raw_bytes))
print(decoded_text, type(decoded_text))
```

## 21. `bytes()` from Integers

`bytes([65, 66, 67])` creates bytes from integer byte values.

Each integer must be between `0` and `255`.

```python

# 65, 66, and 67 are byte values for A, B, and C in ASCII/UTF-8.

raw = bytes([65, 66, 67])

print(raw)
print(raw.decode("utf-8"))
```

## 22. Decimal Casting

`Decimal` is useful for exact decimal arithmetic, especially money.

Prefer creating `Decimal` from strings, not floats.

```python

# Decimal from string avoids importing binary float approximation.

from decimal import Decimal

amount = Decimal("99.95")
tax = Decimal("5.25")

print(amount + tax)
print(type(amount))
```

## 23. Date and Time Casting

Dates often arrive as strings.

Use `datetime.strptime()` to parse a string into a datetime object.

Use `.strftime()` to format a datetime object back into a string.

```python

# Date parsing requires a format string matching the input.

from datetime import datetime

raw_date = "2026-07-09"
parsed_date = datetime.strptime(raw_date, "%Y-%m-%d")
formatted_date = parsed_date.strftime("%d-%b-%Y")

print(raw_date, type(raw_date))
print(parsed_date, type(parsed_date))
print(formatted_date, type(formatted_date))
```

## 24. JSON Casting: String to Python Object

JSON data often arrives as a string from files, APIs, or message queues.

`json.loads()` converts JSON text into Python objects.

`json.dumps()` converts Python objects into JSON text.

```python

# JSON objects become Python dictionaries.
# JSON arrays become Python lists.

import json

raw_json = '{"order_id": 101, "amount": 250.75, "paid": true}'
record = json.loads(raw_json)
json_text = json.dumps(record)

print(record, type(record))
print(json_text, type(json_text))
```

## 25. Handling Failed Casts

Casting can fail when the value is not valid for the target type.

Use `try` and `except` to handle bad data gracefully.

```python

# Never assume external data is clean.

raw_value = "N/A"

try:
    amount = float(raw_value)
    print("Amount:", amount)
except ValueError:
    print("Invalid numeric value:", raw_value)
```

## 26. Safe Integer Parser

In pipelines, it is useful to write small helper functions for repeated conversion logic.

```python

# A parser can return None when conversion fails.
# This makes downstream validation easier.

def parse_int(value):
    try:
        return int(value)
    except (TypeError, ValueError):
        return None

print(parse_int("100"))
print(parse_int("10.5"))
print(parse_int(None))
```

## 27. Casting a Data Record

This is a realistic data engineering example.

A raw record may arrive with everything as strings. We cast fields into useful Python types.

```python

# Data pipelines often normalize raw text into typed fields.

raw_record = {
    "order_id": "101",
    "amount": "250.75",
    "paid": "true",
    "order_date": "2026-07-09",
}

clean_record = {
    "order_id": int(raw_record["order_id"]),
    "amount": Decimal(raw_record["amount"]),
    "paid": parse_bool(raw_record["paid"]),
    "order_date": datetime.strptime(raw_record["order_date"], "%Y-%m-%d").date(),
}

print(clean_record)

for key, value in clean_record.items():
    print(key, "=>", value, "=>", type(value))
```

## 28. Common Casting Matrix

This table is a quick reference for common conversions.

| From | To | Example | Notes |
|---|---|---|---|
| `str` | `int` | `int("10")` | String must be whole number |
| `str` | `float` | `float("10.5")` | String must be numeric |
| `str` | `bool` | custom parser | Do not use `bool("False")` |
| `int` | `str` | `str(10)` | Useful for messages/file names |
| `float` | `int` | `int(10.9)` | Truncates, does not round |
| `int` | `float` | `float(10)` | Adds decimal representation |
| `list` | `tuple` | `tuple([1, 2])` | Mutable to immutable |
| `tuple` | `list` | `list((1, 2))` | Immutable to mutable |
| `list` | `set` | `set([1, 1, 2])` | Removes duplicates |
| `set` | `list` | `list({1, 2})` | Order not guaranteed |
| `dict` | `list` | `list(record)` | Gives keys |
| `pairs` | `dict` | `dict([("a", 1)])` | Needs key-value pairs |
| `str` | `bytes` | `text.encode("utf-8")` | Text to binary |
| `bytes` | `str` | `raw.decode("utf-8")` | Binary to text |
| `str` | `datetime` | `datetime.strptime(...)` | Needs format |
| `dict/list` | JSON string | `json.dumps(...)` | Python to JSON |
| JSON string | `dict/list` | `json.loads(...)` | JSON to Python |

## Quick Recap

- Type casting converts values from one type to another.
- `int()`, `float()`, `str()`, and `bool()` are common scalar casts.
- `list()`, `tuple()`, `set()`, and `dict()` are common collection casts.
- `bool("False")` is `True` because the string is non-empty.
- Use explicit logic for string-to-boolean conversion.
- Use `Decimal` for exact decimal values such as money.
- Use `encode()` and `decode()` for text/bytes conversion.
- Use `try` and `except` when casting external data.
- Real data pipelines should cast fields deliberately and validate failures.
